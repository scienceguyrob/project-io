Pandas is the most widely used Python library for data analysis and is the standard tool for loading, inspecting, cleaning, and transforming tabular data before it is fed into a machine learning model. ### 4.1 Series

Before diving into Pandas, let's recap: an **array** is an ordered collection of elements, and in Python, a **NumPy array (`np.array`)** is a fixed-size, homogeneous (same data type) array that supports efficient numerical operations.

A **Series** in Pandas is a one-dimensional labelled array. It builds on NumPy arrays but adds a critical feature: **every element has an index label** in addition to its positional index. This means you can look up values by name (like a Python dictionary) while retaining the ordered, array-like properties of a list. A Series can hold any Python data type, integers, floats, strings, or even functions, making it incredibly flexible.

In the code below, we'll create Series objects from different data structures (lists, NumPy arrays, and dictionaries) to see how Pandas handles indexing and labels. Read the code and then run it to see how it behaves.

````python
# Listing 2.
# --- Creating a Series from different data structures ---
# pd.Series() is a flexible constructor that can create a Series from:
# - Python lists
# - NumPy arrays
# - Python dictionaries
# When no index is provided, Pandas assigns a default integer index (0, 1, 2, ...)

# Define our input data
labels = ['a', 'b', 'c']  # Custom index labels
my_list = [10, 20, 30]     # Python list of values
arr = np.array([10, 20, 30])  # Equivalent NumPy array
d = {'a': 10, 'b': 20, 'c': 30}  # Dictionary with label:value pairs

# Example 1: From a list with default integer index
s1 = pd.Series(data=my_list)
print("Series from list (default index):")
print(s1)  # Index will be 0, 1, 2
print()

# Example 2: From a list with custom index labels
s2 = pd.Series(data=my_list, index=labels)
print("Series from list (labelled index):")
print(s2)  # Index will be 'a', 'b', 'c'
print()

# Example 3: From a NumPy array (behaves like a list)
s3 = pd.Series(arr)
print("Series from NumPy array:")
print(s3)  # Same output as s1 - default integer index
print()

# Example 4: From a dictionary
# Note: Dictionary keys automatically become the index labels
s4 = pd.Series(d)
print("Series from dictionary:")
print(s4)  # Index will be 'a', 'b', 'c' from dictionary keys
````

The **index** is the heart of a Series. It is the label associated with each element, enabling fast look-ups and automatic alignment during operations.

Pandas uses this index not only for quick label-based access but also to align two Series on their shared index when performing arithmetic. This alignment is what makes combining data from different sources so straightforward, as operations automatically match up corresponding labels.

In the code below, we create two Series representing quarterly sales for different sets of countries. Notice how Pandas handles label-based look-ups and index alignment when performing operations between them, in this case adding two Series together:


````python
# Listing 3.
# --- Using the index for label-based look-up and arithmetic ---
# Here we create two quarterly sales Series for different country sets.
sales_Q1 = pd.Series(data=[250, 450, 200, 150],
                     index=['USA', 'China', 'India', 'Brazil'])

sales_Q2 = pd.Series(data=[260, 500, 210, 100],
                     index=['USA', 'China', 'India', 'Japan'])  # Japan replaces Brazil

print("Q1 Sales:")
print(sales_Q1)
print()

print("Q2 Sales:")
print(sales_Q2)
print()

# Look up a single value by index label
print("Q1 sales for USA:", sales_Q1['USA'])
print()

# Adding the two Series: Pandas aligns on the shared index.
# 'Brazil' and 'Japan' have no match in the other Series, so they become NaN.
# NaN stands for "Not a Number" — Pandas' way of representing missing data.
print("Q1 + Q2 (aligned on index — note NaN for unmatched countries):")
print(sales_Q1 + sales_Q2)
````

In this example, we create two Series, `sales_Q1` and `sales_Q2`, representing quarterly sales for different sets of countries. Notice that while both Series share some countries (`USA`, `China`, `India`), each also includes a unique country (`Brazil` in `sales_Q1` and `Japan` in `sales_Q2`).

When we look up a value by its label (e.g., `sales_Q1['USA']`), Pandas returns the corresponding value (`250`). This demonstrates how the index enables **fast, label-based access** to data.

The real power of the index becomes clear when we add the two Series together. Pandas **aligns the data by index labels**, performing element-wise addition only where the labels match. For countries that don't appear in both Series (`Brazil` and `Japan`), the result is `NaN` (Not a Number), which Pandas uses to represent missing data. This automatic alignment ensures that operations are both intuitive and accurate, even when working with incomplete or mismatched datasets.

### 4.2 DataFrames

A **DataFrame** is a two-dimensional, table-like data structure. Recall from Notebook 3 that a **2-D array (or matrix)** is a rectangular grid of scalars organised into rows and columns. 

A DataFrame builds on this concept: internally, it is a collection of Series objects (1-D arrays with labels) that share the same index. Each **column** in a DataFrame behaves like a Series, while each **row** is a labelled observation across all columns. The entire structure forms a labelled matrix of data, with both rows and columns explicitly named.

The code below demonstrates how to create a DataFrame from a NumPy array, with explicit row and column labels. It uses `np.random.randint` to generate random integers for the data (the data itself doesn't matter). The resulting DataFrame is printed, along with its shape (number of rows and columns) and the names of its columns and index.


````python
# Listing 4.
# --- Creating a DataFrame ---
# np.random.seed() fixes the random number generator so results are reproducible.
# Every time this cell is run with the same seed, the same numbers are produced.
np.random.seed(42)

columns = ['W', 'X', 'Y', 'Z']       # column labels
index   = ['A', 'B', 'C', 'D', 'E']  # row labels

# np.random.randint(low, high, size) generates an array of random integers
# in the range [low, high) with the given shape — here 5 rows × 4 columns.
data = np.random.randint(-100, 100, (5, 4))

df = pd.DataFrame(data, index=index, columns=columns)

print("Our DataFrame:")
print(df)
print()
print(f"Shape: {df.shape}   (rows × columns)")
print(f"Column names: {list(df.columns)}")
print(f"Index names:  {list(df.index)}")
````

### 4.3 Selection and Indexing

Pandas provides several ways to select subsets of a DataFrame, and choosing the right method matters because they behave differently and produce different output types. The most fundamental operation is **column selection**, which can return either a Series or a DataFrame depending on how you specify the columns.

Here's a quick reference for the methods we'll explore:

| Method | Selects by | Returns |
|--------|-----------|---------|
| `df['col']` | Column name | Series |
| `df.loc[row_label, col_label]` | Row and/or column *labels* | DataFrame or Series |
| `df.iloc[row_pos, col_pos]` | Row and/or column *integer positions* | DataFrame or Series |

The code below demonstrates how to select columns from a DataFrame. Notice that selecting a single column by name (e.g., `df['W']`) returns a **Series**, while passing a list of column names (e.g., `df[['W', 'Z']]`) returns a **DataFrame**:

````python
# Listing 5.
# --- Column selection ---
# Selecting a single column by name returns a Series.
print("Single column 'W' (returns a Series):")
print(df['W'])
print()

# Passing a list of column names returns a DataFrame.
print("Multiple columns ['W', 'Z'] (returns a DataFrame):")
print(df[['W', 'Z']])
print()

# Confirm the types
print(f"type(df['W'])        → {type(df['W'])}")
print(f"type(df[['W','Z']]) → {type(df[['W','Z']])}")
````

#### 4.3.1 Adding and Removing Columns in a DataFrame

The code below demonstrates how to **add and remove columns** in a DataFrame. New columns can be created by simple assignment, where the right-hand side expression is applied element-wise across the rows.

To remove columns, use `df.drop()`, specifying `axis=1` to indicate that columns (rather than rows) should be removed. By default, `drop()` does **not** directly modify the original DataFrame. Instead it returns a new DataFrame with the specified columns removed. To update the original DataFrame, you must either:
- Reassign the result back to the original variable (e.g., `df = df.drop('column_name', axis=1)`), or
- Use the `inplace=True` parameter (e.g., `df.drop('column_name', axis=1, inplace=True)`), which modifies the DataFrame directly.

Read the code below then run it to see how it behaves.

````python
# Listing 6.
# --- Creating and removing columns ---
# New columns can be added by assigning a value or expression 
# to a new column name. The right-hand side is applied 
# element-wise across all rows.
df['new'] = df['W'] + df['Y']  # Creates a new column 'new' with the sum of columns 'W' and 'Y'
print("DataFrame after adding a 'new' column (sum of 'W' and 'Y'):")
print(df)
print()

# To remove columns (or rows), use df.drop().
# By default, drop() does not modify the original DataFrame; it returns a new 
# DataFrame with the specified column/row removed. To update the original DataFrame,
# you must either:
#   1. Reassign the result: df = df.drop('column_name', axis=1), or
#   2. Use the inplace=True parameter: df.drop('column_name', axis=1, inplace=True)
df = df.drop('new', axis=1)  # axis=1 specifies that we are dropping a column (not a row)
print("DataFrame after dropping the 'new' column (original DataFrame is now updated):")
print(df)
````

#### 4.3.2 Row Selection with loc - Label-Based

While column selection uses square brackets (`df['col']`), row selection requires the `.loc` accessor. The `.loc` method is **label-based**, meaning it selects rows and columns using their **index or column names** (not their integer positions). This allows you to retrieve specific rows, multiple rows, or even a subset of both rows and columns in a single operation.

The code below demonstrates how to use `.loc` to:
- Select a single row by its label.
- Select multiple rows by their labels.
- Select specific rows *and* columns simultaneously.

````python
# Listing 7.
# --- Row selection with .loc (label-based) ---
# .loc selects rows and columns using their index/column labels,
# not their integer positions.

# Select a single row by its label ('A').
# Returns a Series with the row's data.
print("Single row by label — df.loc['A'] (returns a Series):")
print(df.loc['A'])
print()

# Select multiple rows by their labels ('A' and 'C').
# Returns a DataFrame with the selected rows.
print("Multiple rows by label — df.loc[['A', 'C']] (returns a DataFrame):")
print(df.loc[['A', 'C']])
print()

# Select a subset of rows AND columns simultaneously.
# Syntax: df.loc[row_labels, column_labels]
# Returns a DataFrame with the selected rows and columns.
print("Rows A and C, columns W and Y — df.loc[['A', 'C'], ['W', 'Y']]:")
print(df.loc[['A', 'C'], ['W', 'Y']])
````

#### 4.3.3 Row Selection with iloc - Position-Based

While `.loc` uses labels, `.iloc` is **position-based**, meaning it selects rows and columns using their **integer positions** (0-based), regardless of their index labels. This is similar to how you access elements in a Python list or NumPy array.

The code below demonstrates how to use `.iloc` to:
- Select a single row by its integer position.
- Select a range of rows using slicing (note that the upper bound is exclusive, just like Python lists).

Additionally, we'll see how to **remove rows** using `df.drop()`. By default, `drop()` does not modify the original DataFrame, so we'll print the result of dropping a row and then confirm that the original DataFrame remains unchanged.

````python
# Listing 8.
# --- Row selection with .iloc (integer position-based) ---
# .iloc selects rows and columns using their integer positions (0-based),
# regardless of their index labels.

# Select a single row by its integer position (0).
# Returns a Series with the row's data.
print("Single row at position 0 — df.iloc[0] (returns a Series):")
print(df.iloc[0])
print()

# Select a range of rows using slicing (positions 0 and 1).
# Note: The upper bound (2) is exclusive, so this selects rows 0 and 1.
# Returns a DataFrame with the selected rows.
print("Rows at positions 0 and 1 — df.iloc[0:2] (upper bound is exclusive):")
print(df.iloc[0:2])
print()

# --- Removing rows ---
# Use df.drop() to remove a row by its label.
# By default, drop() does not modify the original DataFrame.
print("DataFrame after dropping row 'C' (not in place, original df unchanged):")
print(df.drop('C', axis=0))
print()

# Confirm that the original DataFrame remains unchanged.
print("Original df is unchanged:")
print(df)
````

### 4.4 Conditional Selection

One of the most useful Pandas features is **Boolean filtering**: selecting rows based on a condition. This works similarly to NumPy Boolean indexing (see Notebook 3), but operates on DataFrame rows using column values as the condition.

Boolean filtering allows you to create a **mask** (a DataFrame of `True`/`False` values) based on a condition, and then use that mask to filter the original DataFrame. Run the code below to see how this works in practice:


````python
# Listing 9.
# --- Boolean filtering ---
# First, let's see the original DataFrame
print("Original DataFrame:")
print(df)
print()

# df > 0 returns a DataFrame of True/False values
print("df > 0 (Boolean mask):")
print(df > 0)
print()

# Passing the Boolean mask back into df returns only the cells that are True;
# cells that are False are replaced with NaN
print("df[df > 0] — only positive values retained:")
print(df[df > 0])
````

Building on the Boolean mask concept, we can also filter **entire rows** based on conditions applied to specific columns. When you apply a condition to a column (e.g., `df['X'] > 0`), it returns a Boolean Series where each element corresponds to a row in the DataFrame. Passing this Boolean Series to the DataFrame filters it to only include rows where the condition is `True`. The code below demonstrates how to:
- Filter rows where a specific column meets a condition.
- Select specific columns from the filtered result.

---

````python
# Listing 10.
# --- Filtering rows based on a column condition ---
# First, let's see the original DataFrame
print("Original DataFrame:")
print(df)
print()

# df['X'] > 0 returns a Boolean Series (one True/False per row)
# Passing it to df selects only the rows where the condition is True
print("Rows where column X > 0:")
print(df[df['X'] > 0])
print()

# We can then select specific columns from the filtered result
print("Column Y from rows where X > 0:")
print(df[df['X'] > 0]['Y'])
print()

print("Columns Y and Z from rows where X > 0:")
print(df[df['X'] > 0][['Y', 'Z']])
````

You can also combine multiple conditions using logical operators (`&` for **and**, `|` for **or**). Each condition must be wrapped in parentheses due to Python's operator precedence rules. This allows you to create complex filters, such as selecting rows where multiple conditions are met simultaneously, or where at least one of several conditions is true.

The code below demonstrates how to:
- Use `&` to filter rows where **both** conditions are `True`.
- Use `|` to filter rows where **either** condition is `True`.


````python
# Listing 11.
# --- Multiple conditions ---
# First, let's see the original DataFrame
print("Original DataFrame:")
print(df)
print()

# Use & (and) or | (or) to combine conditions
# Each condition MUST be wrapped in parentheses because of Python's operator precedence
print("Rows where W > 0 AND Y > 1:")
print(df[(df['W'] > 0) & (df['Y'] > 1)])
print()

# Example with OR — rows where W < 0 or Z > 0
print("Rows where W < 0 OR Z > 0:")
print(df[(df['W'] < 0) | (df['Z'] > 0)])
````

### 4.7 Handling Missing Data

In real-world datasets, missing values are almost unavoidable. A sensor might fail to record a reading, a survey respondent might skip a question, or a database merge might leave some rows incomplete. Before any analysis or modelling can take place, these gaps must be addressed: most machine learning algorithms cannot accept missing values as input and will either crash or produce incorrect results if they are present.

Pandas represents missing data as `NaN` (Not a Number), a special floating-point value inherited from NumPy. You can think of `NaN` as a placeholder that says "there should be a value here, but we don't have one". It is distinct from zero (which is a real value) and from an empty string (which is a real string); `NaN` specifically means *absence of data*.

There are three main strategies for dealing with missing values, and the right choice depends on why the data is missing and how much of it is missing:

- **Dropping missing values**: remove any row or column that contains a `NaN`. This is the simplest approach but should only be used when the missing data is a small fraction of the total and there is no systematic reason for it being absent. Dropping too many rows can introduce bias or leave you with too little data to train a model.
- **Filling missing values (imputation)**: replace `NaN` with a substitute value such as the column mean, median, or a fixed constant. This preserves the size of the dataset but introduces an assumption, that the substitute value is a reasonable stand-in for the true missing value. The mean is commonly used for roughly symmetric distributions; the median is preferred when there are outliers, since it is more robust to extreme values.
- **Interpolation**: estimate the missing value from its neighbours in the dataset. This is particularly useful for time-series data where values change gradually over time and a missing entry can be reasonably estimated from the entries immediately before and after it.

The code below creates a small DataFrame with intentional `NaN` values to illustrate how Pandas represents missing data. Columns A and B each have one or more missing entries, while column C is complete. This controlled example makes it easy to see exactly which cells are affected before exploring the methods above.


````python
# Listing 15.
# --- Create a small DataFrame with missing values ---
# np.nan is NumPy's missing value sentinel. Pandas treats it as "no data here".
df_missing = pd.DataFrame({
    'A': [1,   2,      np.nan, 4],
    'B': [5,   np.nan, np.nan, 8],
    'C': [10,  20,     30,     40]
})

print("DataFrame with missing values (NaN):")
print(df_missing)
````

#### 4.7.1 Dropping Missing Data

One straightforward approach to handling missing values is to **drop** rows or columns that contain them. Pandas provides the `dropna()` method for this purpose, with flexible options to control how much data is removed:

- By default, `dropna()` removes **rows** containing any `NaN` values.
- Using `axis=1` removes **columns** with missing values instead.
- The `thresh` parameter allows you to specify a minimum number of non-`NaN` values required for a row to be retained.

The code below demonstrates these options in practice, showing how to clean a DataFrame by removing incomplete rows or columns.

````python
# Listing 16.
# --- Dropping missing data ---

# df.dropna() removes any ROW that contains at least one NaN.
print("df.dropna()  — drop rows with any NaN:")
print(df_missing.dropna())
print()

# df.dropna(axis=1) removes any COLUMN that contains at least one NaN.
print("df.dropna(axis=1)  — drop columns with any NaN:")
print(df_missing.dropna(axis=1))
print()

# df.dropna(thresh=n) keeps only rows with at least n non-NaN values.
# thresh=2 means: keep rows that have at least 2 real values.
print("df.dropna(thresh=2)  — keep rows with at least 2 non-NaN values:")
print(df_missing.dropna(thresh=2))
````

#### 4.7.2 Filling Missing Data

Instead of dropping missing values, you can **fill** them with substitute values. This approach preserves the structure of your DataFrame and is often preferable when missing data is limited or follows a pattern. Pandas' `fillna()` method offers several ways to replace `NaN` values:

- Replace all `NaN` values with a **constant** (e.g., a placeholder string or zero).
- Use **statistical measures** like the mean, median, or mode to fill missing values, which is especially useful for numeric data.
- Apply fills **column-wise** or to the **entire DataFrame** at once.

The code below demonstrates these strategies, showing how to replace `NaN` with placeholders, zeros, or column-specific means.

````python
# Listing 17.
# --- Filling missing data ---
# First, let's see the DataFrame with missing values
print("Original DataFrame with missing values:")
print(df_missing)
print()

# df.fillna(value) replaces every NaN with the given value
print("Filling all NaN values with 'FILL VALUE':")
print(df_missing.fillna(value='FILL VALUE'))
print()

# Fill a single column's NaNs with zero
print("Filling NaN values in column 'A' with 0:")
print(df_missing['A'].fillna(0))
print()

# Fill with the column mean — a common strategy when NaN values are few
# and the data is roughly normally distributed
col_a_mean = df_missing['A'].mean()  # mean() ignores NaN by default
print(f"Mean of column 'A' (ignoring NaN): {col_a_mean}")
print("Filling NaN values in column 'A' with its mean:")
print(df_missing['A'].fillna(col_a_mean))
print()

# Fill ALL numeric columns with their respective column means in one step
print("Filling NaN values in all numeric columns with their respective means:")
print(df_missing.fillna(df_missing.mean(numeric_only=True)))
````

### 4.8 GroupBy

The `groupby()` method is one of the most powerful tools in Pandas. It allows you to split a DataFrame into groups based on one or more categorical columns, apply an aggregate function to each group (such as `mean`, `sum`, `count`, or `std`), and combine the results into a new DataFrame.

This follows the **Split–Apply–Combine** pattern that is standard in data analysis:

1. **Split** the data into groups based on a key column, for example, grouping all rows that share the same `Sector` value together
2. **Apply** an aggregation or transformation to each group independently, for example, computing the mean number of completions within each sector
3. **Combine** the results back into a single table where each row represents one group

This pattern is extremely common in practice. Rather than writing a loop that manually filters the DataFrame for each unique value in a column, `groupby()` handles all of that internally and returns a clean summary in one line of code. It is the Pandas equivalent of a SQL `GROUP BY` clause, and once you are comfortable with it you will find yourself reaching for it constantly.

The aggregate functions most commonly used with `groupby()` are:

| Function | What it computes |
|---|---|
| `.mean()` | Average value within each group |
| `.sum()` | Total value within each group |
| `.count()` | Number of non-null entries within each group |
| `.std()` | Standard deviation within each group |
| `.min()` / `.max()` | Smallest / largest value within each group |

We will use the Universities dataset here to see `groupby()` on real data. As a reminder, the dataset has the following structure:

| Column | Description | Data Type |
|---|---|---|
| `Sector` | The sector of the institution (e.g. `"Private for-profit, 2-year"`) | String (categorical) |
| `University` | The name of the institution | String |
| `Year` | The academic year of the record | Integer |
| `Completions` | The number of programme completions reported | Integer |
| `Geography` | The US state in which the institution is located | String (categorical) |

The `Sector` and `Geography` columns are both categorical: they take a limited set of repeating string values, which makes them natural candidates for grouping. The code below loads the dataset and inspects its first few rows before applying `groupby()` to answer questions like *"which sector has the highest average completions?"* and *"how does completion volume vary by US state?"*

````python
# Listing 18.
# --- Load the Universities dataset ---
# pd.read_csv() reads a CSV file from disk into a DataFrame.
# The file must be in the 'data' subfolder relative to this notebook.
df_uni = pd.read_csv('data/Universities.csv')

# .head(n) returns the first n rows (5 by default).
# Always inspect the first few rows of a new dataset before doing anything else.
print("First 5 rows of the Universities dataset:")
print(df_uni.head())
print()
print(f"Shape: {df_uni.shape}   ({df_uni.shape[0]} rows × {df_uni.shape[1]} columns)")
````

With the dataset loaded and the structure confirmed, we can now apply `groupby()`. The first example is deliberately simple: grouping by `Year` and computing the mean number of completions within each year. This answers the question *"has the average number of completions per institution changed over time?"* using just one line of code.

Before running it, it is worth understanding what `groupby()` actually returns before an aggregation is applied. Calling `df_uni.groupby('Year')` alone does not produce a table; it creates a **GroupBy object**, which is essentially a set of instructions telling Pandas how the data is split. Nothing is computed until you chain an aggregation method such as `.mean()` onto it. This lazy evaluation is intentional: it allows Pandas to be efficient, only doing the computation once you have specified exactly what you want.

The double bracket notation `[['Completions']]` selects a single column but keeps the result as a **DataFrame** rather than collapsing it into a **Series**. The difference matters: a DataFrame has column labels and is easier to work with in subsequent steps, while a Series is a single unnamed column of values. As a rule of thumb, use double brackets whenever you want to keep the output as a DataFrame.

````python
# Listing 19.
# --- Basic groupby: mean completions per year ---
# Step 1: df_uni.groupby('Year') creates a GroupBy object — a kind of "pending" split.
#         Nothing is computed yet; Pandas is waiting for an aggregation instruction.
# Step 2: [['Completions']].mean() applies the mean() aggregation to the Completions column.
#         The double brackets [[ ]] keep the result as a DataFrame rather than a Series.

print("Mean completions per year:")
print(df_uni.groupby('Year')[['Completions']].mean())
print()

# type() confirms what we get back
print("Return type:", type(df_uni.groupby('Year')[['Completions']].mean()))
````

The result of a `groupby()` aggregation is sorted by the group key in ascending order by default, in this case years from earliest to most recent. That is often not the most useful ordering. Chaining `.sort_index(ascending=False)` onto the result reverses this, placing the most recent year at the top.

Note that we are sorting by the **index** here rather than by a column. After a `groupby()`, the grouping key (`Year`) becomes the index of the resulting DataFrame rather than a regular column, so `sort_index()` is the correct method. If we wanted to sort by the values in the `Completions` column instead, for example to find the year with the highest average completions, we would use `.sort_values('Completions', ascending=False)` instead.


````python
# Listing 20.
# --- Sorting the grouped result ---
# .sort_index(ascending=False) sorts the resulting DataFrame by the index (Year)
# in descending order — most recent year first.
print("Mean completions per year (most recent first):")
print(df_uni.groupby('Year')[['Completions']].mean().sort_index(ascending=False))
````


Rather than computing a single statistic like the mean, it is often useful to get a full summary of each group at once. The code below chains `.describe()` onto a `groupby()` call, which computes count, mean, standard deviation, minimum, maximum, and the 25th, 50th, and 75th percentiles for every numeric column within each group, all in one step.

The raw output of `groupby().describe()` can be wide and difficult to read when there are many statistics and only a few groups. Chaining `.transpose()` flips the rows and columns so that the statistical measures become the row labels and the years become the column headers. This is purely a presentation choice, the underlying data is identical, but the transposed version is often much easier to scan when the number of groups is small relative to the number of statistics.


````python
# Listing 21.
# --- More aggregate functions ---
# .describe() applied to a grouped DataFrame computes count, mean, std, min, max,
# and percentiles for every numeric column within each group.
print("Describe completions by year:")
print(df_uni.groupby('Year').describe())
print()

# .transpose() flips rows and columns — useful when there are many statistics
# and only a few groups, making the table easier to read.
print("Transposed (statistics as rows, years as columns):")
print(df_uni.groupby('Year').describe().transpose())
````

So far we have grouped by a single column at a time. Pandas also supports grouping by multiple columns simultaneously by passing a list of column names to `groupby()`. The code below groups by both `Geography` and `Year` together, so each group represents one unique state-and-year combination. Summing `Completions` within each group answers the question *"how many total completions were recorded in each state in each year?"*

The result has a **MultiIndex**, a hierarchical index with one level for `Geography` and one for `Year`. This is why `.head(10)` is used here rather than printing the full table, which would run to hundreds of rows. When working with multi-level grouped results, `.head()` is a useful sanity check to confirm the structure looks as expected before processing the full output.

````python
# Listing 22.
# --- Groupby with multiple columns ---
# We can group by more than one column at a time. Here we compute the total
# completions per Sector and Year combination.
print("Total completions grouped by Geography and Year:")
print(df_uni.groupby(['Geography', 'Year'])[['Completions']].sum().head(10))
````

The full list of Pandas aggregate functions you can call after `groupby()` includes:

| Function | Description |
|----------|-------------|
| `count()` | Number of non-null observations per group |
| `sum()` | Sum of values |
| `mean()` | Mean of values |
| `median()` | Median value |
| `min()` / `max()` | Minimum / maximum value |
| `std()` / `var()` | Standard deviation / variance |
| `describe()` | All of the above in one call |
| `cumsum()` | Cumulative sum |
| `quantile(q)` | Value at the q-th percentile |

Beyond aggregation with `groupby`, Pandas offers powerful tools for **inspecting, transforming, and recoding** data at the column level. These operations are essential for data cleaning, feature engineering, and exploratory analysis. This section covers three techniques: identifying the **unique values** in a column, **applying** custom functions to transform data row by row, and **mapping** values from one set to another using a dictionary or function.

The code below creates a sample DataFrame to demonstrate these operations in practice.

````python
# Listing 23.
# --- Create a sample DataFrame to work with ---
df_one = pd.DataFrame({
    'k1':   ['A', 'A', 'B', 'B', 'C', 'C'],
    'col1': [100, 200, 300, 300, 400, 500],
    'col2': ['NY', 'CA', 'WA', 'WA', 'AK', 'NV']
})

print("Sample DataFrame:")
print(df_one)
````

Understanding the distinct values in a column is often the first step in exploratory data analysis. Pandas provides methods to inspect the uniqueness and distribution of values in a column.

The `.unique()` method returns all distinct values in a column, while `.nunique()` gives the count of unique values. For a deeper look, `.value_counts()` returns the frequency of each value, sorted by count, which is particularly useful for categorical data.

The code below demonstrates these methods on the `col2` column of our sample DataFrame.

The code below demonstrates how to use this method to clean a DataFrame by removing duplicate entries.

````python
# Listing 25.
# --- Removing duplicate rows ---
# .drop_duplicates() returns a DataFrame with duplicate rows removed.
# "Duplicate" means every column value is identical across two rows.
print("df_one.drop_duplicates()  — rows WA/WA removed:")
print(df_one.drop_duplicates())
````

You can create new columns in a DataFrame by performing **element-wise arithmetic operations** on existing columns. This is a common way to derive new features or transform data for analysis.

The code below demonstrates how to add a new column (`New Col`) by multiplying each value in `col1` by 10.

````python
# Listing 26.
# --- Creating new columns via arithmetic ---
# The operation is applied element-wise across every row.
df_one['New Col'] = df_one['col1'] * 10
print("After adding 'New Col' = col1 × 10:")
print(df_one)
````

The `.apply()` method allows you to apply a **custom function** to each element in a column, enabling flexible transformations that go beyond built-in operations. This is particularly useful for cleaning, recoding, or extracting information from strings, dates, or other complex data types.

The process involves:
1. Defining a function that takes a single value and returns a transformed result.
2. Passing the function (without parentheses) to `.apply()`, which applies it to every element in the column.

The code below demonstrates how to use `.apply()` to extract the first letter from each state code in `col2` and add the result as a new column.

````python
# Listing 27.
# --- Applying a custom function with .apply() ---
# .apply(func) calls func once for every value in the column.
# The return value of func becomes the new element in the resulting Series.

# Step 1: Define the function that will operate on each value.
def grab_first_letter(state):
    """Given a two-letter state code, return the first letter."""
    return state[0]

# First, let's see the original DataFrame
print("Original DataFrame:")
print(df_one)
print()

# Step 2: Pass the FUNCTION OBJECT (no parentheses) to .apply().
# Pandas calls grab_first_letter(value) for each row.
print("First letter of each state code in 'col2':")
print(df_one['col2'].apply(grab_first_letter))
print()

# Save the result as a new column
df_one['first_letter'] = df_one['col2'].apply(grab_first_letter)
print("DataFrame after adding the 'first_letter' column:")
print(df_one)
````

The `.apply()` method can also handle **conditional logic**, allowing you to create new columns based on custom rules. This is useful for categorising data, flagging specific values, or recoding columns based on complex criteria.

The code below demonstrates how to use `.apply()` with a function that checks if a state code starts with `'W'` and returns `'Washington'` if true, or `'Other'` otherwise. The result is stored in a new column, `state_check`.

**Note:** When using `.apply()`, ensure your function is compatible with the data type of the column. For example, indexing with `state_code[0]` works for strings but will raise a `TypeError` if applied to numeric columns.

````python
# Listing 28.
# --- A more complex function with conditional logic ---
# First, let's see the original DataFrame
print("Original DataFrame:")
print(df_one)
print()

def flag_washington_states(state_code):
    """Return the full name for 'W' states, or 'Other' otherwise."""
    if state_code[0] == 'W':
        return 'Washington'
    else:
        return 'Other'

print("Applying flag_washington_states to 'col2' to identify Washington states:")
df_one['state_check'] = df_one['col2'].apply(flag_washington_states)
print(df_one)
print()

# ⚠️ Watch out for data type errors!
# The function tries to index into the value with state_code[0], which works for
# strings but will raise a TypeError if applied to a numeric column.
# For example, the following would fail:
#   df_one['col1'].apply(flag_washington_states)   # TypeError: 'int' is not subscriptable
````

### 4.9.1 Mapping

In Python, a **dictionary** is a collection of key-value pairs, where each key maps to a specific value. The `.map()` method in Pandas leverages this concept to transform values in a Series. When you use `.map()` with a dictionary, it checks each value in the Series against the dictionary's keys. If a match is found, the value is replaced with the corresponding dictionary value. If no match is found, the result is `NaN`. This is a powerful way to recode or rename values in your data efficiently.

Categorical data often needs to be converted into numeric form for machine learning models. The `.map()` method allows you to perform this conversion by substituting each category with a corresponding numeric code. This process, known as **encoding**, ensures that categorical variables can be properly interpreted by algorithms that require numerical input.

The code below demonstrates how to use `.map()` to replace the categorical labels `'A'`, `'B'`, and `'C'` in the `k1` column with numeric codes `1`, `2`, and `3`, respectively.

````python
# Listing 29.
# --- .map() for value substitution ---
# Here we convert the category labels 'A', 'B', 'C' into numeric codes.
# This kind of encoding is sometimes required before feeding categorical data into an ML model.
print("k1 column mapped to numeric codes:")
print(df_one['k1'].map({'A': 1, 'B': 2, 'C': 3}))
````

### 4.10 Locating Min and Max Values

## 5. ML Models as Parameterised Mapping Functions

🔙 [Back to Table of Contents](#Table-of-Contents)

Now that you are comfortable loading and manipulating data with Pandas, we turn to the mathematical ideas that explain what machine learning algorithms are actually doing. The three concepts in this section, parameterised functions, loss landscapes, and optimisation, are the conceptual skeleton of every ML algorithm you will encounter.

### 5.1 Functions as Mappings

In mathematics, a **function** is a rule that takes one or more inputs and produces exactly one output. You already know how to write these in Python:

```python
def f(x):
    return 2 * x + 1
```

Feed in $x = 3$, get back $7$. The function *maps* input 3 to output 7.

**Machine learning models are functions.** They take input data (features) and map it to an output (a prediction). The difference from the simple example above is:

1. The inputs are *vectors*: many features at once
2. The function has **parameters** (written as $\theta$, theta) that are tuned during training
3. "Training" means searching for the $\theta$ values that make the function's outputs match the true labels as closely as possible

Formally, a machine learning model is a **parameterised mapping function**:

$$
\hat{y} = f_{\theta}(X)
$$

where:
- $\hat{y}$ (pronounced "y-hat") is the **predicted output**: the model's best guess at the correct answer
- $X$ is the **input feature matrix**: the data we feed into the model (e.g. fruit diameters)
- $\theta$ (theta) holds all of the model's **tunable parameters**: the internal numbers that the model adjusts during training to improve its predictions

A simple concrete example: in the threshold classifier from Notebook 2, the single parameter $\theta$ was the threshold value $k$. The mapping function was:

$$
\hat{y} = \begin{cases} 1 & \text{if } x \geq k \\ 0 & \text{if } x < k \end{cases}
$$

Training that model meant scanning candidate values of $k$ and picking the one that produced the fewest wrong predictions. Every machine learning algorithm follows this same fundamental pattern; the difference lies in how complex $f_\theta$ is and how $\theta$ is optimised.

The goal of training is always to find the $\theta$ that minimises some **loss function**: a mathematical measure of how wrong the predictions are on average. You have already encountered a loss function in Notebooks 1 and 2: the misclassification error rate, which simply counts the fraction of predictions that do not match the true label. Different model types use different loss functions, but the principle is always the same: training is the process of adjusting $\theta$ to make the loss as small as possible.

### 5.2 A Complete Classifier from Scratch

To make these ideas concrete, let's build and train a complete classifier from scratch. We will use a simple one-parameter model so that every step is transparent.

**The problem:** We have measured the diameter (cm) of a collection of fruit, some apples, some oranges. We want a function that maps diameter to fruit type.

The model we'll make is a **threshold classifier**: predict Orange if diameter $\geq \theta$, else predict Apple. $\theta$ is the single parameter we need to find to split the data in two.

**Step 1: Generate the fruit data**

The cell below generates synthetic fruit diameter measurements and plots them. The data are generated from two normal distributions with different means: apples centred around 7.5 cm and oranges around 9.5 cm. Using synthetic data here is deliberate. It allows us to control exactly how the data was generated, so we know the ground truth and can verify that our model is behaving correctly. Notice that the two distributions overlap slightly: some apples are unusually large and some oranges are unusually small. No threshold will give zero error, but we can find the *best* one.

````python
# Listing 37.
# --- Step 1: Generate synthetic fruit data ---

# np.random.default_rng(seed) creates a random number generator with a fixed seed,
# ensuring the data is the same every time this cell runs.
rng = np.random.default_rng(0)

# rng.normal(loc, scale, size) draws 'size' samples from a normal distribution
# with the given mean (loc) and standard deviation (scale).
apple_diameters  = rng.normal(loc=7.5, scale=0.8, size=30)   # Apples: smaller
orange_diameters = rng.normal(loc=9.5, scale=0.9, size=30)   # Oranges: larger

# Combine into a single feature array x and a label array y.
# np.concatenate joins two arrays end-to-end along a given axis.
x = np.concatenate([apple_diameters, orange_diameters])
y = np.array([0] * 30 + [1] * 30)    # 0 = Apple, 1 = Orange

%matplotlib widget
from visualisations.Figure_1 import show
show()
````

**Step 2: Define `find_theta` and `predict`**

The two functions in the code below implement the complete training and inference pipeline. Read through them carefully: every line maps directly to the formal concepts introduced above.

**`find_theta`: training**

This function implements exhaustive search (also called a grid search) over all candidate threshold values between `start` and `end`. For each candidate threshold $\theta$, it applies the mapping function to every sample and counts how many predictions are wrong. The threshold that produces the fewest errors is returned as the trained parameter.

Each concept from the formal framework maps directly to a line of code:

- The `while` loop is the **search**: it steps through every candidate $\theta$ from `start` to `end` in increments of `step`, evaluating the loss at each one
- The `if/else` inside the loop is the **mapping function** $f_\theta(x)$: it applies the current threshold to a single sample and produces a predicted label
- `min_errors` tracks the **loss**, specifically the number of misclassifications, which is the loss function we minimise during training
- `theta_log` records the loss at every candidate threshold, which we will use shortly to plot the full error landscape

The function returns three things: the best threshold found, the number of errors at that threshold, and the full log of every (threshold, error) pair evaluated during the search.

**`predict`: inference**

Once training is complete, `predict` applies the learned threshold $\theta$ to a new, unseen measurement. This is inference: the parameters are fixed and we are simply evaluating $f_\theta(x)$ for a new input. Notice how much simpler inference is than training: training requires scanning hundreds of candidate values, while inference is a single comparison.

After defining both functions, the final three lines run the training process and print the result: the optimal threshold, the number of misclassifications at that threshold, and the resulting training accuracy.

When ready, read the code, then run it. How does it perform on the data created earlier?

````python
# Listing 38.
# --- Step 2: Define the classifier ---

def find_theta(x, y, start=5.0, end=12.0, step=0.05):
    """
    Search for the best threshold theta by exhaustive search.

    The model predicts label=1 (Orange) if feature >= theta, else label=0 (Apple).
    Every candidate threshold between 'start' and 'end' is evaluated;
    the one with the fewest misclassifications is returned.

    Parameters
    ----------
    x     : array-like, feature values (one per sample)
    y     : array-like, true labels (0 or 1)
    start : float, smallest candidate threshold to try
    end   : float, largest candidate threshold to try
    step  : float, gap between successive candidate thresholds

    Returns
    -------
    best_theta : float, threshold with fewest errors
    min_errors : int, number of misclassifications at best_theta
    theta_log  : list of (theta, error_count) tuples for every candidate
    """
    best_theta = None
    min_errors = float('inf')   # float('inf') = infinity; any real count will be smaller
    theta_log  = []             # records every (theta, error_count) pair for plotting

    theta = start

    while theta <= end:
        errors = 0
        for feature, label in zip(x, y):
            prediction = 1 if feature >= theta else 0  # the mapping function
            if prediction != label:
                errors += 1
        theta_log.append((theta, errors))
        if errors < min_errors:
            best_theta = theta
            min_errors = errors
        theta += step

    return best_theta, min_errors, theta_log


def predict(x_new, theta):
    """
    Apply the trained threshold classifier to a new measurement.

    Parameters
    ----------
    x_new : float, the new feature value (diameter in cm)
    theta : float, the trained decision threshold

    Returns
    -------
    str, 'Apple' or 'Orange'
    """
    if x_new < theta:
        return 'Apple'
    else:
        return 'Orange'


# --- Train: find the best theta ---
best_theta, min_errors, theta_log = find_theta(x, y)

print(f"Best threshold:    theta = {best_theta:.2f} cm")
print(f"Minimum errors:    {min_errors} out of {len(x)}")
print(f"Training accuracy: {1 - min_errors/len(x):.1%}")
````

**Step 3: Visualise the error landscape**

When the `find_theta` function finished running in the code above, it returned three values which Python unpacked into three separate variables on this line:

```python
best_theta, min_errors, theta_log = find_theta(x, y)
```

You have already seen `best_theta` (the optimal threshold) and `min_errors` (how many fruits were misclassified at that threshold) printed in the output. The third variable, `theta_log`, was not printed, but it has been quietly stored in memory and we are about to use it.

`theta_log` is a Python list built up inside `find_theta` by this line:

```python
theta_log.append((theta, errors))
```

Every time the `while` loop tested a new candidate threshold, it added a small tuple, a pair of values in round brackets, to this list. Each tuple contains two numbers: the threshold value that was tested, and the number of misclassifications that resulted. After the loop finished, `theta_log` contains one tuple for every candidate threshold that was evaluated: several hundred entries in total, stepping from 5.0 cm to 12.0 cm in increments of 0.05 cm. For example, the first few entries might look like:

```python
[(5.0, 30), (5.05, 30), (5.1, 30), ..., (8.55, 4), ..., (12.0, 30)]
```

Each tuple is saying: "when I used this threshold, this many fruits were misclassified." The entry `(8.3, 3)` means that at a threshold of 8.3 cm, only 3 fruits were predicted incorrectly: the best result found during the entire search.

In other words, `theta_log` is a complete record of the entire search: every threshold tried, and how badly the model performed at each one. If we plot it with the candidate threshold value on the x-axis and the misclassification error rate on the y-axis, we get the **error landscape**: a curve showing how the loss function changes as $\theta$ moves across the parameter space. Each point on the curve answers the question "if I had chosen *this* threshold, how many fruits would have been misclassified?"

This is one of the most important visualisations in machine learning because it makes optimisation concrete and visual. Rather than just being told "the best threshold is 8.3 cm", you can *see* why: the error curve has a clear minimum at that point, and moving the boundary in either direction increases the number of misclassifications.

The code below produces two panels:

- **Left panel**: the full error landscape, with the x-axis showing every candidate threshold and the y-axis showing the resulting error rate. The red dashed line marks the optimal $\theta$, the lowest point of the curve. Notice that the curve is not perfectly smooth: because we are counting discrete misclassifications rather than computing a continuous quantity, it has a stepped appearance.
- **Right panel**: the raw fruit data from Figure 1 with the optimal decision boundary overlaid as a vertical dashed line. The shaded green region to the left is where the model predicts Apple; the shaded orange region to the right is where it predicts Orange. Points that fall on the wrong side of the boundary are the unavoidable errors, fruits whose diameters happen to fall in the overlap region between the two classes.

Together the two panels tell the complete story of training: the left shows the search, the right shows the result.

````python
# Listing 39.
# --- Step 3: Plot the error landscape and the optimal boundary ---

# Unpack the log into separate lists for plotting.
thetas      = [t for t, e in theta_log]
errors_abs  = [e for t, e in theta_log]
total       = len(x)
error_rates = [e / total for e in errors_abs]  # convert counts to rates (0–1)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
fig.canvas.header_visible = False
fig.canvas.resizable = True
fig.canvas.toolbar_visible = False

# --- Left panel: error landscape ---
ax = axes[0]
ax.plot(thetas, error_rates, color='steelblue', linewidth=2, label='Error rate')
# ax.axvline() draws a vertical line at a fixed x position.
# linestyle='--' makes it dashed; zorder controls layering (higher = on top).
ax.axvline(best_theta, color='red', linewidth=2, linestyle='--',
           label=f'Best theta = {best_theta:.2f} cm  ({min_errors/total:.1%} error)')
ax.set_xlabel('Theta (cm)')
ax.set_ylabel('Misclassification error rate')
ax.set_title('Error landscape: how error changes with theta')
ax.legend()
ax.grid(True, alpha=0.3)

# --- Right panel: data with the optimal decision boundary overlaid ---
ax = axes[1]
ax.scatter(apple_diameters,  [0] * 30, color='green',  s=70, alpha=0.7,
           edgecolors='k', lw=0.4, label='Apple', zorder=3)
ax.scatter(orange_diameters, [0] * 30, color='orange', s=70, alpha=0.7,
           edgecolors='k', lw=0.4, label='Orange', zorder=3)
ax.axvline(best_theta, color='red', linewidth=2.5, linestyle='--',
           label=f'Boundary: {best_theta:.2f} cm')
# ax.fill_betweenx() shades a rectangular region between two x values.
# alpha controls transparency (0=invisible, 1=solid).
ax.fill_betweenx([-0.3, 0.3], 5,          best_theta, alpha=0.08, color='green',  label='Predict Apple')
ax.fill_betweenx([-0.3, 0.3], best_theta, 13,         alpha=0.08, color='orange', label='Predict Orange')
ax.set_xlabel('Diameter (cm)')
ax.set_yticks([])
ax.set_xlim(5, 13)
ax.set_title(f'Optimal boundary | Accuracy = {1 - min_errors/total:.1%}')
ax.legend(fontsize=8)

plt.suptitle('Figure 2: Error landscape (left) and optimal decision boundary (right)', fontsize=11)
plt.tight_layout()
plt.show()
````

**Step 4: Apply the trained model to new measurements**

With `best_theta` found and stored, the hard work of training is complete. We can now classify any new fruit diameter by calling `predict()`: this is inference, applying the fixed, trained parameter $\theta$ to unseen data.

The code below tests the classifier on six new diameter measurements that were not part of the training data. For each one it calls `predict(d, best_theta)` and also computes how far the measurement sits from the decision boundary. This distance is important: a fruit with a diameter very close to `best_theta` is in the overlap region where the two classes are hardest to separate, so we should be less confident in the prediction. A fruit far from the boundary, clearly on one side or the other, can be classified with much more certainty.

This idea of **confidence** based on distance from the boundary is a concept that will reappear in more sophisticated models in later notebooks. For now it is computed simply as `abs(d - best_theta)` and translated into a plain-English label.


````python
# Listing 40.
# --- Step 4: Inference on new data ---
test_cases = [6.5, 7.8, 8.5, 9.0, 10.2, 11.0]

print(f"Trained threshold: theta = {best_theta:.2f} cm")
print()
print(f"{'Diameter (cm)':<18} {'Prediction':<12} {'Confidence'}")
print('-' * 48)

for d in test_cases:
    pred     = predict(d, best_theta)
    distance = abs(d - best_theta)
    # Closeness to the boundary determines how confident we should be.
    if distance > 1.5:
        conf = 'very certain'
    elif distance > 0.8:
        conf = 'certain'
    else:
        conf = 'borderline'
    print(f"  {d:<16.1f} {pred:<12} {conf}")
````


**What we just built is a complete ML pipeline.** Every machine learning algorithm, from logistic regression to a neural network, follows exactly this same four-step structure: generate or load data, define a model with tunable parameters, search for the parameter values that minimise a loss function, and apply the trained model to new inputs.

The differences between algorithms lie in the complexity of $f_\theta$ (a threshold is about as simple as it gets), the number of parameters in $\theta$ (we had just one; a large neural network may have billions), and the sophistication of the search (we used exhaustive grid search; more advanced algorithms use calculus-based techniques like gradient descent to find the minimum far more efficiently). But the underlying logic, train then predict, is universal.


| Step | What happened | ML term |
|------|--------------|---------|
| Generate data | Created `x` (features) and `y` (labels) | Training set |
| Define the model | `prediction = 1 if feature >= theta else 0` | Model architecture |
| Define the loss | Count of misclassifications | Loss function |
| Search for best θ | `while` loop over candidate values | Training / optimisation |
| Apply to new data | `predict(x_new, theta)` | Inference |


### 5.3 Local vs Global Minima

Our `find_theta` function performs an **exhaustive search**: it evaluates every possible threshold and picks the best one. This works well here because we only have a single parameter $\theta$, which means the search space is just a one-dimensional line that can be scanned from one end to the other in a fraction of a second.

But this approach does not scale. If our model had two parameters, we would need to evaluate every combination of both, a two-dimensional grid. With ten parameters the search space has ten dimensions, and the number of evaluations required grows exponentially. This is known as the **curse of dimensionality**: the size of the parameter space explodes as the number of parameters increases, making exhaustive search completely impractical for any real model. A neural network with millions of parameters cannot be trained by trying every possible combination; a smarter search strategy is needed.

A second problem arises even when the search space is manageable: real loss functions are rarely as clean as the one we plotted in Figure 2. Instead of a single smooth valley with one clear minimum, they often have **multiple valleys of different depths**. A search algorithm that is not careful can descend into a shallow valley and stop, believing it has found the best solution, when a deeper valley, a better set of parameters, exists elsewhere in the parameter space.

These two types of minimum have specific names:
- A **local minimum** is a point where the loss is lower than at any nearby point, but not necessarily lower than at all points globally. It is a valley, but not the deepest one.
- The **global minimum** is the single lowest point across the entire parameter space: the optimal solution we are actually trying to find.

Getting trapped in a local minimum is one of the central challenges of training machine learning models, and a great deal of research in the field is devoted to designing optimisation algorithms that are less likely to get stuck. The cell below creates a synthetic multi-valley loss function to visualise both of these concepts directly.

The code below constructs a synthetic loss function with multiple valleys. It is not tied to any specific model, but it is representative of the kind of complex landscape that a real optimiser must navigate. Two panels are produced:

- **Left panel**: the full loss landscape with both the local minimum (orange dot) and the global minimum (red star) marked. Study this panel first and convince yourself that the orange dot is genuinely a valley (the loss rises on both sides of it) but that the red star is deeper. An algorithm that starts its search near the orange dot and always moves downhill will never escape to find the red star.

- **Right panel**: a simulation of a simple gradient descent algorithm starting from a point on the left side of the landscape. Gradient descent (we'll come back to this soon!) is one of the most widely used optimisation algorithms in machine learning: at each step it estimates the slope of the loss function at the current position and takes a small step in the downhill direction. The orange dotted path shows the sequence of steps taken. Notice that the algorithm descends efficiently, but arrives at the local minimum and stops there, never discovering that a better solution exists further to the right.

This simulation is deliberately simplified: the step size (called the **learning rate**) is fixed, and the gradient is estimated numerically rather than computed analytically. In practice, gradient descent is more sophisticated, but the core idea, and the core vulnerability to local minima, is exactly the same.

The right panel illustrates why the **starting point** matters: a search algorithm that begins in the wrong region can converge to a local minimum and never escape.

| Term | Meaning |
|------|---------|
| **Loss landscape** | The shape of the loss function across all possible parameter values |
| **Local minimum** | A valley that is not the deepest — a search algorithm can get stuck here |
| **Global minimum** | The deepest valley — the truly best parameter values |
| **Gradient descent** | A widely used algorithm that moves "downhill" step by step — vulnerable to local minima |
| **Exhaustive search** | Try every value — guaranteed global minimum, but only practical for 1–2 parameters |

For our simple one-dimensional fruit problem the landscape is smooth and exhaustive search trivially finds the global minimum. For a neural network with millions of parameters, the landscape has millions of dimensions, so exhaustive search is completely impractical, which is why gradient descent and its variants are used instead.

### 5.4 Step Size - Precision vs Speed

Recall the `step` parameter in `find_theta` (a few cells up): it controls the gap between successive candidate thresholds evaluated during the search. Choosing the right step size involves a fundamental trade-off that appears in virtually every optimisation algorithm in machine learning:

- **Large step**: the search moves quickly across the parameter space, requiring few evaluations. But it risks jumping over the true optimum entirely: if the best threshold sits between two evaluation points, the search will never find it exactly and will settle for the nearest evaluated value instead.
- **Small step**: the search is slower and requires many more evaluations, but the result is more precise: the found threshold will be closer to the true optimum.

This trade-off does not disappear when you move to more sophisticated models, it just changes form. In gradient-based optimisation, the standard approach used to train neural networks and most modern machine learning models (more on this shortly!), the equivalent of step size is called the **learning rate**. A learning rate that is too large causes the optimiser to overshoot the minimum and bounce around without converging; a learning rate that is too small causes the optimiser to converge correctly but painfully slowly, requiring many more training steps. Choosing the right learning rate is one of the first and most important decisions you will make when tuning a real model, and it is one of the reasons that optimisation in deep learning is as much an art as it is a science.


### 5.5 Gradient Descent - Bonus

So far, `find_theta` has found the optimal threshold by brute force, trying every possible value in a fixed grid and keeping the best one. This works when there is only one parameter, but as we noted in section 5.3, it becomes completely impractical as the number of parameters grows. A neural network might have millions of parameters; evaluating every possible combination is not just slow, it is mathematically impossible.

The algorithm that makes training large models feasible is called **gradient descent**. It is the engine behind almost every modern machine learning system. Understanding it from first principles, which is what this section does, will give you genuine insight into what is happening inside a model during training, rather than just treating it as a black box.

The core idea is simple, and you already have an intuition for it from everyday life.

Imagine you are standing somewhere on a hilly landscape in thick fog. You cannot see the whole landscape; you can only feel the ground immediately beneath your feet. Your goal is to reach the lowest point in the valley. How do you get there?

You feel which direction the ground slopes downward most steeply, and you take a step in that direction. Then you stop, feel the slope again, and take another step. You repeat this process until the ground feels flat in every direction, at which point you have (hopefully) reached the bottom.

This is gradient descent. The "landscape" is the loss function, the surface that measures how wrong the model's predictions are. The "slope" is the **gradient**, a mathematical measure of how steeply the loss changes as each parameter changes. And each "step" is a small adjustment to the model's parameters in the direction that reduces the loss most quickly.

Before we go any further, it is worth being precise about what "slope" means, because this idea underpins everything that follows.

#### What is a slope

You have encountered slopes before: on a road sign warning of a steep hill, or when drawing a straight line on a graph in school. The slope of a line is defined as:

$$\text{slope} = \frac{\text{rise}}{\text{run}} = \frac{\text{how much the y-value changes}}{\text{how much the x-value changes}}$$

For a straight line this is easy: pick any two points, measure how much you go up (rise) and how much you go across (run), and divide. A steep upward line has a large positive slope; a steep downward line has a large negative slope; a flat horizontal line has a slope of zero.

For a curve like a loss function, the slope is not constant; it is different at every point. The slope at a specific point on a curve is defined as the slope of the straight line that just touches the curve at that point without crossing it. This line is called the **tangent line**. You will see it drawn in the interactive figures below.

Gradient descent uses the slope at the current position to decide which direction to step and how far to go. A steep slope means a large step; a shallow slope means a small step. When the slope is exactly zero, when the tangent line is perfectly flat, the algorithm stops, because that means we are at the bottom of the valley.

The figure below illustrates both of these ideas side by side. The left panel shows a straight line with the rise and run clearly labelled. You can read off the slope directly as rise divided by run, and it is the same no matter which two points on the line you choose. The right panel shows a curve, where the slope is different at every point. At the specific point marked, a tangent line is drawn; the slope of that tangent line is the slope of the curve at that exact location, measured using the same rise-over-run idea.

Notice that on the curve, the tangent line would have a different slope if you moved the point left or right. This is the key difference between a straight line and a curve: on a line, one slope value describes the whole thing; on a curve, every point has its own slope. Gradient descent re-measures this slope at every step, which is why it can navigate toward the minimum even on a complex, winding loss surface.


### 5.5.1 The Update Rule

Every step of gradient descent is described by a single equation:

$$
\theta \leftarrow \theta - \alpha \cdot \frac{dL}{d\theta}
$$

This looks intimidating but you have already seen exactly what it does in the visualisation above. Let's break it down symbol by symbol:

- $\theta$ (theta): the current parameter value. In our fruit classifier this was the threshold; in a neural network there are millions of these.
- $\alpha$ (alpha): the **learning rate**. This is the slider you just experimented with. It controls how large each step is. A small $\alpha$ means cautious, small steps; a large $\alpha$ means big, aggressive steps that risk overshooting the minimum.
- $\frac{dL}{d\theta}$: the **gradient**. We will explain this in detail below.
- $\cdot$: means multiplication. The gradient is multiplied by the learning rate before being subtracted.
- $\leftarrow$: means "is updated to". The new value of $\theta$ replaces the old one.

Put together, the equation says: **take the current parameter value, subtract the gradient multiplied by the learning rate, and use the result as the new parameter value**.

In code, this single equation looks like this:

```python
theta = theta - learning_rate * gradient
```

That is the entirety of gradient descent. Everything else, momentum, Adam, RMSProp, and the other optimisers you will encounter, is a variation or refinement of this one line.

---

### 5.5.2 What is the Gradient

The notation $\frac{dL}{d\theta}$ comes from calculus and is called a **derivative**. For students without a calculus background, the important thing is not the notation itself but what it *means*:

> **The gradient tells you the slope of the loss curve at your current position, and therefore which direction is downhill.**

Think back to the hilly landscape analogy. If you are standing on a hillside, the slope of the ground beneath your feet tells you two things: which direction is downhill, and how steep the hill is. The gradient $\frac{dL}{d\theta}$ gives exactly this information for the loss function.

More precisely, the gradient measures **how much the loss $L$ changes when you make a tiny change to $\theta$**. The $d$ notation is mathematical shorthand for "a tiny change in", so $\frac{dL}{d\theta}$ literally means "the change in loss divided by the change in $\theta$", for an infinitesimally small change. You do not need to compute this by hand; in practice, software libraries compute it automatically.

**The sign of the gradient tells you which direction to step:**

| Gradient sign | Meaning | Action taken |
|---|---|---|
| **Positive** ($\frac{dL}{d\theta} > 0$) | The loss curve slopes upward to the right — increasing $\theta$ would increase the loss | Subtract a positive number → $\theta$ moves **left** (downhill) |
| **Negative** ($\frac{dL}{d\theta} < 0$) | The loss curve slopes downward to the right — increasing $\theta$ would decrease the loss | Subtract a negative number → $\theta$ moves **right** (also downhill) |
| **Zero** ($\frac{dL}{d\theta} = 0$) | The curve is flat — we are at a minimum (or maximum, or saddle point) | No movement — the algorithm has converged |

In both the positive and negative cases, subtracting the gradient always moves $\theta$ in the downhill direction. This is not a coincidence; it is precisely why the minus sign appears in the update rule.

**The magnitude of the gradient tells you how steep the slope is:**

A large gradient (far from zero) means the loss is changing rapidly: we are on a steep part of the curve and a single step will produce a large improvement. A small gradient (close to zero) means the loss is nearly flat: we are close to the minimum and steps produce diminishing returns. You can see this in the printed trace from Figure 5: the loss decreases by a large amount in the early steps, then by smaller and smaller amounts as the algorithm approaches the bottom.

### 5.5.4 The Gradient of this Loss Function

Now, the gradient at any position is a measure of how steeply the loss is changing at that exact point. For this particular loss function, there is a formula that tells you the gradient at *any* position: you just plug in your current value of $\theta$ to get the actual slope at that point:

$$\text{gradient at } \theta = 2\theta$$

This is not saying the gradient is always the same fixed number; it is saying that for this problem, "whatever position you are currently at, multiply it by 2 and that gives you the slope of the curve at that exact point." For example:

- At $\theta = 3$: gradient $= 2 \times 3 = 6$, the curve is steeply sloping here
- At $\theta = 1$: gradient $= 2 \times 1 = 2$, less steep
- At $\theta = 0.1$: gradient $= 2 \times 0.1 = 0.2$, nearly flat
- At $\theta = 0$: gradient $= 2 \times 0 = 0$, perfectly flat, this is the minimum

Here is an intuitive way to see where the $2\theta$ formula comes from, without needing any formal calculus. The core idea is: if we give $\theta$ a tiny nudge of size $h$, how much does the loss change?

**Step 1: Compute the loss after nudging $\theta$ by $h$**

We substitute $(\theta + h)$ in place of $\theta$ in the loss formula $L(\theta) = \theta^2$. This just means asking: "what is the loss if our parameter is slightly larger than it currently is?"

$$L(\theta + h) = (\theta + h)^2$$

Now we expand $(\theta + h)^2$. The rule for expanding two brackets multiplied together is: multiply every term in the first bracket by every term in the second bracket, then add the results:

$$(\theta + h)^2 = (\theta + h) \times (\theta + h)$$

$$= \underbrace{\theta \times \theta}_{\theta^2} + \underbrace{\theta \times h}_{\theta h} + \underbrace{h \times \theta}_{\theta h} + \underbrace{h \times h}_{h^2}$$

$$= \theta^2 + \theta h + \theta h + h^2$$

Since $\theta h$ appears twice, we combine those terms:

$$= \theta^2 + 2\theta h + h^2$$

So after the nudge, the new loss is $\theta^2 + 2\theta h + h^2$.

**Step 2: Find how much the loss changed**

The change in loss is simply the new loss minus the old loss, the same way you would calculate how much your bank balance changed by subtracting what you started with from what you ended with:

$$\text{change in loss} = L(\theta + h) - L(\theta) = (\theta^2 + 2\theta h + h^2) - \theta^2$$

The two $\theta^2$ terms cancel (a positive $\theta^2$ and a negative $\theta^2$ sum to zero), leaving:

$$= 2\theta h + h^2$$

**Step 3: Find the rate of change per unit of nudge**

A total change depends on how big the nudge was; a larger nudge produces a larger change, even on a flat surface. To get a fair measure of steepness we divide the change in loss by the size of the nudge $h$, giving us the change *per unit of nudge*:

$$\frac{2\theta h + h^2}{h}$$

We apply the rule that $\frac{a + b}{c} = \frac{a}{c} + \frac{b}{c}$ to split this into two separate fractions:

$$= \frac{2\theta h}{h} + \frac{h^2}{h}$$

In the first fraction, $h$ appears in both the numerator and denominator, so they cancel, just as $\frac{6}{6} = 1$. In the second fraction, $\frac{h^2}{h} = h$ because $h^2 = h \times h$ and one $h$ cancels:

$$= 2\theta + h$$

**Step 4: Make $h$ vanishingly small**

The nudge $h$ was always supposed to be tiny, just a small test to probe the slope. As we make $h$ smaller and smaller (imagine $h = 0.001$, then $0.0001$, then $0.000001$...), the $h$ term at the end contributes less and less to the result and eventually becomes negligible:

$$2\theta + \underbrace{h}_{\approx 0} \approx 2\theta$$

We are left with $2\theta$, the gradient. This is the rate at which the loss changes per unit change in $\theta$, at the exact point where you currently are. The further you are from the minimum, the larger $\theta$ is, and therefore the larger the gradient, which is exactly why gradient descent takes bigger steps when it is far away and smaller steps as it homes in on the answer.

The figure below makes this process visible and interactive. It shows the loss curve $L(\theta) = \theta^2$ with two key lines drawn at your chosen position (chosen using sliders):

- **The red dashed line**: the secant line, which joins two points on the curve, your current position $\theta$ and the nudged position $\theta + h$. Its slope is the approximation $2\theta + h$, what we get when $h$ is still a real measurable size rather than vanishingly small.

The dotted grey lines show the rise (how much the loss changes) and the run (the size of the nudge $h$), making the "rise over run" idea of slope concrete.

### 5.5.5 A Worked Numerical Example

Now that we know the gradient formula for our toy loss function is $2\theta$ (and hopefully we understand where the $2\theta$ comes from!), we can substitute it directly into the update rule:

$$\theta \leftarrow \theta - \alpha \cdot \frac{dL}{d\theta} \quad \text{becomes} \quad \theta \leftarrow \theta - \alpha \cdot 2\theta$$

Let's work through this step by step with a learning rate of $\alpha = 0.1$, starting at $\theta = 3$. At each step we need to do three things: compute the gradient at the current position, multiply it by the learning rate, then subtract that from the current $\theta$.

---

**Step 1** — starting at $\theta = 3$

First, compute the gradient at the current position:
$$\text{gradient} = 2\theta = 2 \times 3 = 6$$

Then multiply the gradient by the learning rate to get the step size:
$$\text{step size} = \alpha \times \text{gradient} = 0.1 \times 6 = 0.6$$

Then subtract the step size from the current $\theta$:
$$\theta \leftarrow 3 - 0.6 = 2.4$$

We moved from 3 to 2.4, closer to the minimum at 0. Check the loss improved:
$$L(3) = 3^2 = 9 \qquad \rightarrow \qquad L(2.4) = 2.4^2 = 5.76 \quad \checkmark$$

---

**Step 2** — now at $\theta = 2.4$

Compute the gradient at the new position:
$$\text{gradient} = 2\theta = 2 \times 2.4 = 4.8$$

Multiply by the learning rate:
$$\text{step size} = 0.1 \times 4.8 = 0.48$$

Subtract from current $\theta$:
$$\theta \leftarrow 2.4 - 0.48 = 1.92$$

Check the loss improved:
$$L(2.4) = 2.4^2 = 5.76 \qquad \rightarrow \qquad L(1.92) = 1.92^2 = 3.69 \quad \checkmark$$

---

**Step 3** — now at $\theta = 1.92$

Compute the gradient:
$$\text{gradient} = 2\theta = 2 \times 1.92 = 3.84$$

Multiply by the learning rate:
$$\text{step size} = 0.1 \times 3.84 = 0.384$$

Subtract from current $\theta$:
$$\theta \leftarrow 1.92 - 0.384 = 1.536$$

Check the loss improved:
$$L(1.92) = 1.92^2 = 3.69 \qquad \rightarrow \qquad L(1.536) = 1.536^2 = 2.36 \quad \checkmark$$


Look carefully at the step sizes across the three steps: 0.6, then 0.48, then 0.384. Each one is smaller than the last. This is not a coincidence; it is a direct consequence of the gradient formula $2\theta$. As $\theta$ gets smaller (closer to zero), the gradient $2\theta$ gets smaller too, so $\alpha \times \text{gradient}$ shrinks automatically. Gradient descent is **self-braking**: it takes large confident steps when it is far from the minimum and cautious small steps as it closes in, without needing any extra logic to make that happen. This property makes it a well-behaved and reliable algorithm for minimising loss functions.

To summarise: gradient descent is a simple but powerful algorithm that works by repeating three things over and over until the loss stops improving:

1. **Measure the slope**: compute the gradient $\frac{dL}{d\theta}$ at the current position. This tells you how steeply the loss is changing and in which direction.
2. **Take a step downhill**: subtract the gradient multiplied by the learning rate from the current $\theta$. The sign of the gradient ensures the step always goes in the downhill direction; the learning rate controls how far.
3. **Repeat**: use the new $\theta$ as the starting point and go back to step 1.

That is the entirety of gradient descent. The three worked steps above are not a special case; they are exactly what happens inside every neural network during training, repeated not three times but thousands or millions of times, across millions of parameters simultaneously. The loss function is more complex, the parameters are more numerous, and the gradient is computed automatically rather than by hand, but the underlying logic is identical to what you just worked through.

The self-braking property is worth pausing on. We did not write any code to slow the algorithm down as it approached the minimum; it slowed down automatically, as a mathematical consequence of the gradient formula. This is a general feature of gradient descent on smooth loss functions: steps are naturally proportional to how far you are from the minimum, which makes the algorithm stable and well-behaved without any extra engineering.

### 5.5.6 Gradient Descent with More Than One Parameter

Everything we have worked through so far has involved a single parameter $\theta$. Real models have many parameters: a simple linear classifier might have two or three, while a neural network might have millions. The natural question is: does gradient descent still work?

The answer is yes, and the core idea is exactly the same. With multiple parameters, each one has its own gradient, and each one is updated independently using the same rule:

$$\theta_i \leftarrow \theta_i - \alpha \cdot \frac{dL}{d\theta_i}$$

Think back to the hilly landscape analogy. With one parameter, the landscape was a curve, a 1D line, and gradient descent slid along it toward the lowest point. With two parameters, the landscape becomes a 2D surface, like a real hilly terrain, and gradient descent now moves across that surface. With three parameters it becomes a 3D volume, and so on. We cannot visualise these spaces for more than two or three parameters, but the mathematics works the same way regardless.

Each gradient $\frac{dL}{d\theta_i}$ measures the slope of the loss in the direction of one specific parameter: how much would the loss change if we nudged just $\theta_i$ while leaving all other parameters fixed? This is called a **partial derivative**, and it is the multi-parameter version of everything you have already seen. The word "partial" just means "with respect to one variable at a time".

The self-braking property carries over too. As gradient descent approaches the minimum of the loss surface, the lowest point across all parameter dimensions simultaneously, the gradient in every direction shrinks toward zero, and all steps become smaller automatically. The algorithm slows down and converges without any extra logic, just as it did in the single-parameter case.

This also connects directly to the curse of dimensionality in section 6 below. With millions of parameters, the loss surface has millions of dimensions, and finding the global minimum of that surface is genuinely hard. Local minima, saddle points (flat regions that are not true minima), and poorly chosen learning rates all become more likely problems as the parameter space grows. This is why training large neural networks requires careful tuning and why optimisation is an active research area: the basic algorithm is simple, but navigating a million-dimensional loss landscape reliably is not.

The key takeaway is that nothing conceptually new happens when you add more parameters. The update rule stays the same; the gradient still points downhill; the learning rate still controls step size. The only things that change are the number of gradient calculations needed per step and the complexity of the surface being navigated, both of which are handled automatically by modern software libraries.

### 5.5.7 Convex vs Non-Convex Optimisation

Everything we have seen so far has used a well-behaved loss function, one with a single, clearly defined lowest point. Gradient descent found it reliably regardless of where we started. This is not always the case in practice, and understanding why is important for anyone working with machine learning models.

#### Convex Loss Functions

A loss function is **convex** if its surface has a single bowl-like shape: one minimum, no valleys, no bumps, no flat regions that could trap the algorithm. The squared error function $L(\theta) = \theta^2$ that we worked through earlier is convex: it has exactly one lowest point, and gradient descent will always find it eventually, regardless of where you start. The 3D surface in Figure 6c is also convex; the bowl has one bottom.

Convexity is a mathematically precise property, but the intuition is simple: **draw a straight line between any two points on the surface, and that line always sits above the surface**. If that is true everywhere, the surface is convex and has only one minimum. Gradient descent on a convex function is guaranteed to converge to the global minimum, the best possible solution, given a small enough learning rate and enough steps. This is one of the most reassuring theoretical results in optimisation.

Many classical machine learning algorithms, logistic regression, support vector machines, linear regression, have convex loss functions. This is one of the reasons they are reliable and well-understood.

#### Non-Convex Loss Functions

Neural networks, by contrast, almost always have **non-convex** loss functions: surfaces with multiple valleys, ridges, flat plateaus, and saddle points (locations where the surface is flat but not at a minimum, like the centre of a mountain pass). You saw a simplified version of this in Figure 3, where the synthetic loss function had both a local minimum and a deeper global minimum.

The practical consequences are significant:

- **Local minima**: gradient descent may converge to a valley that is not the deepest one. The solution found depends on where the algorithm started, meaning two runs with different random initialisations can produce models with different quality.
- **Saddle points**: in high-dimensional spaces these are actually more common than local minima. At a saddle point the gradient is zero (so gradient descent stops moving) but the algorithm has not found a true minimum; it has found a flat region where some directions go up and others go down. Advanced optimisers include tricks to escape saddle points.
- **Flat plateaus**: large regions of the loss surface where the gradient is close to zero. Gradient descent slows almost to a stop in these regions, making training feel stuck even when the algorithm has not converged.
- **Sensitivity to initialisation**: because the final solution depends on the starting point, the initial values of the parameters matter. Modern neural networks use carefully designed initialisation strategies rather than random starting points to increase the chance of finding a good solution.

Despite all of these challenges, gradient descent works surprisingly well in practice on non-convex neural network loss surfaces. Current research suggests that in very high dimensions, most local minima of neural network loss functions have similar loss values, so even if the global minimum is not found, the solution reached is often nearly as good. This is an active area of research and not fully understood.

### 5.5.8 How are Gradients Computed in Practice

In the examples above, we knew the exact formula for the loss function, so the gradient at any point could be worked out by hand. In practice, when training a neural network, the loss function involves millions of parameters and hundreds of mathematical operations stacked on top of each other; no one could differentiate that by hand. Instead, two approaches are used:

1. **Automatic differentiation** (used by PyTorch, TensorFlow, and similar libraries): the software keeps a record of every mathematical operation performed during the forward pass and uses this record to compute the gradient automatically. You do not need to understand how this works internally; the library handles everything. When you call `.backward()` in PyTorch, for example, this is what is happening behind the scenes.

2. **Numerical differentiation** (used for illustration and checking): the gradient is approximated by evaluating the loss at two nearby points and measuring the slope between them, exactly like estimating the steepness of a hill by walking two small steps and measuring how much your altitude changed:

$$\frac{dL}{d\theta} \approx \frac{L(\theta + h) - L(\theta - h)}{2h}$$

In plain English: nudge $\theta$ a tiny amount to the right (by $h$) and measure the loss; nudge it the same tiny amount to the left and measure the loss again; divide the difference by the total distance moved ($2h$). The smaller $h$ is, the more accurate the approximation. You saw this exact approach used inside `Figure_3.py` when simulating the gradient descent path; the variable `h = 0.01` in that code is exactly this small nudge.

> **Key terms introduced in this section:**
> - **Loss function** $L(\theta)$: a formula that takes the current parameter value(s) and returns a single number measuring how wrong the model's predictions are, the quantity gradient descent tries to minimise
> - **Gradient** ($\frac{dL}{d\theta}$): the slope of the loss function at the current $\theta$. Positive means the curve is rising to the right, negative means it is falling, zero means it is flat (at or near a minimum)
> - **Gradient formula**: a rule that gives the gradient at *any* position. For $L(\theta) = \theta^2$ this is $2\theta$, meaning the slope depends on where you currently are
> - **Derivative**: the mathematical name for the gradient, a measure of how steeply one quantity changes in response to a tiny change in another
> - **Learning rate** ($\alpha$): scales how large each step is. Too large causes overshooting, too small causes very slow convergence
> - **Update rule**: $\theta \leftarrow \theta - \alpha \cdot \frac{dL}{d\theta}$, applied once per iteration to nudge $\theta$ downhill
> - **Convergence**: when the gradient is close to zero and further steps produce negligible improvement, the algorithm has effectively found the minimum
> - **Automatic differentiation**: the technique used by modern ML libraries to compute gradients automatically without any manual calculus

## 6. The Curse of Dimensionality

🔙 [Back to Table of Contents](#Table-of-Contents)

So far the fruit classifier we built uses a single feature: diameter. In real datasets we often have dozens or hundreds of features: colour, texture, weight, ripeness score, and so on. Intuitively, more features should always help; more information means better predictions. This intuition is wrong, and understanding *why* it is wrong is one of the most important ideas in machine learning.

**The curse of dimensionality** describes a set of deeply counterintuitive problems that emerge as the number of features (dimensions) grows. The core issue is that the volume of the feature space, the space of all possible combinations of feature values, grows *exponentially* with the number of dimensions. As a result, any fixed dataset of training examples becomes increasingly sparse: the same number of points that would feel dense in 2D feels empty and scattered in 100D. Algorithms that rely on finding nearby training examples, such as the nearest neighbour classifier you will encounter in later notebooks, quietly break down in high dimensions, and understanding why is the first step toward choosing the right tools for the job.

Before we look at the data consequences, it helps to build a geometric intuition. Consider a simple thought experiment in 2D: draw a circle inside a square, both centred at the origin, with the circle touching the edges of the square. The circle covers roughly 78% of the square's area, most of the space is inside the circle, close to the centre. Now do the same in 3D with a sphere inside a cube. The sphere still covers about 52% of the cube's volume. So far so good.

````python
# Listing 50.
%matplotlib widget
from visualisations.Figure_7 import show
show()
````

Now imagine doing the same in 10, 20, or 100 dimensions. In 10 dimensions, the equivalent sphere covers less than 0.25% of the hypercube's volume. By 20 dimensions it is essentially zero. Almost all of the space in a high-dimensional hypercube lies in the "corners", far from the centre, even though those corners are hard to visualise or reason about intuitively. This is the geometry of high-dimensional space, and it is nothing like our everyday experience of 2D or 3D.

### 6.1 Growing Volume - Everything Ends Up in the Corners

In two dimensions, a circle drawn inside a square covers most of the area. In three dimensions, a sphere inside a cube still covers a reasonable fraction of the volume. But as dimensions increase, the fraction of the hypercube's volume that lies *near the centre* shrinks toward nothing, almost all of the space migrates to the outer shell, far from the centre. Any algorithm that assumes "nearby" means "within a reasonable distance of the centre" will be operating in an almost entirely empty region of the space.

Figure 7 illustrates this with two panels:

- **Left panel**: the volume of a unit hypersphere (a sphere with radius 1) as the number of dimensions increases. You might expect the volume to keep growing as we add dimensions, but it does not: it peaks around 5 dimensions and then shrinks back toward zero. A unit sphere in 20 dimensions has a volume so close to zero it is effectively invisible. The surrounding hypercube, meanwhile, always has volume 1 by definition, so the sphere occupies a vanishingly small fraction of the available space.

- **Right panel**: a direct simulation of this effect. For each dimensionality, 50,000 random points are scattered uniformly inside the unit hypercube. The y-axis shows what fraction of those points fall within distance 0.5 of the centre, i.e. how many are "near the middle". In 2D, roughly three-quarters of points are near the centre. By dimension 10, fewer than 1 in 1,000 are. By dimension 20, the fraction is so small it rounds to zero. The printed output below the figure shows this numerically for the first ten dimensions so you can see exactly how rapidly the collapse happens.

The bar chart on the left and the bar chart printed below do not require any background in higher mathematics to read; just notice how quickly the numbers fall. This is the curse of dimensionality in its most direct geometric form, and it explains why techniques like dimensionality reduction (covered in later notebooks) are not just convenient but sometimes essential.


### 6.2 Data Sparsity - How Many Samples Do You Need

A direct consequence of the volume explosion described above is that the amount of data needed to adequately *cover* the feature space, to have enough training examples that any new test point is reasonably close to at least one of them, grows **exponentially** with the number of dimensions.

To see why, think about what it means to have good coverage in one dimension. Imagine your single feature (say, fruit diameter) ranges from 0 to 1. If you divide that range into 5 equal bins and want at least 10 training examples in each bin, you need $5 \times 10 = 50$ samples. Manageable.

Now add a second feature (say, colour intensity, also ranging from 0 to 1). The feature space is now a 2D grid of $5 \times 5 = 25$ cells, and you need $25 \times 10 = 250$ samples to cover it with the same density. Add a third feature and you need $5^3 \times 10 = 1{,}250$ samples. By the time you have 10 features the grid has $5^{10} \approx 10$ million cells and you would need 100 million samples to cover them, more than most real-world datasets contain. This is not a quirk of the grid-based argument; it reflects a genuine geometric reality.

The practical consequence is stark: with a fixed training set, every additional feature you add makes the existing data *sparser*. The same 200 training examples that felt abundant in 2 dimensions feel hopelessly scattered in 20.

A closely related problem affects any algorithm that relies on finding nearby training examples. The intuition behind many classifiers is simple: if you want to predict the label of a new, unseen data point, find the training example that is most similar to it, the *nearest neighbour*, and use its label as the prediction. If the nearest neighbour is close, this works well: a fruit with a diameter of 8.1 cm is probably the same class as the training fruit with a diameter of 8.0 cm. But if the nearest neighbour is far away, the prediction becomes unreliable: a fruit with a diameter of 8.1 cm tells you very little if the closest training example has a diameter of 4.3 cm. In low dimensions, with a reasonable amount of data, most test points will find a genuinely close training neighbour. In high dimensions, even the nearest training point may be so far away that using it for prediction is little better than guessing. "Nearby" stops meaning anything useful.

Figure 7b illustrates both of these effects directly:

- **Left panel**: the number of training samples needed to maintain a fixed coverage density as dimensions increase. The y-axis uses a logarithmic scale, meaning each gridline represents a tenfold increase, because the growth is so rapid that a linear scale would be unreadable. The horizontal reference lines at 1,000, 1,000,000, and 1 billion samples give a sense of what is actually achievable in practice.
- **Right panel**: a simulation with 200 training points and one test point, repeated across different dimensionalities. The y-axis shows the distance to the nearest training neighbour. The coloured bands mark three zones: green for distances where the neighbour is genuinely close and informative, orange for borderline cases, and red for distances where the nearest neighbour is so far away that using it for prediction is questionable. Watch how quickly the line moves from green into red as the number of dimensions grows.

### 6.3 Irrelevant Features Dilute the Signal

The previous section showed that data becomes sparse as dimensions grow. A third, subtler problem is that adding features which carry **no useful information** actively harms a model, even when the informative features remain present and unchanged.

To understand why, we first need to introduce how distance is measured between data points. The most common measure is **Euclidean distance**, the straight-line distance between two points, generalised from the familiar distance formula you may have seen in school. In two dimensions, the distance between points $(x_1, y_1)$ and $(x_2, y_2)$ is:

$$d = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}$$

This simply says: find how far apart the two points are in each dimension, square each difference, add them up, and take the square root. In three dimensions you add a third squared difference; in $d$ dimensions you add $d$ squared differences. The formula is the same regardless of how many dimensions there are.

The problem is what happens when you add dimensions that contain only random noise: measurements that have nothing to do with what you are trying to predict. Each noise dimension adds a random squared difference to the distance calculation. Since these differences are random and always positive, they push *all* distances upward, distances between points of the same class and distances between points of different classes alike. The informative signal (the genuine separation between classes) gets diluted by the accumulating noise.

We measure this effect using the **within/between distance ratio**: the average distance between two points of the *same* class (within-class) divided by the average distance between two points of *different* classes (between-class). When this ratio is small, say 0.3, the two classes are well separated relative to their internal spread and a distance-based classifier can easily tell them apart. As we add noise dimensions, both the within-class and between-class distances grow, but they grow toward each other. When the ratio approaches 1.0, within-class distances and between-class distances are roughly equal, the two classes look identical to any algorithm that relies on distance, and classification becomes no better than random guessing.

Additional noise dimensions are then added: random numbers with no relationship to the class labels whatsoever. The four panels show what happens to the within/between ratio as the total number of dimensions grows from 2 to 20, even though the two informative features remain unchanged throughout. Notice that the scatter plots look identical across all four panels (the informative features have not changed), yet the ratio printed in each title tells a very different story about how separable the classes appear to a distance-based model operating in the full feature space.

It is one of the most intuitive classifiers imaginable: if you want to know what class a new fruit belongs to, look at the five most similar fruits in your training data and take a majority vote.

**Practical solutions to the curse of dimensionality:**

| Approach | How it helps |
|----------|-------------|

| **Collecting more data** | Increases density, but is often impractical in high dimensions |
