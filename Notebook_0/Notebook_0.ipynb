## 3. Introducing Notebooks

🔙 [Back to Table of Contents](#Table-of-Contents)

The resource you're reading is a [Jupyter](https://jupyter.org/) notebook (originally called an [IPython](https://en.wikipedia.org/wiki/IPython) notebook). A notebook is a document made of **cells**: some cells contain text (like this one), others contain executable Python code. You can run individual code cells in any order and see their output appear immediately below them.

This interactive style differs from running a "normal" Python script, where you would:

1. Write all your code in a single file (e.g., `script.py`).
2. Hand the file to the Python interpreter.
3. Watch it execute top-to-bottom in one pass.

A notebook hides those steps. When you run a cell, its code is sent to a long-lived Python interpreter called the **kernel**. Crucially, **the kernel keeps state between cells**: variables and functions defined in one cell remain available in all later cells until you restart the kernel. This is what makes notebooks ideal for teaching, for exploratory data analysis, and for ML research. You can load a dataset once and then poke at it across dozens of cells without reloading it each time.

One consequence of the shared kernel that catches everyone out at some point is that **the order you run cells matters**. If you scroll up and re-run an earlier cell after defining variables below it, you may overwrite state in surprising ways. The numbers in the `[ ]` to the left of each code cell tell you the execution order, not the position in the notebook. If something behaves oddly, a clean fix is often "Restart Kernel & Run All".

## 4. Python - the language

🔙 [Back to Table of Contents](#Table-of-Contents)

Programming means writing instructions that direct a computer to do something useful. Those instructions ultimately have to reach the CPU as **machine code**, the very low-level numeric instructions the CPU's circuits are wired to execute. We don't write machine code directly; we write in a high-level language and let a tool translate. There are two broad strategies for this translation.

**Compiled languages.** A *compiler* reads your entire source file and translates it ahead of time into a standalone executable file of machine code. You compile once, then run the executable as many times as you like. C, C++, and Rust work this way. The compiled binary is fast but is tied to a specific operating system and CPU architecture.

**Interpreted languages.** An *interpreter* reads your source file and translates it on the fly, executing each piece as it goes. There is no separate "build" step that produces a standalone executable. Python, Ruby, and JavaScript work this way. (In Python's case, the interpreter actually compiles to an intermediate bytecode first, then executes that, but you never see the bytecode.) Interpreted code is slower to run than compiled code, but it is more flexible and far easier to develop in.

Here we use [Python 3](https://en.wikipedia.org/wiki/Python_(programming_language)). Depending on how you're running this notebook (Google Colab, Jupyter Lab, or local Jupyter), an interpreter is already provided for you.

The Python interpreter executes your code **line by line**, top to bottom. When it meets a line, it first checks whether the line is a **comment** (we'll cover comments in the next section). Comments are ignored. For real code, it processes the statement, assigning a variable, calling a function, evaluating a loop, and then moves on.

### 4.1 Why Python for machine learning

🔙 [Back to Table of Contents](#Table-of-Contents)

Python's relative slowness is, on paper, a drawback. So why does essentially every ML framework, including NumPy, PyTorch, TensorFlow, JAX, and scikit-learn, present a Python interface? Two reasons:

1. **It's easy to read and write.** ML research moves fast; you want a language that gets out of your way.
2. **The slow bits don't run in Python.** When you call `numpy.dot(A, B)`, Python is just a thin wrapper that hands the work to highly optimised compiled C and Fortran code (and, on a GPU, to CUDA kernels). Python orchestrates; the heavy lifting happens elsewhere.

So in this primer you are learning the *glue* language of modern ML. Master the glue, and the libraries will fit.

### 4.2 Trade-offs of being interpreted

🔙 [Back to Table of Contents](#Table-of-Contents)

| Advantages | Disadvantages |
|------------|---------------|
| You can declare variables without specifying their type. | You cannot control memory layout, so writing the very fastest code in pure Python is hard. |
| You can run individual statements interactively (as you will in this notebook). | Pure-Python loops over large data are slow, so for ML, vectorise with NumPy or use a JIT compiler. |
| Code is portable: any machine with a Python interpreter can run it. | Because types aren't declared, type-mismatch bugs only show up at runtime. |

### 4.3 PEP 8 - writing readable Python

🔙 [Back to Table of Contents](#Table-of-Contents)

Readability matters. A widely-adopted style guide for Python, [PEP 8](https://peps.python.org/pep-0008/), spells out conventions for indentation, naming, line length, spacing around operators, and so on. The most important ones to adopt now:

- **4 spaces per indentation level** (not tabs, not 2 spaces).
- **`snake_case`** for variable and function names (`user_count`, not `UserCount` or `userCount`).
- **`UPPER_SNAKE_CASE`** for constants.
- Lines under **~79–99 characters** where reasonable.
- Spaces around binary operators (`x = a + b`, not `x=a+b`).

Code in this notebook follows PEP 8 throughout. When in doubt, consult [the standard](https://peps.python.org/pep-0008/).

## 5. Comments

🔙 [Back to Table of Contents](#Table-of-Contents)

Comments are notes in your source code that the interpreter ignores. They exist for human readers: your collaborators, and, most often, **future you** trying to remember what past you was thinking. Useful reasons to comment:

- Explain *why* the code does what it does, especially for non-obvious choices. (The *what* is usually clear from the code itself; the *why* is not.)
- Flag assumptions, edge cases, or known limitations.
- Make a complex pipeline easier to skim by labelling sections.

Good comments are **informative, clear, and brief**. Avoid restating the obvious: `i += 1  # add 1 to i` is noise. Clarify the surprising instead.

Python has two main kinds of comment:

- **Inline / single-line comments**, introduced with a hash `#`. Everything on the line after `#` is ignored.
- **Docstrings**, which are triple-quoted strings placed at the top of a module, class, or function. Strictly speaking these are string literals, not comments, but they serve as documentation and can be retrieved at runtime with `help(...)`.

You will also see triple-quoted strings used as scratch "block comments". This works because an unassigned string literal is harmless: Python just evaluates it and throws the result away. It is **not the recommended style** for ordinary commenting, though. Use `#` for that.

````python
# Listing 1.
# This is a single-line comment.

# A short comment explaining the code below. Here we add two numbers.
number_1 = 10
number_2 = 15
result = number_1 + number_2

# You can stack `#` comments to span multiple lines. Use this for short
# blocks. For longer explanations that document a function, prefer a
# docstring (we cover those in the Functions section).
#
# Below is a small example: we find the index of each character in a
# string. Strings in Python are sequences of characters, and each
# character has a position starting from 0:
#
#   index:   0  1  2
#            |  |  |
#            v  v  v
#          " R  o  b "
#
# So asking for the index of "R" in "Rob" gives 0.
text = "Rob"
print("Position of 'R':", text.find('R'))
print("Position of 'o':", text.find('o'))
print("Position of 'b':", text.find('b'))
````

## 6. Variables

🔙 [Back to Table of Contents](#Table-of-Contents)

A variable is a name that refers to a value stored somewhere in your computer's memory. Python takes care of the memory management for you; you just say "this name refers to this value" and the interpreter sorts out the rest.

Python is **dynamically typed**: you do not declare what type a variable will hold. The interpreter figures the type out from the value you assign, and the type of a variable can even change over the program's lifetime. That flexibility is convenient, but it also means it is your responsibility, as the programmer, to keep types consistent.

The core scalar (single-value) types you need to know now:

| Type     | Examples                          | How you create one     |
|----------|-----------------------------------|------------------------|
| `int`    | -2, -1, 0, 1, 2, ...              | `x = 1`                |
| `float`  | -2.1, -1.4, 0.01, 1.2, 2.8, ...   | `x = 0.5675`           |
| `str`    | `"hello"`                         | `x = "hello"`          |
| `bool`   | `True`, `False`                   | `x = True`             |
| `NoneType` | `None` (the "no value" value)    | `x = None`             |

Let's explore variables in code. Read the cell, then run it.

````python
# Listing 6.
# Create an integer variable. Integers are whole numbers (..., -1, 0, 1, ...).
number_1 = 1

# `type(...)` tells us the type of any object.
print("Type of 'number_1':", type(number_1))

# Every object in Python also has a unique identifier, returned by `id(...)`.
# You can think of the id as the object's address in memory. It is mainly
# useful as a teaching tool — in real code you rarely call `id` directly.
print("ID of 'number_1':", id(number_1))

# Now reassign `number_1` to a different value.
number_1 = 2
print("ID after reassignment:", id(number_1))
````

You should see the ID change after the reassignment. Here is the subtle but important point: **the variable `number_1` is a name, and the value `1` is a separate object in memory.** When we wrote `number_1 = 2` we did not modify the object that held `1`; we made the name `number_1` refer to a different object that holds `2`. The old object is still there (until Python's garbage collector decides nobody wants it).

What if we reassign back to the original value?

````python
# Listing 7.
number_1 = 1
print("ID after reassigning back to 1:", id(number_1))
````

You will usually find the ID matches the very first one we printed. Small integers like `1` are *cached* by the Python interpreter, so there is a single shared `1` object that every name pointing to `1` reuses. We'll come back to this in the **Immutability** section.

Now some other types:

````python
# Listing 8.
# A float (number with a fractional component).
number_1 = 1.999
print("Type of 'number_1':", type(number_1))

# A string (text). Strings can use single or double quotes.
text = "a string"
print("Type of 'text':", type(text))

# Strings have a length (number of characters), accessible via `len`.
print("Length of 'text':", len(text))

# A boolean (True/False).
flag = True
print("Type of 'flag':", type(flag))

# None — Python's way of saying "no value here yet". Common as a default
# for function arguments and as a placeholder.
nothing = None
print("Type of 'nothing':", type(nothing))
````

If you need to check the type of a variable in your own code, `isinstance(value, type)` returns a boolean. Prefer `isinstance` over comparing `type(x) == int` directly, since `isinstance` correctly handles subclasses, which matters as your code grows.

````python
# Listing 9.
text = "string"
number_1 = 1
number_2 = 1.0

print("Is 'text' a str?    ", isinstance(text, str))
print("Is 'number_1' an int?", isinstance(number_1, int))
print("Is 'number_2' a float?", isinstance(number_2, float))
print("Is 'number_1' a float?", isinstance(number_1, float))
````

## 7. Casting

🔙 [Back to Table of Contents](#Table-of-Contents)

[Casting](https://en.wikipedia.org/wiki/Type_conversion) means converting a value from one type to another. In Python, you cast by calling the type's constructor function: `int(x)`, `float(x)`, `str(x)`, `bool(x)`, and so on. Watch the video below before continuing.

````python
# Listing 10.
from IPython.display import IFrame
IFrame(src='https://www.youtube.com/embed/ALvbltAPOcI', width=560, height=315)
````

Now apply that. Read the code below, predict what will print, then run it.

````python
# Listing 11.
# Start with an integer.
number_1 = 1
print("number_1:", number_1, "type:", type(number_1))

# Cast to float.
number_2 = float(number_1)
print("number_2:", number_2, "type:", type(number_2))

# Cast to string.
text = str(number_1)
print("text:", text, "type:", type(text))

# And back to int.
number_3 = int(text)
print("number_3:", number_3, "type:", type(number_3))
````

All easy so far. Now a subtle one. When you cast a float to an int, the fractional part is **truncated toward zero**, *not* rounded.

````python
# Listing 12.
# Truncation toward zero (not rounding).
print(int(1.6732))   # 1, not 2
print(int(-1.6732))  # -1, not -2

# If you actually want to round, use `round(...)`. With one argument it
# rounds to the nearest integer (using banker's rounding for .5 cases).
print(round(1.6732))   # 2
print(round(-1.6732))  # -2
print(round(2.5))      # 2  (banker's rounding: round-half-to-even)
print(round(3.5))      # 4
````

Mixing numeric types in arithmetic is usually safe: Python silently widens the result to the more general type. Multiplying or dividing by a float produces a float.

````python
# Listing 13.
# An int multiplied by a float gives a float.
number_1 = 1 * 1.0
print("Value:", number_1, "Type:", type(number_1))

# Integer division of integers — note the `/` always produces a float in
# Python 3, even when the result is mathematically a whole number.
print(10 / 2, type(10 / 2))  # 5.0  <class 'float'>

# `//` is floor division — it gives you the integer quotient.
print(10 // 3, type(10 // 3))  # 3   <class 'int'>
````

Casting can also fail. If you try to convert a string that does not look like an integer, Python raises an error. The cell below deliberately triggers one; run it and read the error.

````python
# Listing 14.
# A string containing what looks like a number.
text = "1.0"
print("text:", text, "type:", type(text))

# Try to convert directly to int. This fails because `"1.0"` is a valid
# *float* literal, not a valid *int* literal — `int(...)` is strict.
print("Casting to int...")
number_1 = int(text)
````

The error message — `invalid literal for int() with base 10: '1.0'` — is Python saying: "I tried to read `'1.0'` as a base-10 integer and could not." The fix is to cast in two steps: string to float to int. Try it yourself first, then check the solution.

````python
# Listing 15.
text = "1.0"

# Cast `text` to an integer, store in `number_1`, then print both
# the value and its type.
print("Your code here.")
````

**Solution**

````python
# Listing 16.
text = "1.0"
print("text:", text, "type:", type(text))

# Two-step cast: str -> float -> int.
number_1 = int(float(text))
print("number_1:", number_1, "type:", type(number_1))
````

Here is an everyday example: computing an average.

````python
# Listing 17.
# Some integers.
x = [1, 2, 3, 4, 5, 6, 7, 8]

# `sum` adds them up; `len` counts them.
total = sum(x)
items = len(x)

# Divide. In Python 3, `/` always produces a float.
mean = total / items
print(f"Sum: {total}, count: {items}, mean: {mean} (type: {type(mean).__name__})")
````

You will notice I used an **f-string** there: `f"Sum: {total}, count: {items}, ..."`. F-strings (formatted string literals) are the modern way to embed values inside strings. They are far cleaner than concatenation with `+` or repeated commas in `print`, and you will see them constantly in ML code. We'll cover them properly in the next section.

A historical note: in **Python 2**, dividing two integers used `/` and produced an integer (truncating the remainder), so `7 / 2` gave `3`, not `3.5`. This is the kind of silent bug that haunted older numerical code. Python 3 fixed this by making `/` always produce a float and providing `//` for explicit integer division. You may still see Python 2 idioms like `float(total) / float(items)` in old code, now you know why.

## 8. Strings

🔙 [Back to Table of Contents](#Table-of-Contents)

A string is a sequence of characters. You already know you can print them:

````python
# Listing 18.
print("A string")
````

You can read a single character by indexing with square brackets. Indexing in Python starts at **0**, not 1.

````python
# Listing 19.
text = "Hello"
print(text[0])   # 'H'
print(text[4])   # 'o'  (the last character; index 4 because length is 5)
print(text[-1])  # 'o'  (negative indices count from the end)
````

You can extract a substring (a "slice") using the colon syntax `s[start:stop]`. The slice is **half-open**: it includes `start`, excludes `stop`. So `text[0:2]` gives the characters at positions 0 and 1, which is `"He"`.

````python
# Listing 20.
text = "Hello"
print(text[0:2])    # 'He'   characters 0 and 1
print(text[1:4])    # 'ell'  characters 1, 2, 3
print(text[:3])     # 'Hel'  start defaults to 0
print(text[2:])     # 'llo'  stop defaults to len(text)
print(text[:])      # 'Hello'  a full copy
print(text[::-1])   # 'olleH'  step of -1 reverses the string
````

You'll meet the same slicing notation again when we get to lists, and again when you start using NumPy arrays and pandas DataFrames. Learning it well now pays off many times over.

### 8.1 f-strings - the modern way to build strings

🔙 [Back to Table of Contents](#Table-of-Contents)

Before we look at the rest of the string API, let's properly introduce **f-strings**. They are the recommended way to build strings that include variable values. The syntax is `f"... {expression} ..."`, where any expression inside curly braces is evaluated and inserted into the string.

````python
# Listing 22.
name = "Rob"
employer = "Bob"

# The clean, modern way: f-strings.
print(f"My name is {name}. My employer is {employer}.")

# You can put any expression inside the braces, not just a bare name.
x = 7
print(f"x squared is {x**2}")

# Format specifiers come after a colon. Some useful ones:
pi = 3.141592653589793
print(f"pi to 2 decimal places: {pi:.2f}")
print(f"pi to 5 significant figures: {pi:.5g}")
print(f"a number with commas: {1_000_000:,}")
print(f"left-padded to width 8: '{42:>8}'")
print(f"as a percentage: {0.7563:.1%}")
````

You may still see two older styles in books and tutorials:

- `"%s is %d" % ("age", 30)` — the C-style `%` operator. Outdated; avoid in new code.
- `"{} is {}".format("age", 30)` — the `str.format` method. Still works fine, but f-strings are clearer.

In this notebook from here on, we will use f-strings whenever we build strings containing values.

### 8.2 Strings as sequences

🔙 [Back to Table of Contents](#Table-of-Contents)

Strings behave like ordered, immutable sequences of characters. That means many of the operations you'll later use on lists (indexing, slicing, length, iteration, membership testing) also work on strings.

````python
# Listing 23.
text = "Rob"

# Length of the sequence.
print(f"Length: {len(text)}")

# Membership test.
print(f"Is 'o' in text? {'o' in text}")

# Iteration: looping over the characters.
for ch in text:
    print(ch)
````

What happens if we ask for a position that doesn't exist? Python raises an `IndexError`. Run the cell: the error is *expected*.

````python
# Listing 24.
text = "Rob"
print(text[3])  # 'Rob' has indices 0, 1, 2 — there is no index 3.
````

Get used to reading error messages: they are diagnostic, not punitive. The traceback tells you:

1. **The exception type** (`IndexError`) — what category of problem occurred.
2. **The message** (`string index out of range`) — a short human explanation.
3. **The line and file** where it happened.

We can defend against this kind of error by checking the index ourselves before using it.

````python
# Listing 25.
def char_at(text, index):
    """Return the character at `index` in `text`, or a friendly message
    if the index is out of range. Handles negative indices via len(text)."""
    if index < -len(text) or index >= len(text):
        return f"Index {index} is outside the range of a string of length {len(text)}."
    return text[index]

text = "Rob"
print(char_at(text, 1))    # 'o'
print(char_at(text, 3))    # message
print(char_at(text, -1))   # 'b'
print(char_at(text, -10))  # message
````

````python
# Listing 27.
test_index = 3
text = "Rob"

# Valid indices for a string of length n are: -n, -n+1, ..., n-1.
# So `test_index` is out of range iff it is < -n or >= n.
if test_index < -len(text) or test_index >= len(text):
    print(f"Index {test_index} is out of range for {text!r}.")
else:
    print(f"text[{test_index}] = {text[test_index]!r}")

# Bonus: wrap that logic in a reusable function.
def safe_get(seq, index):
    """Return seq[index] if valid, else None."""
    if index < -len(seq) or index >= len(seq):
        return None
    return seq[index]

for i in [0, 1, 2, 3, -1, -3, -4]:
    print(f"safe_get('Rob', {i}) -> {safe_get('Rob', i)!r}")
````

### 9.2 List comprehensions

🔙 [Back to Table of Contents](#Table-of-Contents)

If there is one Python idiom you must learn early for ML, it is the **list comprehension**. It builds a new list from an existing iterable in a single expression. The general shape is:

```python
[expression for item in iterable]                  # transform every item
[expression for item in iterable if condition]     # transform a filtered subset
```

You will see this notation absolutely everywhere in ML code: building lists of file paths, transforming labels, batching, you name it. The two-line `for` loop version works fine, but the comprehension is shorter, clearer, and often faster.

The cell below shows the same computation written two ways and is meant to make the **list comprehension** syntax click by direct comparison. The first block builds the list of squares `0², 1², …, 9²` the long way: start with an empty list `squares_long = []`, then loop over `range(10)` and `append` each `n ** 2` one at a time.

The second block does the identical job in a single expression, `[n ** 2 for n in range(10)]`, which reads almost like the English sentence "the value `n ** 2` for each `n` in `range(10)`". Printing both lists, and then printing `squares_long == squares_short`, confirms they produce exactly the same result; the comprehension is just a more concise way to spell the same loop.

The third block adds a filter: `[n ** 2 for n in range(10) if n % 2 == 0]` keeps only those `n` for which `n % 2 == 0` (the even numbers), so we get the squares of `0, 2, 4, 6, 8`. The final block shows that comprehensions aren't limited to numbers. Given `words = ['hello', 'world', 'python']`, the expression `[len(w) for w in words]` produces a list of word lengths and `[w.upper() for w in words]` produces a list of uppercased copies. The pattern in all four cases is the same: take an iterable, apply an expression to each element, optionally filter, and collect the results into a new list.

Run the code when ready.

````python
# Listing 36.
# Squares of 0..9 — the loop version vs the comprehension version.
squares_long = []
for n in range(10):
    squares_long.append(n ** 2)

squares_short = [n ** 2 for n in range(10)]

print(squares_long)
print(squares_short)
print(squares_long == squares_short)  # True

# With a filter: only even squares.
even_squares = [n ** 2 for n in range(10) if n % 2 == 0]
print(even_squares)

# Transforming a list of strings.
words = ['hello', 'world', 'python']
lengths = [len(w) for w in words]
uppers = [w.upper() for w in words]
print(lengths)
print(uppers)
````

**One warning.** Comprehensions are wonderful, but a comprehension that spans three lines and computes something intricate is a comprehension that should have been a regular `for` loop. Readability beats cleverness.

## 10. Dictionaries

🔙 [Back to Table of Contents](#Table-of-Contents)

A **dictionary** (`dict`) is a mutable collection of **key–value pairs**. You look up values by their key, not by a positional index. Keys must be **hashable** (immutable types like `str`, `int`, `float`, `tuple` are fine; lists and other dicts are not). Values can be anything.

Dictionaries are everywhere in real Python code: configuration objects, JSON data, model hyperparameters, lookup tables. 

Create a dictionary with curly braces:

```python
person = {'name': 'Rob', 'employer': 'Bob', 'id': 1}
```

Access a value by its key:

```python
print(person['name'])  # 'Rob'
```

Assign a new value (creating the key if it doesn't yet exist, replacing it if it does):

```python
person['name'] = 'Robert'
```

````python
# Listing 37.
# Create a dictionary.
person = {'name': 'Rob', 'employer': 'Bob', 'id': 1}
print(person)

# Access values by key.
print(f"name:     {person['name']}")
print(f"employer: {person['employer']}")
print(f"id:       {person['id']}")

# Check whether a key exists. `in` tests against keys (not values).
print(f"'id' in person?    {'id' in person}")
print(f"'salary' in person? {'salary' in person}")

# Safer lookup: `.get(key, default)` returns the default if the key is missing,
# instead of raising a KeyError. Very common in ML code that handles optional fields.
print(f"person.get('salary', 'unknown'): {person.get('salary', 'unknown')}")

# Add a new key.
person['salary'] = 42000
print(person)

# Update a key. Same syntax as assigning — Python figures out the rest.
person['salary'] = 45000
print(person)

# Remove a key.
del person['salary']
print(person)
````

### 10.1 Iterating over dictionaries

🔙 [Back to Table of Contents](#Table-of-Contents)

There are three common loop patterns. Learn all three.

````python
# Listing 38.
person = {'name': 'Rob', 'employer': 'Bob', 'id': 1}

# 1. Iterate over keys (the default).
print("--- keys ---")
for key in person:
    print(key)

# 2. Iterate over values.
print("--- values ---")
for value in person.values():
    print(value)

# 3. Iterate over key-value pairs. This is the most common — use it whenever
#    you need both. `.items()` returns (key, value) tuples.
print("--- items ---")
for key, value in person.items():
    print(f"{key}: {value}")
````

## 11. Tuples

🔙 [Back to Table of Contents](#Table-of-Contents)

A **tuple** is an ordered, **immutable** collection. Think of it as a list you cannot modify. Tuples are written with parentheses (or, often, with no brackets at all):

```python
point = (3.0, 4.0)
also_a_point = 3.0, 4.0   # the parentheses are optional in most contexts
```

Why bother with tuples when we have lists? Two reasons that matter for ML:

1. **Multiple return values.** Python functions return one object, but that object is very often a tuple. `model.train_test_split(...)` returns one tuple containing four arrays; you destructure it.
2. **Use as dictionary keys.** Tuples are hashable (because immutable), so you can use a tuple like `(row, column)` or `("epoch_5", "validation")` as a dictionary key. Lists cannot do this.

A wart in the syntax: a one-element tuple needs a trailing comma, because `(x)` is just `x` in parentheses. Hence `(x,)`.

````python
# Listing 40.
# Creating and indexing a tuple.
point = (3.0, 4.0)
print(f"x = {point[0]}, y = {point[1]}")
print(f"type: {type(point).__name__}, length: {len(point)}")

# Tuple unpacking — the common Pythonic style.
x, y = point
print(f"unpacked: x={x}, y={y}")

# The trailing-comma wart.
not_a_tuple = (5)
a_tuple = (5,)
print(f"(5)  is {type(not_a_tuple).__name__}")   # int
print(f"(5,) is {type(a_tuple).__name__}")       # tuple

# Tuples are immutable. The line below would raise TypeError.
# point[0] = 99  # uncomment to see the error

# Common idiom: swap two variables using tuple packing/unpacking.
a, b = 1, 2
a, b = b, a
print(f"swapped: a={a}, b={b}")
````

## 12. Mutability and Immutability

🔙 [Back to Table of Contents](#Table-of-Contents)

This is one of the topics that trips up almost every new Python programmer. It's worth slowing down.

**Mutable** objects can be modified in place after they are created. **Immutable** objects cannot: once they exist, their value is fixed, and any operation that "changes" them actually produces a brand-new object.

| Type | Mutable? |
|------|----------|
| `bool`, `int`, `float`, `str` | No |
| `tuple` | No |
| `frozenset` | No |
| `list` | **Yes** |
| `set` | **Yes** |
| `dict` | **Yes** |

The key distinction is between **rebinding a name** and **mutating an object**. When you write `x = 5` and then `x = 6`, you have *rebound* `x` to a different object: the integer `5` was not modified. When you write `lst.append(7)`, you have *mutated* the list: there is still only one list object, but its contents have changed.

This matters in ML for two practical reasons:

1. **NumPy arrays and PyTorch tensors are mutable.** Two variables pointing to the same tensor can produce surprises if you mutate through one of them.
2. **Function arguments.** Passing a list to a function and having the function call `.append` on it changes the *caller's* list. This is both useful and dangerous.

### 12.2 Mutation through aliases

### 12.3 == vs is

🔙 [Back to Table of Contents](#Table-of-Contents)

These two comparisons answer different questions:

- `a == b` — are the *values* equal? (Calls `__eq__`.)
- `a is b` — are these the **same object**? (Compares identities, i.e. `id(a) == id(b)`.)

Almost always you want `==`. The only common exception is comparing to a singleton like `None`, where the convention is `x is None` (because there is only ever one `None`).

````python
# Listing 45.
list_a = [1, 2, 3]
list_b = [1, 2, 3]

print(f"list_a == list_b? {list_a == list_b}")   # True  — same contents
print(f"list_a is list_b? {list_a is list_b}")   # False — different objects

x = None
print(f"x is None? {x is None}")   # True — correct check
print(f"x == None? {x == None}")   # also True, but `is None` is the convention
````

### 12.4 Integer and string caching - a teaching curiosity

🔙 [Back to Table of Contents](#Table-of-Contents)

You may have noticed that small integers and short strings sometimes share an identity even when written as separate literals. This is a memory optimisation: the CPython interpreter pre-allocates integers from -5 to 256 and reuses them, and it sometimes interns short strings too. **You should never rely on this in real code.** Use `==`, not `is`, for equality. But it's worth seeing once so you don't get confused by it.

````python
# Listing 46.
# Small ints in CPython are pre-allocated and reused. The literal `256`
# always refers to the same cached object.
a = 256
b = 256
print(f"a, b both = 256:    a is b? {a is b}")   # True (cached)

# For 257 we have to be careful: the Python *compiler* folds identical
# literals inside a single statement to the same constant, so even
# `257 is 257` can look True. To see the underlying behaviour, we build
# the value in a way the compiler cannot fold — e.g. by parsing it from
# a string at runtime.
c = int("257")
d = int("257")
print(f"c, d via int('257'): c is d? {c is d}")  # typically False — not cached

# A string with non-identifier characters is unlikely to be interned.
s1 = "python is cool!"
s2 = "".join(["python", " is cool!"])  # build at runtime; defeats interning
print(f"runtime-built strings equal:  {s1 == s2}")
print(f"runtime-built strings same?:  {s1 is s2}")

# Empty immutables are typically shared singletons in CPython.
print(f"empty tuple equality:        {() == ()}")
print(f"two (1,) tuples are equal:   {(1,) == (1,)}")

# The takeaway: use `==` for value equality. Reserve `is` for
# `x is None` and similar singleton checks. Don't rely on caching.
````

### 12.5 A subtlety - + vs += on lists

🔙 [Back to Table of Contents](#Table-of-Contents)

Both expressions below produce a list `[1, 2, 3, 4]`, but they behave differently in terms of which object ends up where.

````python
# Listing 47.
# Case 1: concatenation with `+`. Creates a NEW list; the original is untouched.
a = [1, 2, 3]
print(f"id before: {id(a)}")
a = a + [4]
print(f"id after  + [4]: {id(a)}  # a now refers to a NEW list")

# Case 2: in-place extension with `+=`. Mutates the existing list.
b = [1, 2, 3]
print(f"id before: {id(b)}")
b += [4]
print(f"id after += [4]: {id(b)}  # same id — same object, now with 4 appended")
````

The implementation reason: `a + [4]` calls the list's `__add__` method, which by contract returns a new list. `b += [4]` calls `__iadd__` (in-place add), which mutates `b` and returns it.

This distinction matters because if there are two names pointing at the list, the two cases differ visibly:

````python
# Listing 48.
x = [1, 2, 3]
y = x

x = x + [4]      # rebinds x; y still sees the old list
print(f"x: {x}, y: {y}")

x = [1, 2, 3]
y = x
x += [4]         # mutates the shared list; y sees the change
print(f"x: {x}, y: {y}")
````

## 14. Operators and precedence

🔙 [Back to Table of Contents](#Table-of-Contents)

Python provides several families of operators.

**Arithmetic** — `+`, `-`, `*`, `/` (float division), `//` (floor division), `%` (modulo / remainder), `**` (exponent).

**Comparison** — `<`, `<=`, `>`, `>=`, `==`, `!=`.

**Logical** — `and`, `or`, `not`.

**Identity** — `is`, `is not`.

**Membership** — `in`, `not in`.

Operator **precedence** is the order in which the interpreter evaluates a mixed expression. The rules largely follow ordinary maths: exponents before products before sums. Here are the most relevant rules, **high precedence at the top**:

| Group | Operators |
|-------|-----------|
| Parentheses | `( ... )` |
| Exponent | `**` |
| Unary | `+x`, `-x`, `not x` |
| Multiplicative | `*`, `/`, `//`, `%` |
| Additive | `+`, `-` |
| Comparisons | `<`, `<=`, `>`, `>=`, `==`, `!=` |
| Membership / identity | `in`, `not in`, `is`, `is not` |
| Boolean `not` | `not` |
| Boolean `and` | `and` |
| Boolean `or` | `or` |

When in doubt, **add parentheses**. They cost nothing and remove all ambiguity for the reader.

````python
# Listing 52.
# Same precedence as ordinary maths: * and / bind tighter than + and -.
# So `2 + 5 - 2 * 8 / 4` is evaluated as:
#   2 + 5 - ((2 * 8) / 4)
# = 2 + 5 - (16 / 4)
# = 2 + 5 - 4.0
# = 3.0
print(2 + 5 - 2 * 8 / 4)
````

**Your turn.** Use parentheses to change the order of evaluation in the expression `2 + 5 - 2 * 8 / 4` so the answer comes out as **8**, *without* changing any of the numbers or operators. (Hint: think about which sub-expression needs to be evaluated first to produce 6, so that `2 + 6 = 8`.) The solution is below.

````python
# Listing 53.
# Add parentheses (no other changes) so this prints 8.0.
print(2 + 5 - 2 * 8 / 4)
````

**Solution.** Several bracketings work; the cleanest is `2 + (((5 - 2) * 8) / 4) = 2 + (24 / 4) = 2 + 6.0 = 8.0`.

````python
# Listing 54.
# Several valid bracketings that produce 8:
print("2 + (((5 - 2) * 8) / 4)  =", 2 + (((5 - 2) * 8) / 4))   # 8.0
print("2 + ((5 - 2) * (8 / 4))  =", 2 + ((5 - 2) * (8 / 4)))   # 8.0

# For contrast, here are a few other reachable values:
print("\nOther bracketings of the same expression give very different results:")
print("2 + 5 - 2 * 8 / 4         =", 2 + 5 - 2 * 8 / 4)        # 3.0  (the default)
print("(2 + 5 - 2) * 8 / 4       =", (2 + 5 - 2) * 8 / 4)      # 10.0
print("(2 + 5 - 2 * 8) / 4       =", (2 + 5 - 2 * 8) / 4)      # -2.25
````

Rule of thumb: when you have an expression with a mix of operators and you would have to pause to work out the precedence, add parentheses. Your future self and your collaborators will thank you.

### 14.1 Short-circuit evaluation of and and or

🔙 [Back to Table of Contents](#Table-of-Contents)

One last property of the boolean operators that catches people out: they **short-circuit**. `a and b` only evaluates `b` if `a` is truthy; `a or b` only evaluates `b` if `a` is falsy. The whole expression returns one of the original operands, not necessarily `True`/`False`.

````python
# Listing 55.
# `and` returns the first falsy operand, or the last operand.
print(0 and 1)        # 0  (short-circuits at 0)
print(1 and 2)        # 2  (1 is truthy, evaluates 2, returns 2)

# `or` returns the first truthy operand, or the last operand.
print(0 or 5)         # 5  (0 is falsy, falls through to 5)
print(7 or 8)         # 7  (7 is truthy, short-circuits)

# A practical use: a cheap default value.
name = ""
display = name or "anonymous"
print(display)        # 'anonymous'
````


The `elif` (short for "else if") clause is optional and can be repeated; `else` is also optional. The conditions are evaluated in order, and only the first matching branch's block runs.

What counts as "truthy"? Any non-zero number, any non-empty container (string, list, dict, tuple), and `True`. Falsy values are `False`, `0`, `0.0`, `""`, `[]`, `{}`, `()`, and `None`. This means you usually write `if items:` rather than `if len(items) > 0:`.

### 15.1 A running example - a login system

🔙 [Back to Table of Contents](#Table-of-Contents)

We'll build up a login validator step by step over the next few sections, motivating loops and functions naturally. Here is the specification.

Every email address has the form:

```
<user-number>@acme.<department>.com
```

- `<user-number>` is exactly four digits. Numbers from `0000` to `8000` are **employees**; numbers from `8001` upward are **contractors**.
- `<department>` is one of `Engineering`, `Sales`, or `Operations`.

For each address we want to print:

1. Whether it is valid, and if not, why.
2. If valid, whether the holder is an employee or contractor, and which department.

**Your turn first.** Try to write a single `if/elif/else` chain that handles one address (start with the cell below). The solution follows.


````python
# Listing 59.
user_1 = '0001@acme.Engineering.com'
departments = ['Engineering', 'Sales', 'Operations']

# Check, in order:
#   1. Is '@' present?
#   2. Are there exactly 3 dots?
#   3. Are there exactly 1 '@'?
#   4. Does it contain 'acme'?
#   5. Does it end with '.com'?
#   6. Is the department one of the valid ones?
# If all checks pass, extract the user number and department, then
# report employee/contractor status.
````


**Solution**

A few things to notice in my version below:

- The validity checks short-circuit: as soon as one fails, the chain stops, so later checks safely assume earlier ones hold.
- I use `s.count(".") != 2` rather than the chained inequality `2 > s.count(".") > 2`. **The chained version is always false**, since no number is both less than 2 and greater than 2, so that check would never trigger. (This is a common bug; if you spotted it, well done.)
- The department check uses `email.lower()` to make the test case-insensitive.
- We split on `.` to get `['0001@acme', 'Engineering', 'com']`, so the department sits at index 1. The user number is the first 4 characters of the whole address.


````python
# Listing 60.
user_1 = '0001@acme.Engineering.com'
valid_departments = ['engineering', 'sales', 'operations']

if '@' not in user_1:
    print(f"Invalid: {user_1!r} is missing the '@' symbol.")
elif user_1.count('.') != 2:
    print(f"Invalid: {user_1!r} should have exactly 2 dots.")
elif user_1.count('@') != 1:
    print(f"Invalid: {user_1!r} should have exactly one '@'.")
elif 'acme' not in user_1.lower():
    print(f"Invalid: {user_1!r} is not an Acme address.")
elif not user_1.lower().endswith('.com'):
    print(f"Invalid: {user_1!r} does not end in '.com'.")
elif user_1.split('.')[1].lower() not in valid_departments:
    print(f"Invalid: {user_1!r} has an unknown department.")
else:
    # All checks passed; safe to extract fields.
    number = int(user_1[:4])
    department = user_1.split('.')[1]
    if number <= 8000:
        print(f"Valid employee account: {user_1}")
        print(f"  Department: {department.capitalize()}")
        print(f"  Employee number: {number}")
    else:
        print(f"Valid contractor account: {user_1}")
        print(f"  Department: {department.capitalize()}")
        print(f"  Contractor number: {number}")
````

## 16. While loops

🔙 [Back to Table of Contents](#Table-of-Contents)

A `while` loop repeats its body as long as a condition stays true. The general shape:

```python
while condition:
    ...
```

`while` loops are right when:

- You don't know in advance how many iterations you need.
- You're waiting for some external event (data arriving, user input, convergence of an optimisation).

Two pitfalls:

1. **Infinite loops.** If the condition never becomes false, the loop never ends. (`Kernel → Interrupt` in Jupyter to stop one.)
2. **Off-by-one errors.** Make sure you update the variables in the condition correctly inside the loop body, and pick `<` vs `<=` deliberately.

Two control statements work inside any loop:

- `break` — exit the loop immediately.
- `continue` — skip the rest of this iteration and go back to the condition.

Let's process several addresses by reusing our validation logic in a loop. We iterate over the list using an index counter (a pattern that will look much cleaner once we meet `for` loops in the next section).

````python
# Listing 61.
users = [
    '0001@acme.Engineering.com',
    '0002@acme.engineering.com',
    '8000@acme.Sales.com',
    '8001@acme.Operations.com',
    '8002@acme.sales.com',
    '8003@acme.sales.com',
    '8004@acme.Engineering.com',
    '8005@acme.engineering.co',
    '8006@acme.com',
    '',
]
valid_departments = ['engineering', 'sales', 'operations']

i = 0
while i < len(users):
    user = users[i]
    if '@' not in user:
        print(f"Invalid: {user!r} missing '@'.")
    elif user.count('.') != 2:
        print(f"Invalid: {user!r} should have exactly 2 dots.")
    elif user.count('@') != 1:
        print(f"Invalid: {user!r} should have exactly one '@'.")
    elif 'acme' not in user.lower():
        print(f"Invalid: {user!r} not an Acme address.")
    elif not user.lower().endswith('.com'):
        print(f"Invalid: {user!r} doesn't end in '.com'.")
    elif user.split('.')[1].lower() not in valid_departments:
        print(f"Invalid: {user!r} unknown department.")
    else:
        number = int(user[:4])
        department = user.split('.')[1].capitalize()
        role = "employee" if number <= 8000 else "contractor"
        print(f"Valid {role} account: {user}  (Dept: {department}, Number: {number})")
    i += 1
````


I used a **ternary expression**, `role = "employee" if number <= 8000 else "contractor"`, to pick a value based on a condition in a single line. Use it when the choice is simple; reach for a full `if/else` block when it's not.

The same shape of loop generalises: process items one at a time, advance a counter, stop when the counter reaches the length. Looping like this is **iteration** (not "recursion", which is when a function calls itself, a different idea entirely).

**Your turn.** Write a `while` loop that simulates a thermostat. Starting at 20.0°C, the room loses 0.05°C per minute. If the temperature drops below 19.0°C, the heating system kicks in for that minute and adds 0.1°C. Simulate 30 minutes.


````python
# Listing 62.
temperature = 20.0       # current room temperature, °C
heat_loss_rate = 0.05    # °C lost per minute
heating_rate = 0.1       # °C added per minute when heating is on
time_in_mins = 0
test_runtime = 30        # minutes

# Your loop here.
````

**Solution**

````python
# Listing 63.
temperature = 20.0
heat_loss_rate = 0.05
heating_rate = 0.1
time_in_mins = 0
test_runtime = 30

while time_in_mins < test_runtime:
    # Passive heat loss this minute.
    temperature -= heat_loss_rate

    # Thermostat kicks in if too cold.
    if temperature < 19.0:
        temperature += heating_rate
        print(f"t={time_in_mins:>2} min  temp={temperature:.3f}°C  (heating on)")
    else:
        print(f"t={time_in_mins:>2} min  temp={temperature:.3f}°C")

    time_in_mins += 1
````

Notice how I formatted the output: `{time_in_mins:>2}` right-aligns the number in a field of width 2, and `{temperature:.3f}` shows three decimal places. These small touches make output much easier to read in a long simulation log.

### 17.2 enumerate - when you need the index too

🔙 [Back to Table of Contents](#Table-of-Contents)

If you want the index *and* the item, don't fall back to `for i in range(len(...))`. Use `enumerate`, which yields `(index, item)` tuples. (You can start the count at any value with `start=...`; the default is 0.)

````python
# Listing 67.
users = ['alice', 'bob', 'carol', 'dave']

for i, user in enumerate(users):
    print(f"User #{i}: {user}")

# Start at 1 for human-friendly numbering.
print("---")
for i, user in enumerate(users, start=1):
    print(f"User #{i}: {user}")
````

### 17.3 zip - iterating two or more iterables in lockstep

🔙 [Back to Table of Contents](#Table-of-Contents)

When you have parallel lists, for example labels and predictions, `zip` pairs them up element by element. `zip` stops at the shortest input.

````python
# Listing 68.
labels      = ['cat', 'dog', 'cat', 'dog']
predictions = ['cat', 'dog', 'dog', 'dog']

correct = 0
for label, pred in zip(labels, predictions):
    matched = label == pred
    correct += matched
    print(f"label={label!r}  pred={pred!r}  match={matched}")

print(f"\nAccuracy: {correct}/{len(labels)} = {correct/len(labels):.0%}")
````

These two idioms, `enumerate` for "index and value" and `zip` for "pair things up", replace the great majority of cases where a beginner would reach for `range(len(...))`. Use them.

### 18.1 Defensive programming and exceptions

🔙 [Back to Table of Contents](#Table-of-Contents)

The function above is fragile. What if the caller passes in something that doesn't start with four digits?

````python
# Listing 73.
def get_user_number(email_address):
    return int(email_address[:4])

# This raises a ValueError.
print(get_user_number('abcd@acme.Engineering.com'))
````

When a piece of code we write may encounter situations it cannot handle, we have two choices:

1. **Let the exception propagate.** This is often correct: don't hide a real problem.
2. **Catch the exception** with a `try / except` block and respond gracefully, for instance by returning a sentinel value or a useful default.

The choice depends on whether the calling code is in a position to do something sensible about the failure.

Here is `get_user_number` written defensively, returning `-1` on any kind of failure.

````python
# Listing 74.
def get_user_number(email_address):
    """Extract the user number from an email address.

    Expects an address of the form '<NNNN>@acme.<department>.com' where
    <NNNN> is four digits. Returns the integer NNNN, or -1 if the address
    does not start with four digits.
    """
    if email_address is None or len(email_address) < 4:
        return -1
    try:
        return int(email_address[:4])
    except ValueError:
        # The first four characters were not parseable as an integer.
        return -1

# Test it.
for addr in ['0001@acme.Engineering.com', 'abcd@acme.Engineering.com', '', None]:
    print(f"{addr!r:35} -> {get_user_number(addr)}")
````

### 18.2 try and except in more detail

🔙 [Back to Table of Contents](#Table-of-Contents)

The `try / except` block lets your code attempt something that might fail and respond to specific failures.

```python
    try:
        # code that might raise
    except SomeError:
        # handle this kind of error
    except (OtherError, AnotherError) as exc:
        # handle several at once, and bind the exception object to `exc`
    else:
        # ran only if NO exception was raised
    finally:
        # always runs, exception or not; useful for cleanup
```

**Catch specific exceptions.** Bare `except:` (or `except Exception:`) catches everything, including bugs you'd rather see crash loudly. Almost always, name the exception you actually expect.

### 18.3 Docstrings

🔙 [Back to Table of Contents](#Table-of-Contents)

Every function you write that does anything non-trivial should have a **docstring**, a string literal placed on the first line of the function body. It documents the function's purpose, its parameters, and its return value. Other tools (and `help(your_function)`) will pick it up automatically.

````python
# Listing 76.
def get_user_department(email_address):
    """Extract the department name from a company email address.

    Expects an address of the form '<NNNN>@acme.<department>.com'.
    Returns the department string if extraction succeeds, else 'Unknown'.

    Parameters
    ----------
    email_address : str
        The address to parse. May be None.

    Returns
    -------
    str
        The department name (preserving its original case), or 'Unknown'
        if the address is malformed.
    """
    if email_address is None or '.' not in email_address:
        return "Unknown"
    parts = email_address.split('.')
    if len(parts) < 2:
        return "Unknown"
    return parts[1]

# `help` reads the docstring.
help(get_user_department)
print()
print(get_user_department('0001@acme.Engineering.com'))
print(get_user_department('garbage'))
````

### 18.5 A classic Python gotcha - mutable default arguments

🔙 [Back to Table of Contents](#Table-of-Contents)

Default argument values are evaluated **once**, when the `def` statement runs, *not* each time the function is called. If the default is a mutable object like a list or dict, all calls that don't supply that argument **share the same object**. This causes one of Python's most famous bugs.

````python
# Listing 78.
# WRONG — shared mutable default.
def add_item_wrong(item, basket=[]):
    basket.append(item)
    return basket

print(add_item_wrong('apple'))    # ['apple']
print(add_item_wrong('banana'))   # ['apple', 'banana']  <- not what we wanted!
print(add_item_wrong('cherry'))   # ['apple', 'banana', 'cherry']
````

The fix: use `None` as the default and create a fresh container inside the function.

````python
# Listing 79.
# Correct — `None` sentinel, fresh list per call.
def add_item(item, basket=None):
    if basket is None:
        basket = []
    basket.append(item)
    return basket

print(add_item('apple'))    # ['apple']
print(add_item('banana'))   # ['banana']
print(add_item('cherry'))   # ['cherry']
````

Remember this. It will catch you out at least once.

### 18.6 Putting it all together

🔙 [Back to Table of Contents](#Table-of-Contents)

Now we wrap the entire login pipeline in a single, well-documented function and run it over a list of test cases.

````python
# Listing 80.
def get_user_number(email_address):
    """Return the integer at the start of an address, or -1 on failure."""
    if email_address is None or len(email_address) < 4:
        return -1
    try:
        return int(email_address[:4])
    except ValueError:
        return -1

def get_user_department(email_address):
    """Return the department name from an address, or 'Unknown' on failure."""
    if email_address is None or '.' not in email_address:
        return "Unknown"
    parts = email_address.split('.')
    if len(parts) < 2:
        return "Unknown"
    return parts[1]

def process_login(user):
    """Validate a login attempt and print the outcome.

    Expects an address of the form '<NNNN>@acme.<department>.com'.
    Numbers up to 8000 are employees; 8001+ are contractors.
    """
    valid_departments = ['engineering', 'sales', 'operations']

    if '@' not in user:
        print(f"Invalid: {user!r} missing '@'.")
        return
    if user.count('.') != 2:
        print(f"Invalid: {user!r} should have exactly 2 dots.")
        return
    if user.count('@') != 1:
        print(f"Invalid: {user!r} should have exactly one '@'.")
        return
    if 'acme' not in user.lower():
        print(f"Invalid: {user!r} not an Acme address.")
        return
    if not user.lower().endswith('.com'):
        print(f"Invalid: {user!r} doesn't end in '.com'.")
        return
    if user.split('.')[1].lower() not in valid_departments:
        print(f"Invalid: {user!r} unknown department.")
        return

    # All checks passed — extract and report.
    number = get_user_number(user)
    department = get_user_department(user)
    role = "employee" if number <= 8000 else "contractor"
    print(f"Valid {role} account: {user}  (Dept: {department.capitalize()}, Number: {number})")

# Run the pipeline over a list of test cases.
users = [
    '0001@acme.Engineering.com',
    '0002@acme.engineering.com',
    '8000@acme.Sales.com',
    '8001@acme.Operations.com',
    '8002@acme.sales.com',
    '8003@acme.sales.com',
    '8004@acme.Engineering.com',
    '8005@acme.engineering.co',
    '8006@acme.com',
    '',
]

for user in users:
    process_login(user)
````

Notice how readable this version is compared with the giant `while` loop earlier. Each function does one thing, has a clear name, and can be tested or reused independently. This is the payoff of breaking work into functions: code that reads more like specification than instruction.

A common pattern in the function above is **early return**: as soon as a precondition fails, return immediately rather than nesting deeper. It keeps the "happy path" flat and the failure paths concise.

## 19. Scope

🔙 [Back to Table of Contents](#Table-of-Contents)

**Scope** is the region of code in which a name is visible. Python resolves names according to the **LEGB rule**: when it sees a name, it searches:

1. **L**ocal — the current function's namespace.
2. **E**nclosing — any enclosing function's namespace (for nested functions).
3. **G**lobal — the namespace of the module (i.e. the notebook / script).
4. **B**uilt-in — names like `print`, `len`, `range`, `int`.

The first match wins. If none match, you get `NameError`.

Scope catches beginners out because **assignment inside a function creates a local name**, even if a same-name variable exists in an outer scope. To explicitly modify an outer name from inside a function, you use the `global` or `nonlocal` keyword. We'll see all of this in code.

Watch the video, then work through the examples. Scope is best learned by running these and watching what happens.

````python
# Listing 81.
from IPython.display import IFrame
IFrame(src='https://www.youtube.com/embed/QVdf0LgmICw', width=560, height=315)
````

**Example 1 — reading an outer variable from inside a function.**

If a function only *reads* a name, Python looks it up via LEGB and finds it in the global scope. No special declaration is needed.

````python
# Listing 82.
a = 5

def show():
    print(f"inside the function, a = {a}")   # reads global `a`

show()
print(f"after the call, a = {a}")
````

**Example 2 — assigning to a name inside a function.**

The moment you write `a = 3` inside a function, Python decides that `a` is **local** to that function. The outer `a` is untouched.

````python
# Listing 83.
a = 5

def assigner():
    a = 3                  # creates a NEW local name `a`
    print(f"inside the function, a = {a}")

assigner()
print(f"after the call, a = {a}")   # global `a` is still 5
````

**Example 3 — same thing, with a parameter.**

A function parameter is a local name. Reassigning it does not affect the caller's variable.

````python
# Listing 84.
name = "Rob"

def change_name(new_name):
    name = new_name        # local `name` — global `name` unaffected
    print(f"inside the function, name = {name}")

print(f"before: name = {name}")
change_name("James")
print(f"after:  name = {name}")
````

**Example 4 — `global` to explicitly modify a global.**

If you really do want to modify a global from inside a function, declare it with `global`. (You usually shouldn't, since passing values in and returning them out is cleaner, but the mechanism exists.)

````python
# Listing 85.
name = "Rob"

def change_name(new_name):
    global name            # `name` here refers to the global one
    name = new_name

print(f"before: name = {name}")
change_name("James")
print(f"after:  name = {name}")
````

**Example 5 — nested functions, default behaviour.**

Each function gets its own local namespace. The inner function's assignment doesn't escape into the outer function's namespace, nor into the module's.

````python
# Listing 86.
x = "a"

def outer():
    x = "b"                # local to `outer`

    def inner():
        x = "c"            # local to `inner`
        print(f"inner: x = {x}")

    inner()
    print(f"outer: x = {x}")

outer()
print(f"global: x = {x}")
````

**Example 6 — `nonlocal` to modify the enclosing function's variable.**

`nonlocal` says "use the `x` from the nearest enclosing function scope". It doesn't reach all the way to the module scope; that's what `global` is for.

````python
# Listing 87.
x = "a"

def outer():
    x = "b"

    def inner():
        nonlocal x         # refers to `outer`'s x
        x = "c"
        print(f"inner: x = {x}")

    inner()
    print(f"outer: x = {x}")   # now 'c' — modified by inner

outer()
print(f"global: x = {x}")      # still 'a'
````

**Example 7 — `global` inside a nested function.**

`global` skips all enclosing function scopes and goes straight to the module level.

````python
# Listing 88.
x = "a"

def outer():
    x = "b"

    def inner():
        global x           # refers to the MODULE-level x
        x = "c"
        print(f"inner: x = {x}")

    inner()
    print(f"outer: x = {x}")   # still 'b' — its local x was never touched

outer()
print(f"global: x = {x}")      # now 'c'
````

**Practical advice.** In day-to-day code, you should very rarely need `global` or `nonlocal`. They make code harder to reason about, since values flow in unpredictable directions. The clean pattern is: **pass values in via parameters, return values out**. Reach for `global`/`nonlocal` only when you have a clear reason.

## 20. Where to go next

🔙 [Back to Table of Contents](#Table-of-Contents)

You now have the foundations of Python. That is genuinely enough to start writing useful programs. But real ML code quickly takes you beyond bare Python: almost no real ML code is "just Python". Here is a quick map of what to learn next, in roughly the order to learn it.

### 20.1 NumPy - numerical arrays

🔙 [Back to Table of Contents](#Table-of-Contents)

NumPy is *the* foundation library of scientific Python. Its core object, the `ndarray`, is a multi-dimensional array of uniform type, with operations that run as fast as C. Almost everything in ML, including model parameters, batches of features, gradients, and predictions, is, at some level, a NumPy array (or something that behaves like one).

Pay attention when you learn it to:

- **Vectorisation.** A Python `for` loop over a million elements is slow; a single NumPy expression operating on the whole array is fast. Most NumPy idioms exist to *avoid* explicit loops.
- **Broadcasting.** The rules by which arrays of different shapes are combined elementwise. Powerful, but easy to get wrong silently.
- **Shapes and axes.** Most NumPy bugs come from arrays of an unexpected shape. Always know the shape of every array you touch.

### 20.2 pandas - tabular data

🔙 [Back to Table of Contents](#Table-of-Contents)

pandas builds on NumPy and gives you `Series` (a labelled 1-D array) and `DataFrame` (a labelled 2-D table, basically a spreadsheet that you can program). It's used for loading, cleaning, exploring, and transforming datasets before feeding them into a model.

### 20.3 matplotlib and seaborn and plotly - plotting

🔙 [Back to Table of Contents](#Table-of-Contents)

You will plot things constantly: training curves, error distributions, decision boundaries, embeddings. Matplotlib is the standard low-level plotting library; seaborn is a high-level wrapper that produces good-looking statistical plots with one line.

### 20.4 scikit-learn - classical ML

🔙 [Back to Table of Contents](#Table-of-Contents)

The de-facto library for classical (non-deep) machine learning: linear and logistic regression, random forests, SVMs, k-means, PCA, model selection, evaluation, pipelines. Its API is famously consistent: once you know how to fit one model, you know how to fit all of them.

### 20.5 PyTorch or JAX or TensorFlow - deep learning

🔙 [Back to Table of Contents](#Table-of-Contents)

For neural networks. PyTorch is the most popular framework in research and increasingly in industry. The mental shift from NumPy is small: tensors behave a lot like arrays, but they remember the operations performed on them so that gradients can be computed automatically.
