## 4. Comparison Operators

🔙 [Back to Table of Contents](#Table-of-Contents)

Comparison operators allow you to compare two values. The result is always a **Boolean**, either `True` or `False`. Booleans are the foundation of all decision-making in Python: every `if` statement, every loop condition, and every filter you write relies on evaluating an expression to `True` or `False`. The table below provides a complete reference for all six comparison operators.

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `>` | Greater than | `7 > 3` | `True` |
| `<` | Less than | `2 < 5` | `True` |
| `>=` | Greater than or equal to | `4 >= 4` | `True` |
| `<=` | Less than or equal to | `3 <= 2` | `False` |
| `==` | Equal to | `8 == 8` | `True` |
| `!=` | Not equal to | `4 != 5` | `True` |

### 4.1 Relational Operators

Relational operators compare the *magnitude* or *ordering* of two values.

The cells below demonstrate each operator. Run each one and confirm the result matches your expectation before moving on.

**Greater than (`>`):** Returns `True` if the left operand is strictly greater than the right.

````python
# Listing 2.
4 > 8    # Is 4 greater than 8? Expected: False
````

**Less than (`<`):** Returns `True` if the left operand is strictly less than the right.

````python
# Listing 3.
4 < 8    # Is 4 less than 8? Expected: True
````

**Greater than or equal to (`>=`):** Returns `True` if the left operand is greater than *or equal to* the right. This means the result is `True` even when both sides are identical.

````python
# Listing 4.
4 >= 4   # Is 4 greater than or equal to 4? Expected: True
````

**Less than or equal to (`<=`):** The mirror of `>=`.

````python
# Listing 5.
3 <= 3   # Is 3 less than or equal to 3? Expected: True
````

### 4.2 Equality and Inequality

**Equality (`==`):** Tests whether two values are *equal*. Note the double equals sign: a common mistake for beginners is to write a single `=`, which is the *assignment* operator and has a completely different meaning.

````python
# Listing 6.
8 == 8   # Is 8 equal to 8? Expected: True
````

````python
# Listing 7.
4 == 8   # Is 4 equal to 8? Expected: False
````

Equality can also be used with strings. Python is **case-sensitive**, so `'data'` and `'Data'` are treated as different strings.

````python
# Listing 8.
'data' == 'data'    # Identical strings: Expected: True
````

````python
# Listing 9.
'Data' == 'data'    # Different capitalisation: Expected: False
````

**Inequality (`!=`):** Returns `True` when two values are *not* equal. The `!` symbol means 'not' in many programming languages.

````python
# Listing 10.
2 != 4   # Is 2 not equal to 4? Expected: True
````

````python
# Listing 11.
4 != 4   # Is 4 not equal to 4? Expected: False
````

## 5. Logical Operators

🔙 [Back to Table of Contents](#Table-of-Contents)

Logical operators allow you to combine multiple **Boolean expressions** (statements that evaluate to either `True` or `False`, such as *"flipper length is below 200"* or *"body mass exceeds 4000 g"*) into a single condition. This is essential whenever a decision depends on more than one condition, like classifying a penguin as Species A only if its flipper length is below 200 *and* its body mass is below 4000 g.

Python has three logical operators: `and`, `or`, and `not`.

### 5.1 The `and` Operator

The `and` operator returns `True` only when **both** conditions are `True`. If *either* side is `False`, the whole expression is `False`.

The cells below illustrate this. Look at each cell, try to work out what it will do, and then execute it to confirm your understanding.

Both conditions are `True`, so the result is `True`.

````python
# Listing 13.
(4 == 4) and ('data' == 'data')   # True AND True → Expected: True
````

In the next example, the second condition is `False` (case mismatch), so even though the first part is `True`, the result is `False`.

````python
# Listing 14.
(4 == 4) and ('Data' == 'data')   # True AND False → Expected: False
````

### 5.2 The `or` Operator

The `or` operator returns `True` when **at least one** condition is `True`. It only returns `False` when *both* sides are `False`.

In the code sample below, the first condition is `True`, so the whole expression is `True`, even though the second condition is `False`.

````python
# Listing 15.
(4 == 4) or ('Data' == 'data')    # True OR False → Expected: True
````

### 5.3 The `not` Operator

The `not` operator **inverts** a Boolean value: `True` becomes `False` and `False` becomes `True`. It is a *unary* operator, meaning it acts on a single value rather than combining two.

````python
# Listing 16.
not True    # Inverts True → Expected: False
````

````python
# Listing 17.
not (4 == 8)    # 4 == 8 is False; not False → Expected: True
````

### 5.4 Logical Operator Reference Table

| Operator | Description | Example | Result |
|----------|-------------|---------|--------|
| `and` | Both conditions must be `True` | `(5 > 3) and (2 < 4)` | `True` |
| `and` | One condition is `False` | `(5 > 3) and (4 < 2)` | `False` |
| `or` | At least one condition is `True` | `(5 > 3) or (4 < 2)` | `True` |
| `or` | Both conditions are `False` | `(1 > 3) or (4 < 2)` | `False` |
| `not` | Inverts the Boolean value | `not (5 > 3)` | `False` |

A **truth table** shows all possible combinations of Boolean values (`True` or `False`) for inputs and the resulting output of a logical operation. For example, the `and` operator only returns `True` if *both* inputs are `True`, while the `or` operator returns `True` if *at least one* input is `True`. These truth tables can help you better understand how logical operators behave.

**Truth table: `and`**

| A | B | A `and` B |
|---|---|-----------|
| `True` | `True` | `True` |
| `True` | `False` | `False` |
| `False` | `True` | `False` |
| `False` | `False` | `False` |

**Truth table: `or`**

| A | B | A `or` B |
|---|---|----------|
| `True` | `True` | `True` |
| `True` | `False` | `True` |
| `False` | `True` | `True` |
| `False` | `False` | `False` |

## 6. Conditional Statements

🔙 [Back to Table of Contents](#Table-of-Contents)

Comparison and logical operators produce Boolean values. **Conditional statements** use those values to decide which block of code to run. This is the core mechanism by which programs make decisions and via which you, as a programmer, can control the execution of an application.

### 6.1 The `if` Statement

The simplest conditional statement runs a block of code only when a condition is `True`. The syntax is:

```python
if <condition>:
    <code to run when condition is True>
```

Notice that the code block is **indented**: Python uses indentation (typically four spaces) to define which lines belong to the `if` block. This is different from languages like Java or C++ that use curly braces `{}`. Incorrect indentation is one of the most common errors for new Python programmers.

In the first cell below, the condition `4 == 4` is `True`, so the `print` statement runs.

````python
# Listing 18.
if 4 == 4:
    print("True")   # This line runs because the condition is True
````

In the next cell, the condition `4 == 8` is `False`, so the indented block is skipped entirely. Nothing is printed.

````python
# Listing 19.
if 4 == 8:
    print("True")   # This line is skipped because 4 == 8 is False
````

You can use any comparison operator in an `if` condition. Here `4 < 8` is `True`, so the block runs.

````python
# Listing 20.
if 4 < 8:
    print("True")   # 4 < 8 is True, so this runs
````

### 6.2 The `else` Clause

An `else` clause provides a block of code to run when the `if` condition is `False`. It acts as a 'catch-all': if none of the preceding conditions were met, run this instead.

```python
if <condition>:
    <run when True>
else:
    <run when False>
```

Since `4 == 8` is `False`, the `else` block runs and prints `"False"`.

````python
# Listing 23.
if 4 == 8:
    print("True")    # Skipped — 4 == 8 is False
else:
    print("False")   # This runs instead
````

### 6.3 The `elif` Clause

When you need to check **more than two** possibilities, use `elif` (short for 'else if'). Python evaluates each condition in order and runs the first block whose condition is `True`. If no condition is `True`, the `else` block (if present) runs.

```python
if <condition 1>:
    <run when condition 1 is True>
elif <condition 2>:
    <run when condition 2 is True>
else:
    <run when no condition is True>
```

You can chain as many `elif` clauses as you need. The cell below demonstrates a three-branch conditional.

````python
# Listing 24.
a = 8
if a == 4:
    print("First branch: a equals 4")    # Skipped — False
elif a == 8:
    print("Second branch: a equals 8")   # This runs — True
else:
    print("No conditions were met")
````

## 7. Loops

🔙 [Back to Table of Contents](#Table-of-Contents)

A **loop** repeats a block of code multiple times. Without loops, you would have to write the same code out over and over again, once for each repetition. Python provides two types of loop: `for` and `while`.

### 7.1 The `for` Loop

A `for` loop iterates over the items in a **sequence** (a list, a tuple, a string, or any other iterable object) and runs the indented block below the ```for``` once for each item.

```python
for <variable> in <sequence>:
    <code to run for each item>
```

The variable name after `for` is your choice; it is just a temporary name that holds the current item during each iteration. Common choices are `item`, `element`, `i`, or a name that describes what the sequence contains.

Let's try this out for ourselves. First, create a list to iterate over.

````python
# Listing 27.
# Create a list of integers to iterate over
new_seq = [1, 2, 3, 4, 5]
````

Iterate over the list using the variable name `item`, printing each element.

````python
# Listing 28.
# The variable 'item' takes the value of each list element in turn
for item in new_seq:
    print(item)
````

The variable name can be anything. Here we use `element`; the behaviour is identical.

````python
# Listing 29.
# 'element' works just as well as 'item' — the name is arbitrary
for element in new_seq:
    print(element)
````

You can perform operations on the loop variable inside the block. Here we print the square of each element.

````python
# Listing 30.
# Perform a calculation on each element during the loop
for element in new_seq:
    print(element * element)   # Print the square of each number
````

The `break` statement exits the loop immediately, before it has finished all iterations. The loop below stops as soon as `item` equals 4, so 5 is never printed.

````python
# Listing 31.
# 'break' stops the loop when a condition is met
for item in new_seq:
    print(item)
    if item == 4:    # When item reaches 4, exit the loop
        break        # 5 will never be printed
````

### 7.2 The `while` Loop

A `while` loop continues repeating as long as its condition is `True`. You use it when you do not know in advance how many iterations are needed.

```python
while <condition>:
    <code to repeat>
```

> ⚠️ **Important:** You must update the loop variable inside the block. If the condition never becomes `False`, the loop runs forever (an **infinite loop**) and you will need to interrupt the kernel. In the cell below, `i = i + 1` ensures the loop eventually terminates.

````python
# Listing 32.
i = 1                                           # Initialise the counter
while i < 8:                                    # Continue while i is less than 8
    print('This is loop number {}'.format(i))   # Print the current loop number
    i = i + 1                                   # Increment — essential to avoid an infinite loop
````

### 7.3 The `range()` Function

The function `range()` generates a sequence of integers (whole numbers, e.g. 1, 2, 3 etc). It is most commonly used to control how many times a `for` loop runs, but it can also be used on its own to inspect the sequence it produces.

The function `range()` accepts one, two, or three arguments. The table below summarises how each form behaves.

| Call | Produces | Notes |
|------|----------|-------|
| `range(n)` | `0, 1, 2, ..., n-1` | Start defaults to 0, step defaults to 1 |
| `range(start, stop)` | `start, start+1, ..., stop-1` | Stop is **exclusive**, meaning the range includes all values *up to, but not including*, the stop value. |
| `range(start, stop, step)` | `start, start+step, ...` | Custom increment |

On its own, `range()` returns a `range` object. Cast it to a `list` to see its contents. Take a look at the cells below, try to understand them and then execute them. You'll see how range behaves for different input parameters.

````python
# Listing 35.
# range(n): generates integers from 0 up to (but not including) n
list(range(8))    # Produces [0, 1, 2, 3, 4, 5, 6, 7]
````

````python
# Listing 36.
# range(start, stop): generates integers from start up to (but not including) stop
list(range(4, 8))    # Produces [4, 5, 6, 7]
````

````python
# Listing 37.
# range(start, stop, step): generates integers with a custom increment
list(range(4, 20, 2))    # Produces [4, 6, 8, ..., 18] — every second integer
````

## 8. List Comprehensions

🔙 [Back to Table of Contents](#Table-of-Contents)

A **list comprehension** is a compact, single-line way to create a new list by performing an operation on each item in an existing sequence (like a list or range). Instead of writing a loop to process each item one by one and append it to a new list, you can do it all in one line. Say you have a list of numbers and want to square each one: a list comprehension lets you generate the new list directly by specifying the operation (`x ** 2`) and the source of the items (`for x in [1, 2, 3, 4]`) in a single expression. It's more concise than the loop version and arguably easier to read once you're used to the syntax, and it's particularly handy for simple transformations or filtering tasks, like creating a list of squared numbers or pulling out only the even numbers from a list.

Instead of writing:
```python
squares = []
for x in [1, 2, 3, 4]:
    squares.append(x ** 2)
```

You can achieve the same result in one line:

```python
squares = [x ** 2 for x in [1, 2, 3, 4]]
```

The syntax for a list comprehension is:

```python
new_list = [<expression> for <variable> in <sequence>]
```

You can also add an optional filter:

```python
new_list = [<expression> for <variable> in <sequence> if <condition>]
```

The example below first shows the traditional `for` loop approach, then the equivalent list comprehension, so you can see exactly how they relate.

````python
# Listing 40.
# Starting list
my_list = [1, 2, 3, 4, 5, 6, 7, 8]
````

**Traditional approach:** create an empty list, loop over `my_list`, and append each squared value.

````python
# Listing 41.
# Traditional for-loop approach to creating a list of squares
out = []                        # Start with an empty list
for num in my_list:
    out.append(num ** 2)        # Square each number and add it to the list
print(out)
````

**List comprehension equivalent:** the same result in a single line.

````python
# Listing 42.
# List comprehension: [expression for item in sequence]
squares = [num ** 2 for num in my_list]    # Square every element in one line
print(squares)
````

**Comprehension with a filter:**

You can also add a condition to a list comprehension to *selectively* include items in the new list. For example, to generate a list of squares **only for even numbers**, you can add an `if` statement inside the comprehension. This means the operation (`num ** 2`) is only applied to items in `my_list` that meet the condition (`num % 2 == 0`). The result is a new list containing only the squares of the even numbers from the original list.

````python
# Listing 43.
# List comprehension with a filter condition
even_squares = [num ** 2 for num in my_list if num % 2 == 0]
#               ^^^^^^^^   ^^^^^^^^^^^^^^^^^^  ^^^^^^^^^^^^^
#               expression  iterates over list   keep only even numbers
print(even_squares)
````

## 12. Lines as Decision Boundaries

🔙 [Back to Table of Contents](#Table-of-Contents)

In Notebook 1 I introduced the equation of a straight line and measured how well a line fits data. In this section we deepen that intuition and connect it directly to classification. By the end of this notebook you will be able to:

- Use a **single vertical boundary** ($x = k$) to classify 1-D data.
- Use the **general line equation** $y = mx + c$ as a 2-D decision boundary.
- Combine **two lines** to create richer boundaries that separate three classes.
- Appreciate why simple lines are sometimes insufficient, motivating more complex ML models.
- Quantify how well a boundary works using **classification error**.
- Use Python to **search for a better boundary** by minimising that error.

### 12.1 The Simplest Boundary - x = k

The line $x = k$ divides a 1-D measurement into two regions: everything to the left ($x < k$) and everything to the right ($x \geq k$). We can use this directly as a classifier.

**Example:** Suppose we are trying to distinguish two species of penguin based solely on their flipper length. We hypothesise that a single threshold (a vertical line) can separate them. The **classification rule** is:

$$
\text{if } x < k \Rightarrow \text{Species A}; \quad \text{else} \Rightarrow \text{Species B}
$$

This is, in essence, what a **decision stump** does (the simplest possible decision tree, more on those in future notebooks).

For the next example, we generate **synthetic** data so that we can control exactly what our problem looks like and focus on the ideas rather than data wrangling (I'll often do this).

The cell below generates the data and plots it with a candidate boundary at $x = 195$. Note that `np.random.default_rng(0)` creates a random number generator with a fixed **seed**, which ensures the same data is produced every time the cell is run. This particular cell contains interactive code, which makes the underlying Python quite long if you go and look at it directly, so I've embedded it in a separate file rather than pasting it into the notebook. I'll often do this for the more involved interactive figures; you're free to open the file and look at the code if you're curious.

````python
# Listing 79.
%matplotlib widget
from visualisations.Figure_1 import show
show()
````

The initial boundary of `x=195` does a reasonable job, but some points are on the wrong side. We can quantify this using the **misclassification error rate**:

$$
\text{error rate} = \frac{\text{number of misclassified points}}{\text{total number of points}}
$$

An error rate of 0.0 means perfect classification; 1.0 means every point is classified wrong.

Now that I have a way of measuring the misclassification rate, I'm going to write my own function that counts how many mistakes a specific split point would make if I were to use it on flipper data. I've created that function in the cell below. All it does is use a decision `threshold` to classify each data point: any point with an `x` value less than the threshold is classified as Species A, and any point with an `x` value greater than or equal to the threshold is classified as Species B. Sometimes the chosen threshold will get this right, sometimes it will be wrong. But since we know the correct species for each penguin, this function can calculate how many mistakes would be made on the data. It compares predictions made using the simple threshold against the true species labels, and returns the proportion of misclassified points as the error rate.

As you can see, it uses the `if`/`else` statements and `for` loops you practised in Sections 6 and 7. So that wasn't a waste of time after all!!

````python
# Listing 80.
def classify_1d(species_a, species_b, threshold):
    '''
    Classifies 1-D data using a vertical threshold boundary.
    Rule: x < threshold → predicted Species A; x >= threshold → predicted Species B.

    Parameters
    ----------
    species_a  : list of float  — flipper lengths for Species A
    species_b  : list of float  — flipper lengths for Species B
    threshold  : float          — the x-position of the decision boundary

    Returns
    -------
    float : misclassification error rate (proportion of incorrectly classified points)
    '''
    errors = 0
    total  = len(species_a) + len(species_b)

    # Species A points should be LEFT of the threshold (x < threshold)
    for x in species_a:
        if x >= threshold:    # Right of boundary → misclassified
            errors += 1

    # Species B points should be RIGHT of the threshold (x >= threshold)
    for x in species_b:
        if x < threshold:     # Left of boundary → misclassified
            errors += 1

    return errors / total     # Return fraction of mistakes

# Fixed seed (0) means every user sees identical data points.
rng = np.random.default_rng(0)

# Species A: shorter flippers centred around 180 mm
species_a = np.array(rng.normal(loc=180, scale=8, size=40))
# Species B: longer flippers centred around 210 mm
species_b = np.array(rng.normal(loc=210, scale=8, size=40))

# Test the boundary at x = 195
threshold  = 195
error_rate = classify_1d(species_a, species_b, threshold)
print(f'Threshold: x = {threshold} mm')
print(f'Misclassification rate: {error_rate:.1%}')
print(f'Accuracy:               {1 - error_rate:.1%}')
````

In any case, searching manually like this is a huge waste of time. Instead of guessing, we can **automate the search**: loop over many candidate thresholds and record the error rate for each one. This is a simple form of **model selection**, where we systematically evaluate many configurations and choose the best one.

The cell below does this for us. It searches every integer threshold from 160 to 229 using our ```classify_1d``` function from above and plots the **error landscape**: a visual representation of how the misclassification error rate changes as you adjust the decision threshold (or any other parameter) in a classification model. It shows the ups and downs of error values across a range of possible thresholds, helping you spot where the model performs best (lowest error) or worst (highest error). For a 1-D classifier like the one above, this means plotting the error rate for every possible threshold value, which reveals the optimal split point where the fewest mistakes are made.

Run the code below to search for the best threshold value and plot the error landscape.

````python
# Listing 82.
# ── Threshold search ───────────────────────────────────────────────────────
# list(range(160, 230, 1)) creates a list of every integer from 160 to 229
candidates  = list(range(160, 230, 1))
error_rates = []

# Evaluate the classifier at each candidate threshold
for k in candidates:
    rate = classify_1d(species_a, species_b, k)
    error_rates.append(rate)

# Find the threshold that produced the lowest error
best_idx       = error_rates.index(min(error_rates))   # Index of the minimum value
best_threshold = candidates[best_idx]
best_error     = error_rates[best_idx]

# ── Plot error rate vs threshold ────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))
fig.canvas.header_visible = False
fig.canvas.resizable = True
fig.canvas.toolbar_visible = True
fig.canvas.toolbar_position = 'right'

ax.plot(candidates, error_rates, color='steelblue', linewidth=2)
ax.axvline(x=best_threshold, color='red', linestyle='--',
           label=f'Best threshold: x = {best_threshold} mm  (error = {best_error:.1%})')
ax.set_xlabel('Threshold value k  (mm)')
ax.set_ylabel('Misclassification error rate')
ax.set_title('Figure 2: Error rate as a function of the decision threshold')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f'Best threshold found: x = {best_threshold} mm')
print(f'Error rate at best threshold: {best_error:.1%}')
````

We see here that the search process found the optimal threshold value for us, `x=192`. In order to carry out this search, I had to bound the search process by restricting the possible threshold values to a range between `160` and `230`, and testing each value in steps of `1`. This restricted search space makes it quick and efficient to find a good solution, as there are only 71 possible values to test (230 - 160 + 1). However, if I were to extend the range, for example, from 100 to 300, the number of values to test would increase significantly, making the search slower. Similarly, using a smaller step size (e.g., 0.1 instead of 1) would also increase the number of values to test, providing a more precise result but at the cost of additional computation time.

> **Key insight:** Looping over possible values of a parameter and picking the one that minimises error is fundamentally what many machine learning training algorithms do, just with far more sophisticated mathematics under the hood.

### 12.2 A 2-D Decision Boundary - y = mx + c

When data has two features (e.g. both flipper length and body mass) a single vertical line is no longer sufficient, so we need a line in 2-D space instead. You'll recall the general line equation from Notebook 1:

$$
y = mx + c
$$

A point $(x_i, y_i)$ can be classified according to **which side of the line it falls on**:

| Region | Condition | Predicted class |
|--------|-----------|----------------|
| Above the line | $y_i > mx_i + c$ | Class B |
| Below the line | $y_i < mx_i + c$ | Class A |


The cell below generates two 2-D clusters and draws a candidate boundary. This code sample looks complex, **because it is**: I've created an interactive version for you to experiment with, so don't be put off by it; you aren't expected to understand every line. Rather than making you change the values of `m` and `c` directly, I let you adjust the angle and offset of the line, and the code recalculates the equation of the line and the error rate for you automatically. This is a simplified, educational setup, not something you'd typically build this way in the real world.

````python
# Listing 83.
%matplotlib widget
from visualisations.Figure_3 import show
show()
````

Now let's write a classifier that uses a line like this and measures its accuracy. The function uses `zip()` to iterate over the x- and y-coordinates of each point simultaneously, allowing us to check each point's position relative to the decision boundary `y = mx + c`. Here's how the code below works, assuming a line with the parameters `m` (slope) and `c` (intercept):

- For **Class A points**, the classifier checks if each point lies *below* the line (`yi < m*xi + c`). If a point is above or on the line, it's misclassified.
- For **Class B points**, the classifier checks if each point lies *above* the line (`yi > m*xi + c`). If a point is below the line, it's misclassified.
- The function then returns the proportion of misclassified points as the error rate.

By default I use `m=40` and `c=-3500` but you can tweak this as you see fit.

````python
# Listing 84.
def classify_2d(xa, ya, xb, yb, m, c):
    '''
    Classifies 2-D data using y = mx + c as the decision boundary.
    Points below the line are predicted Class A; points above are predicted Class B.

    Parameters
    ----------
    xa, ya : arrays — coordinates of Class A points
    xb, yb : arrays — coordinates of Class B points
    m      : float  — slope of the boundary line
    c      : float  — intercept of the boundary line

    Returns
    -------
    float : misclassification error rate
    '''
    errors = 0
    total  = len(xa) + len(xb)

    # zip(xa, ya) pairs up matching x and y coordinates so we can iterate both at once
    # Class A should be BELOW the line (yi < m*xi + c)
    for xi, yi in zip(xa, ya):
        predicted_y = m * xi + c
        if yi >= predicted_y:    # Above or on the line → misclassified
            errors += 1

    # Class B should be ABOVE the line (yi > m*xi + c)
    for xi, yi in zip(xb, yb):
        predicted_y = m * xi + c
        if yi < predicted_y:     # Below the line → misclassified
            errors += 1

    return errors / total

# Fixed seed so every user sees the same point clouds.
rng = np.random.default_rng(1)

# Create the data used in the plot
class_a_x = rng.normal(loc=180, scale=8,    size=50)
class_a_y = rng.normal(loc=3500, scale=300, size=50)
class_b_x = rng.normal(loc=210, scale=8,    size=50)
class_b_y = rng.normal(loc=5000, scale=300, size=50)
    
error = classify_2d(class_a_x, class_a_y, class_b_x, class_b_y, m=40, c=-3500)
print(f'Line: y = 40x - 3500')
print(f'Misclassification rate: {error:.1%}')
print(f'Accuracy:               {1 - error:.1%}')
````

### 12.4 When Lines Are Not Enough

Consider a dataset where Class A sits in the middle and Class B forms a **ring** around the outside. No straight line can separate these two groups regardless of how it is positioned or rotated. This is the key motivation for more sophisticated models (SVMs with non-linear kernels, decision trees, neural networks) that can learn **curved** boundaries.

The figure below shows the failure of a linear boundary and the non-linear boundary that solves the problem.

This is not simply a case of choosing the wrong values of m and c, it is a deeper, structural problem. A straight line divides the plane into exactly two half-planes: everything to one side, and everything to the other. But a ring-shaped dataset cannot be separated that way; the inner class is *surrounded*, not *beside* the outer class. No half-plane can capture that.

The code below constructs the two classes using **polar coordinates**, a natural way to describe circular structure. Each point is defined by a distance from the centre (`r`) and an angle (`ang`), then converted to the Cartesian `(x, y)` coordinates that Matplotlib expects:

```
x = r · cos(angle)
y = r · sin(angle)
```

Class A uses small radii (0–1.5), placing all its points close to the origin. Class B uses larger radii (2.5–4.0), forming the surrounding ring. The gap between radius 1.5 and 2.5 is precisely what the ideal circular boundary exploits: it sits at radius 2.0, cleanly in the middle of that gap.

The two panels of Figure 5 tell the full story side by side: on the left, the best linear boundary we can draw still slices through both classes; on the right, a circle of radius 2.0 separates them perfectly. Same data, very different outcomes. The only thing that changed is the *shape* of the boundary.

| Classes form curved or complex shapes (*not linearly separable*) | SVM (RBF kernel), Decision Trees, Random Forests, Neural Networks |

### 12.5 Error Minimisation - Searching for the Best Line

In the 1-D example we searched over candidate thresholds and picked the best. The same idea extends to 2-D: we loop over candidate **slopes** (and later **intercepts**) and record the misclassification error for each combination.

This concept, **evaluating a loss function over a parameter space and finding its minimum**, is the heart of all supervised machine learning.

The approach here is a form of **brute-force search**: we do not derive the best line mathematically; instead we simply try many candidates and keep the one that makes the fewest mistakes. This is computationally expensive for large parameter spaces, but it has one important virtue: it makes the optimisation process completely transparent. There is no calculus, no matrix algebra, just *try every option, measure the error, pick the best*.

This same logic underlies far more sophisticated algorithms. Gradient descent, the engine behind neural networks, is essentially a smarter version of this search: rather than trying every possible value, it follows the slope of the error curve downhill towards the minimum. Understanding the brute-force version first makes gradient descent much easier to grasp when you encounter it later.

#### The loss function

The function `error_rate_2d` defined in the cell below is our **loss function**, a number that summarises how badly a given line performs. It counts two types of mistake:

- **False positives**: Class A points that fall *above* the boundary (predicted Class B)
- **False negatives**: Class B points that fall *below* the boundary (predicted Class A)

It then divides the total number of mistakes by the total number of points to give a rate between 0 (perfect) and 1 (every point wrong). Minimising this rate is the optimisation objective.

#### Part A: Searching over slope only

The cell below fixes the intercept at $c = -3500$ and searches over slopes from 10.0 to 99.5 in steps of 0.5, a total of 178 candidate lines. This is a *partial* search: we only vary one parameter at a time, holding the other fixed. It demonstrates the principle clearly but is not guaranteed to find the globally optimal line, because the best slope when $c = -3500$ may not be the best slope at a different intercept.

The two-panel figure below shows exactly why this search is useful. **You do not need to understand every line of the code**; focus instead on what the plots are telling you.

#### The left panel: the loss landscape

The left panel plots the **error rate** (the fraction of points misclassified) on the vertical axis against the **slope** of the boundary line on the horizontal axis. Each point on this curve answers the question: *"if I used this slope, how many mistakes would my classifier make?"*

The shape of this curve is called the **loss landscape** (or error landscape). The goal of training a machine learning model is always to find the lowest point on this landscape, the parameter value that produces the fewest mistakes. Here the landscape is a simple 1-D curve and the minimum is visually obvious. In real models, the landscape extends across millions of parameter dimensions simultaneously and cannot be visualised directly, but the principle is identical.

The **red dashed line** marks the slope that minimises the error rate. The **black dot** shows where your current boundary sits on the landscape; drag the sliders and watch it move.

#### The right panel: the boundary over the data

The right panel shows the actual boundary line drawn over the scatter plot, so you can connect the abstract error number on the left to a concrete geometric object on the right. When the black dot on the left sits at the bottom of the curve, the line on the right should visually separate the two classes as cleanly as possible.

#### Part B: Full 2-D grid search over both slope and intercept

The search above was *partial*: we fixed $c = -3500$ and only varied $m$. You could change the line via the sliders, but the underlying search itself only ever explored slopes, with the intercept held at $c = -3500$ throughout. The truly optimal line could have a different intercept entirely. Below, we run a **full 2-D grid search**: every combination of $m$ and $c$ within a reasonable range is evaluated.

Note how the nested `for` loop directly mirrors the concept of a 2-D parameter space: the outer loop iterates over slopes, and the inner loop iterates over intercepts.

#### What the grid search is actually doing

In Part A, the parameter space was a single line: one axis for slope, one curve of error values. In Part B, the parameter space becomes a **grid**: every slope paired with every intercept. The code evaluates:

$$101 \text{ slopes} \times 101 \text{ intercepts} = 10{,}201 \text{ candidate lines}$$

For each of those lines it asks the same question as before: *how many points does this line misclassify?* The pair $(m, c)$ that produces the lowest error rate across the entire grid is the globally optimal boundary, at least within the resolution of the search.

This is still brute force. A finer grid (smaller step sizes) would find a more precise answer but take longer to run. A coarser grid is faster but may miss the true minimum. This **resolution vs. cost trade-off** is a recurring theme in numerical optimisation and one of the reasons more efficient methods (gradient descent, Newton's method) were developed.

#### What the plots show

The left panel now shows something subtly different from Part A. Rather than the error at a fixed $c$, each point on the curve shows the **best error achievable at that slope across all intercepts**: in other words, for each slope, we have already optimised over $c$ and are only reporting the minimum. This collapses the 2-D grid down to a 1-D curve so it can be visualised, but the red dashed line now marks the truly globally optimal slope rather than just the best slope at one fixed intercept.

Compare the minimum error on this curve to the minimum from Part A: it should be equal or lower, because we have given the search strictly more freedom.

The right panel shows the winning boundary over the data. Notice that the intercept $c$ is now a large number (in the tens of thousands); this is not unusual. When features like body mass are measured in the thousands, the intercept must be correspondingly large to position the line correctly in that feature space. This is one reason why **feature scaling** (normalising inputs to a common range) is standard practice in machine learning: it keeps parameter values at a manageable scale and makes optimisation behave more predictably.

**What the search is doing, step by step:**

`m_values` covers slopes from $-200$ to $+200$. Negative slopes are now included because there is no reason in principle why the optimal separating line must slope upward.

`c_values` covers intercepts from $-50{,}000$ to $+50{,}000$. The wide range is necessary because the y-axis (body mass) sits in the thousands, so a useful line will have a large intercept.

The nested `for` loop evaluates `error_rate_2d` for every `(m, c)` pair. If there are, say, 100 slopes and 100 intercepts, that is 10,000 individual error evaluations. The pair that produces the fewest mistakes is stored as `(best_m, best_c)`.

**Why this matters: the curse of dimensionality in parameters**

| Number of parameters | Grid size (100 values each) |
|----------------------|-----------------------------|
| 1 | 100 evaluations |
| 2 | 10,000 evaluations |
| 3 | 1,000,000 evaluations |
| 10 | $10^{20}$ evaluations, computationally impossible |

A real machine learning model that is trying to separate data, e.g. logistic regression, may have hundreds of parameters. Exhaustive grid search becomes impossible very quickly. This is precisely why the algorithms you will meet from Notebook 5 onwards use smarter strategies, such as **gradient descent**, closed-form solutions, and probabilistic updates, rather than testing every possible combination.
