## 5. Python Numbers and Arithmetic

🔙 [Back to Table of Contents](#Table-of-Contents)


Numbers are the most fundamental data type in any programming language, and Python is no exception. In machine learning, almost everything eventually reduces to numerical computation, so it's essential to be comfortable with how Python handles numbers.

### 5.1 Integers and Floats

Python distinguishes between two main numeric types:

| Type | Description | Example |
|------|-------------|---------|
| `int` | A whole number (no decimal point). Can be positive, negative, or zero. | `5`, `-3`, `0` |
| `float` | A number with a decimal point (floating-point number). | `0.5`, `3.14`, `-2.7` |

The distinction matters because some operations behave differently depending on the type. For example, in Python 3 (the current up-to-date version of Python), dividing two integers with the division symbol `/` always returns a `float`. Run the cells below to see both types:


````python
# Listing 7.
# An integer literal — a whole number with no decimal point
print(5)
print(5/5) # Will return a float.
````

You can see from running the cell above that 5 is an integer (no decimal point), but using the division symbol on an integer returns a float. You don't need to run the cell immediately below this one, it's just there to show you an example of a float.

````python
# Listing 8.
# A float literal — a number that includes a decimal point
0.5
````

#### Checking the Type of an Object

Python's built-in `type()` function returns the type of any object. This is very useful for debugging and for understanding what kind of data you are working with. Run the cell below to see what the function returns.


````python
# Listing 9.
# type() is a built-in function that returns the data type of its argument.
# Here we check the type of the integer 5 and the float 0.5.
print(type(5))     # Will print: <class 'int'>
print(type(0.5))   # Will print: <class 'float'>
````

The function is explaining to me, in its own way, that `5` belongs to the int class (an integer) and `0.5` belongs to the float class (a float). If you were to run this function on me, my type would be `<class 'human'>`, well, hopefully.

### 5.2 Arithmetic Operators

Python supports all standard arithmetic operations. The table below summarises each operator:

| Operator | Operation | Example | Result |
|----------|-----------|---------|--------|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Subtraction | `5 - 3` | `2` |
| `*` | Multiplication | `5 * 3` | `15` |
| `/` | Division (always returns float) | `10 / 2` | `5.0` |
| `//` | Floor division (returns integer) | `10 // 3` | `3` |
| `**` | Exponentiation (power) | `2 ** 4` | `16` |
| `%` | Modulus (remainder after division) | `8 % 3` | `2` |

The **modulus** operator (`%`) is particularly useful in programming, for example, to check whether a number is even (`number % 2 == 0`) or to wrap values around a range.

Run the cell below to see all operators in action:


````python
# Listing 10.
# Demonstrating all arithmetic operators in Python

print("Addition:          5 + 5  =", 5 + 5)
print("Subtraction:       10 - 3 =", 10 - 3)
print("Multiplication:    5 * 5  =", 5 * 5)
print("Division:          10 / 2 =", 10 / 2)   # Note: returns a float
print("Floor division:    10 // 3 =", 10 // 3)  # Rounds down to integer
print("Exponentiation:    2 ** 4 =", 2 ** 4)
print("Modulus:           8 % 2  =", 8 % 2)    # Remainder: 8 / 2 = 4 remainder 0
print("Modulus (odd):     9 % 2  =", 9 % 2)    # Remainder: 9 / 2 = 4 remainder 1
````

See how the modulus operator gives you the remainder after division?

### 5.3 Operator Precedence (Order of Operations)

Python follows the standard mathematical rules for the **order of operations**, sometimes remembered by the acronym **BODMAS**, **BIDMAS** or **PEMDAS**. This means if you give Python some mathematical description or definition to evaluate, it will do so in the following order:

1. **Brackets** / **Parentheses** `()`
2. **Exponents** / **Powers** `**`
3. **Multiplication** and **Division** `*`, `/`, `//`, `%`
4. **Addition** and **Subtraction** `+`, `-`

This means Python evaluates `*` and `/` **before** `+` and `-`, unless you use parentheses to override the order.

Run the two cells below and compare the results: both use the same numbers, but the parentheses change the order and therefore the answer.


````python
# Listing 11.
# Without parentheses: Python evaluates 8 * 8 first (= 64), then 2 + 64 + 8
# Expected result: 74
result_no_parentheses = 2 + 8 * 8 + 8
print("2 + 8 * 8 + 8  =", result_no_parentheses)
````

And with the parentheses...

````python
# Listing 12.
# With parentheses: Python evaluates (2 + 8) = 10, then (8 + 8) = 16, then 10 * 16
# Expected result: 160
result_with_parentheses = (2 + 8) * (8 + 8)
print("(2 + 8) * (8 + 8)  =", result_with_parentheses)
````

**Now try it yourself.** Try to calculate the following expressions by hand. This will help you build confidence with operator precedence:

- `2 + 4 * 4 + 4`: what does Python evaluate first?
- `5 + 5 * 5 + 5`: same question.
- `(2 + 4) * (4 + 4)`: how does adding parentheses change the result?

Once you've tried this out, run the code cell below to determine if you were correct.

````python
# Listing 13.
# Try the expressions below — work out the expected answer first, then run the cell.
# Expression 1: what does Python evaluate first?
print("2 + 4 * 4 + 4  =", 2 + 4 * 4 + 4)

# Expression 2
print("5 + 5 * 5 + 5  =", 5 + 5 * 5 + 5)

# Expression 3: how do parentheses change the result?
print("(2 + 4) * (4 + 4) =", (2 + 4) * (4 + 4))
````

### 5.4 Variables

A **variable** is a named label that stores a value in memory. You create (or *assign*) a variable using the `=` (equals) operator. In Python, variable names should use **snake_case**: all lowercase letters with underscores separating words (e.g. `my_variable`, `age`). This is the convention specified in **PEP 8**, the official Python style guide.

> **Note:** Python uses `=` for *assignment* (storing a value), and `==` for *comparison* (checking whether two values are equal). These are easy to confuse, so remember that `=` is for storing, `==` is for testing equality.

Read the code below and try to understand it. When ready, execute the code to test your understanding.

````python
# Listing 14.
# Assigning a value to a variable using snake_case naming (PEP 8 convention)
my_variable = 5
print("my_variable =", my_variable)

# Assigning values to two separate variables
x = 2
y = 3

# Variables can be used in arithmetic expressions just like raw numbers
print("x + y =", x + y)         # Addition
print("x * y =", x * y)         # Multiplication
print("x ** y =", x ** y)       # x to the power of y: 2^3 = 8
````

## 6. Python Strings

So far we've worked with numbers. The other data type you'll use constantly is the **string**: a sequence of characters enclosed in either single quotes (`'...'`) or double quotes (`"..."`). Strings are used to represent and store text: labels, messages, file names, and much more. ### 6.1 Creating Strings

You can create a string using either single or double quotes, Python treats them identically. Using double quotes is convenient when your text contains an apostrophe (e.g. `"it's a string"`), and vice versa.


````python
# Listing 15.
# Strings can be created with single quotes or double quotes — both are equivalent
string_single = 'Hello, World!'
string_double = "Hello, World!"

print(string_single)
print(string_double)

# You can check the type — both are 'str'
print(type(string_single))
print(type(string_double))
````

### 6.2 Printing and String Formatting

The built-in `print()` function outputs text to the screen. It is one of the most useful tools for understanding what your code is doing: you can insert `print()` statements anywhere to inspect the value of a variable at some point during the execution of your code. This makes it invaluable for **debugging** (removing bugs!).

#### Basic Printing


````python
# Listing 16.
# print() outputs its argument to the screen.
# It can display strings, numbers, or any other Python object.
print('Welcome to Rob\'s notebook!')

# You can print multiple values by separating them with commas.
# print() will add a space between them automatically.
print('The answer is:', 42)
````

#### String Formatting with `.format()`

Simply printing a fixed string is useful, but in practice we often need to **insert variable values into a string**. Python's `.format()` method lets you do this. You place curly braces `{}` inside the string as **placeholders**, and `.format()` fills them in left to right.


````python
# Listing 17.
# .format() replaces the curly braces {} with the supplied values, in order.
# This is useful for building informative output messages in your code.
print("My name is: {}".format('Rob'))

# You can supply multiple values — each {} is filled in left to right
print("My name is: {} and I work at {}".format('Rob', 'bOB'))
````

You can also use **named placeholders** to make the format string more readable and to avoid errors when there are many values to insert:


````python
# Listing 18.
# Named placeholders make the format string easier to read and maintain.
# The keyword argument name inside {} matches the argument name in .format().
print("My name is: {name} and I work at {place}".format(
    name='Rob',
    place='Bob'
))
````

#### f-strings (Modern Alternative)

Python 3.6 introduced **f-strings** (formatted string literals) as a cleaner alternative to `.format()`. To use them you prefix the string with `f` and place variable names (or expressions) directly inside the curly braces:


````python
# Listing 19.
# f-strings: prefix the string with f and embed variables/expressions directly
name = 'Rob'
workplace = 'Bob'
year = 2026

# The variable is evaluated and inserted at runtime — cleaner than .format()
print(f"My name is: {name} and I work at {workplace}")
print(f"The year is {year} and 2 + 2 = {2 + 2}")  # Expressions work too!
````

So far we have worked with single values: one number or one string at a time. In real-world programming and data science, we almost always need to work with **collections of values**. Python provides four built-in data structures for this purpose, each with different properties:

| Structure | Ordered? | Changeable? | Allows duplicates? | Syntax |
|-----------|----------|-------------|-------------------|--------|
| **List** | ✅ Yes | ✅ Yes | ✅ Yes | `[1, 2, 3]` |
| **Dictionary** | ✅ Yes (Python 3.7+) | ✅ Yes | Keys: ❌ No | `{'key': 'value'}` |
| **Tuple** | ✅ Yes | ❌ No | ✅ Yes | `(1, 2, 3)` |
| **Set** | ❌ No | ✅ Yes | ❌ No | `{1, 2, 3}` |

### What is a Data Structure?
A **data structure** is a way of organising and storing data in a computer so it can be used efficiently. Each of Python's built-in data structures behaves differently:

- **Lists** are ordered, changeable, and allow duplicates, ideal for collections where you need flexibility.
- **Dictionaries** store data as key-value pairs, making them perfect for labelled or structured data.
- **Tuples** are ordered and unchangeable, useful for fixed data that shouldn't be modified.
- **Sets** are unordered, changeable, and do not allow duplicates, great for unique items or membership testing.

The differences in ordering, mutability (changeability), and how they handle duplicates, determine which structure is best for a given task. Choosing the right structure for the right job is a key programming skill. As you progress through the notebooks you will see all four used in different contexts.

### 7.1 Lists

A **list** is Python's most versatile data structure. It is an **ordered, changeable** collection that allows duplicate values and can hold mixed data types. Lists are defined using square brackets `[]`.

In machine learning, lists (and their NumPy equivalents, arrays) are used extensively to store feature values, class labels, and model predictions.


````python
# Listing 22.
# ── Creating Lists ────────────────────────────────────────────────────────────

# A list of integers — could represent, for example, ages of patients
ages = [25, 30, 22, 45, 30]
print("Ages:", ages)

# A list of strings — could represent class labels
labels = ['cat', 'dog', 'cat', 'bird', 'dog']
print("Labels:", labels)

# Lists can mix data types (though this is less common in data science)
mixed = [1, 'two', 3.0, True]
print("Mixed:", mixed)
````

#### Indexing and Slicing
To work with data structures like lists, we often need to access specific items or subsets of the data. For example, if we have the following list (repeated from above):

```PYTHON
# A list of integers — could represent, for example, ages of patients
ages = [25, 30, 22, 45, 30]
print("Ages:", ages)
```

How can we access the first item, the last, the third, and so on? There are two ways: **indexing** and **slicing**. Before we get into those, some terminology first:

- **Index**: A position in an ordered collection (like a list or string), used to access a specific element. In Python, indices start at `0` for the first element. In the list above, the item at index `0` would be `25`.
- **Indices**: The plural of "index", referring to multiple positions in a collection (e.g. `0`, `1`, `2`).

With that terminology out of the way:

* **Indexing** lets you retrieve a single item by its position using **zero-based indexing**: the first element is at index `0`, the second at index `1`, and so on. This is a very common source of confusion for beginners, so take care. You can also use **negative indices** to count from the end: `-1` gives the last element, `-2` the second to last, etc.
* **Slicing** lets you extract a range of items.

Read the code sample below and try to understand what it's doing. When you're confident that you know, run the code cell to check your understanding.


````python
# Listing 23.
# Create a list of fruit names to demonstrate indexing
fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry']

# Zero-based indexing: the FIRST element is at index 0
print("First element  (index 0):", fruits[0])   # apple
print("Second element (index 1):", fruits[1])   # banana
print("Third element  (index 2):", fruits[2])   # cherry

# Negative indexing counts from the END of the list
print("Last element   (index -1):", fruits[-1])  # elderberry
print("Second to last (index -2):", fruits[-2])  # date
````

You can clearly see here how we've used indexing to access items from the list we created.

#### Modifying Lists

Because lists are **mutable** (changeable), you can update, add, and remove elements after the list has been created.

Python list functions follow a consistent naming convention. The most important ones are summarised below:

| Method | Description |
|--------|-------------|
| `list.append(x)` | Adds `x` to the **end** of the list |
| `list.insert(i, x)` | Inserts `x` at position `i` |
| `list.remove(x)` | Removes the **first** occurrence of `x` |
| `list.pop(i)` | Removes and returns the element at position `i` (default: last) |
| `len(list)` | Returns the number of elements (this is a built-in function, not a method) |

Read the code in the cell below and try to understand it. When you feel you do, execute it to see how the list functions behave in practice.

````python
# Listing 24.
# Start with a simple list
shopping = ['milk', 'eggs', 'bread']
print("Original:", shopping)

# append() adds a new item to the END of the list
shopping.append('butter')
print("After append('butter'):", shopping)

# insert(index, value) adds an item at a specific position
# Here we insert 'coffee' at position 0 (the very beginning)
shopping.insert(0, 'coffee')
print("After insert(0, 'coffee'):", shopping)

# Change a value by assigning to a specific index
shopping[1] = 'oat milk'   # Replace 'milk' (now at index 1) with 'oat milk'
print("After updating index 1:", shopping)

# remove() deletes the first occurrence of a specified value
shopping.remove('eggs')
print("After remove('eggs'):", shopping)

# len() returns the number of items in the list
print("Number of items:", len(shopping))
````

#### Looping Through a List and Membership Testing

Two very common operations on lists are iterating (looping) through all elements, and checking whether a specific value exists in the list.

Read the code in the cell below and try to understand it. When you feel you do, execute it to see how we can loop through a list and check if specific values are within the list.


````python
# Listing 25.
# A for loop iterates through every item in the list in order.
# 'item' is just a loop variable — you can name it anything.
print("Items in the shopping list:")
for item in shopping:
    print(" -", item)

# The 'in' keyword tests for membership — returns True or False
if 'coffee' in shopping:
    print("\nCoffee is on the list!")
else:
    print("\nCoffee is NOT on the list.")

# You can also check for absence with 'not in'
if 'eggs' not in shopping:
    print("Eggs have been removed from the list.")
````

#### Nested Lists

A list can contain other lists as elements: this is called a **nested list**. You will encounter this pattern when working with matrices and 2D data structures later in the notebook series. Read the code in the cell below and try to understand it. Then execute it to see how nested lists behave in practice.


````python
# Listing 26.
# A nested list: the last element is itself a list
# This structure could represent a 2D data point [x, y] embedded in a dataset
data_point = [1, 2, 3, 4, [100, 150]]
print("Full list:", data_point)

# To access elements in a nested list, chain two index operations:
# data_point[4]  -> [100, 150]  (the inner list)
# data_point[4][0] -> 100       (first element of the inner list)
# data_point[4][1] -> 150       (second element of the inner list)
print("Inner list:", data_point[4])
print("First element of inner list:", data_point[4][0])
print("Second element of inner list:", data_point[4][1])
````

### 7.2 Dictionaries

A **dictionary** stores data as **key–value pairs**. Instead of accessing values by a numeric index (as in a list), you access them by their **key**, which can be any immutable type (usually a string or integer). Dictionaries are defined using curly brackets `{}`.

Dictionaries are extremely useful in data science for storing configuration settings, mapping class labels to numeric codes, and representing structured records (similar to a row in a database table).

#### Example: Working with a Patient Record
The code cell below creates a dictionary to represent a patient's medical record. Each **key** (e.g. `'name'`, `'age'`) describes the type of data, and each **value** (e.g. `'Alice'`, `32`) holds the actual information.

Run the cell to see how dictionaries allow you to:
- Store different types of data (strings, integers, etc.) in one structured object.
- Access values directly using their keys (e.g. `patient['name']`).

This is a practical way to organise and retrieve data efficiently, especially when working with real-world datasets.

````python
# Listing 27.
# Create a dictionary representing a single patient record.
# Keys are descriptive strings; values can be any type.
patient = {
    'name':      'Alice',         # string value
    'age':        32,             # integer value
    'diagnosis': 'hypertension',  # string value
    'bp_systolic': 145            # integer value
}

# Access a value by providing its key in square brackets
print("Patient name:", patient['name'])
print("Age:", patient['age'])
print("Diagnosis:", patient['diagnosis'])
````

#### Modifying and Extending a Dictionary
Dictionaries are **mutable**, meaning you can change their contents after creation. The code below demonstrates three common operations:

- **Updating a value**: Assign a new value to an existing key (e.g. updating a patient's blood pressure).
- **Adding a new entry**: Introduce a new key-value pair (e.g. adding a medication field).
- **Iterating through the dictionary**: Loop over all keys to access and print each value.

Run the cell to see how dictionaries can be dynamically updated and traversed, making them ideal for managing structured data like patient records.

````python
# Listing 28.
# Modifying a dictionary value: assign a new value to an existing key
patient['bp_systolic'] = 138
print("Updated blood pressure:", patient['bp_systolic'])

# Adding a new key-value pair — simply assign to a new key name
patient['medication'] = 'amlodipine'
print("Added medication:", patient['medication'])

# Looping through a dictionary: iterating over keys and accessing values
print("\nFull patient record:")
for key in patient:
    print(f"  {key}: {patient[key]}")
````

### 7.3 Tuples

A **tuple** is similar to a list, but it is **immutable**: once created, its contents cannot be changed. Tuples are defined using parentheses `()`.

Tuples are used when you want to store a fixed collection of values that should not be modified, for example, the coordinates of a fixed point, or the (rows, columns) shape of a dataset. Their immutability makes them slightly faster than lists and prevents accidental modification.

#### Example: Using Tuples for Fixed Data
The code below shows two practical uses of tuples:
- Storing **(x, y) coordinates** for a point in 2D space.
- Representing the **shape of a dataset** (rows, columns), such as the Iris dataset (150 samples, 4 features), a famous machine learning dataset that we might come across at some point.

Run the cell to see how tuples work. Notice that, like lists, you can access individual elements using indexing (e.g. `point[0]` for the x-coordinate). However, unlike lists, you cannot modify the tuple after creation.

````python
# Listing 29.
# Create a tuple representing the (x, y) coordinates of a point
# This is a common use of tuples in geometry and data science
point = (3.5, 7.2)
print("Point:", point)
print("x-coordinate:", point[0])
print("y-coordinate:", point[1])

# Another common use: storing the dimensions of a dataset
# (number of rows, number of columns)
dataset_shape = (150, 4)   # e.g. the Iris dataset: 150 samples, 4 features
print("\nDataset shape:", dataset_shape)
print("Number of samples:", dataset_shape[0])
print("Number of features:", dataset_shape[1])
````

The key property of a tuple is that it **cannot be modified**. Attempting to change a value raises a `TypeError`. Run the cell below to see this in action: this error is **expected and intentional**.


````python
# Listing 30.
# Attempting to change a value in a tuple raises a TypeError.
# This is intentional — it demonstrates that tuples are immutable.
# The error message tells us: 'tuple' object does not support item assignment.
try:
    point[0] = 99   # This will raise a TypeError
except TypeError as e:
    print(f"TypeError caught: {e}")
    print("Tuples cannot be modified after creation — this is by design.")
````

### 7.4 Sets

A **set** is an **unordered** collection of **unique** values, duplicates are automatically removed. Sets are defined using curly brackets `{}` (like dictionaries, but without key–value pairs).

Sets are useful when you only care about the *distinct* elements in a collection, for example, finding all unique class labels in a dataset, or checking for overlap between two groups.

#### Example: Working with Unique Values
The code below demonstrates how sets handle uniqueness:
- Creating a set from a list of class labels (duplicates like `'cat'` and `'dog'` appear only once).
- Converting a list of survey responses to a set to extract only the unique values.

Run the cell to see how sets automatically remove duplicates and provide the count of unique items. Note that the order of elements in a set is not guaranteed.

````python
# Listing 31.
# Create a set of unique class labels
# Even if duplicates are included, only unique values are stored
labels = {'cat', 'dog', 'cat', 'bird', 'dog', 'cat'}
print("Unique labels:", labels)
# Note: the order of elements in a set is not guaranteed

# Sets automatically deduplicate — useful for finding unique values in data
survey_responses = [3, 5, 3, 2, 5, 1, 4, 2, 3, 5]
unique_responses = set(survey_responses)  # Convert list to set to get unique values
print("Unique survey responses:", unique_responses)
print("Number of unique responses:", len(unique_responses))
````

#### Mathematical Operations with Sets
Sets in Python support standard mathematical operations, making them powerful tools for data analysis tasks like comparing groups, finding overlaps, or identifying unique elements.

The code below demonstrates three fundamental set operations:
- **Union (`∪`)**: Combines all unique elements from both sets.
- **Intersection (`&`)**: Returns only elements present in both sets.
- **Difference (`-`)**: Returns elements in the first set that are not in the second.

Run the cell to see how these operations work in practice.

````python
# Listing 32.
# Sets support mathematical set operations — very useful in data analysis

set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

# Union: all elements in either set
print("Union (A ∪ B):", set_a | set_b)

# Intersection: elements in BOTH sets
print("Intersection (A ∩ B):", set_a & set_b)

# Difference: elements in A but NOT in B
print("Difference (A - B):", set_a - set_b)
````

### 8.1 A Vertical Line - x = k

The simplest possible line is a **vertical line** defined by a single value of `x`. The line `x = k` passes through every point where `x` equals `k`, regardless of what `y` is.

**Why does this matter in machine learning?** Suppose you measure the height of plants and want to classify them as healthy or unhealthy. If you observe that all plants taller than 10 cm are healthy and all plants shorter than 10 cm are unhealthy, then the line **`x = 10`** perfectly **separates** the two groups. This is the most basic form of a **decision boundary**: a threshold that assigns a data point to one class or another based on where it falls around some line.

The plot below visualises exactly this scenario. Take a moment to read through each comment before running the cell:


````python
# Listing 35.
# Plant heights (in cm) for two groups
healthy   = [12, 15, 18, 22, 11, 14, 20]   # Heights > 10 cm
unhealthy = [3,  6,  8,  5,  9,  4,  7]    # Heights < 10 cm

# plt.subplots() creates a Figure and one or more Axes (subplots).
# - The Figure is the overall canvas (the window/image).
# - The Axes is the individual plot area within the canvas.
# - figsize=(8, 4) sets the width and height of the figure in inches.
# We unpack the return value into two variables: fig (the figure) and ax (the axes).
fig, ax = plt.subplots(figsize=(8, 5))
fig.canvas.header_visible = False
fig.canvas.resizable = True
fig.canvas.toolbar_visible = False

# ax.scatter() draws a scatter plot — individual points at (x, y) coordinates.
# We plot all healthy plants at y=1 and unhealthy plants also at y=1
# (we only care about the x-axis here — y is just used to spread the points out).
# s=100 sets the marker size; zorder=3 ensures points are drawn on top of the gridlines.
ax.scatter(healthy,   [1] * len(healthy),   color='green', label='Healthy',   s=100, zorder=3)
ax.scatter(unhealthy, [1] * len(unhealthy), color='red',   label='Unhealthy', s=100, zorder=3)

# ax.axvline() draws a vertical line at a specified x value.
# Unlike ax.plot() (which draws between two points), axvline() spans the full height.
# x=10 — the x position of the line (our decision threshold)
# linestyle='--' — dashed line style
ax.axvline(x=10, color='blue', linewidth=2, linestyle='--', label='Decision boundary: x = 10')

# Label the x-axis and remove the y-axis ticks (they have no meaning here)
ax.set_xlabel('Height (cm)')
ax.set_yticks([])   # ax.set_yticks([]) removes tick marks from the y-axis

ax.set_title('Figure 1: A vertical decision boundary separating healthy and unhealthy plants')
ax.legend()
plt.tight_layout()  # Automatically adjusts spacing to prevent labels from being cut off
plt.show()
````

Every green point (healthy) falls to the right of the boundary line, and every red point (unhealthy) falls to the left. The line at x = 10 **perfectly separates** the two classes in this simple example.

**Try it yourself:** Change the value `x=10` inside `ax.axvline(x=10, ...)` to `x=8` or `x=12` and re-run the cell. How does shifting the boundary change which plants are misclassified? Can you find a value of `x` that misclassifies the fewest points?

### 8.2 The Equation of a Straight Line - y = mx + c

Most decision boundaries in machine learning are **not** vertical lines. Some could be diagonal lines that cut across two-dimensional space. The general equation of a straight line capable of drawing such a diagonal line is:

$$y = mx + c$$

- **m** is the **slope** (or gradient): it controls how steeply the line rises or falls as x increases. A positive slope means the line rises left to right; a negative slope means it falls.
- **c** is the **y-intercept**: the value of y when x = 0, i.e. where the line crosses the vertical axis.

This equation is the foundation for **linear regression**, **logistic regression**, and **support vector machines**. The table below summarises how different values of `m` and `c` affect the line:

| Parameter | Value | Effect on the line |
|-----------|-------|-------------------|
| **m** | m > 0 | Line rises as x increases (positive slope) |
| **m** | m = 0 | Flat, horizontal line |
| **m** | m < 0 | Line falls as x increases (negative slope) |
| **m** | Large \|m\| | Steep line |
| **m** | Small \|m\| | Shallow line |
| **c** | c > 0 | Line shifted upwards |
| **c** | c = 0 | Line passes through the origin (0, 0) |
| **c** | c < 0 | Line shifted downwards |

### 8.3 Using a Line to Separate Data - Decision Boundaries

Now let us combine the ideas from Sections 8.1 and 8.2. Suppose we have two groups of data points in two-dimensional space, for example, two species of flower, each described by two measurements: petal length (x) and petal width (y). A line **```y = mx + c```** can act as a **decision boundary**: points above the line are predicted to belong to one class; points below the line to the other.

This is the geometric foundation for many classifiers you will encounter later in this notebook series, including:

- **Logistic Regression** (Notebook 5): finds the optimal linear boundary for binary classification.
- **Support Vector Machines** (Notebook 6): finds the boundary with the maximum margin between classes.

### 8.4 The Notion of Error - Residuals and MAE

Real data is messy, points rarely sit exactly on a line. When we fit a line to data, each point will generally be some distance away from the line. The **vertical distance** from a data point to the line is called the **residual** or **error**:

$$	error_i = y_i - \hat{y}_i$$

where:
- $y_i$ is the **actual** (observed) value of the i-th data point.
- $\hat{y}_i = m x_i + c$ is the **predicted** value from the line at the same x (we call $\hat{y}$ y-hat).

A **positive residual** means the point lies *above* the line; a **negative residual** means it lies *below*.

The goal in linear regression (Notebook 6) is to find the m and c that **minimise the total error** across all data points. To do this, we need a single number that summarises how well the line fits. One such measure is the **Mean Absolute Error (MAE)**:

$$ \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i| $$

The **MAE** is the average of all the absolute residuals. In simple terms, for each data point, we measure the vertical distance (the error) from the point to the line, ignore whether it's above or below (by taking the absolute value), and then calculate the average of all these distances. A smaller MAE means the line fits the data better.

We take the **absolute value** of each residual before averaging because positive and negative residuals would otherwise cancel each other out: a line that is above half the points and below the other half might appear to have zero error, even though it is a poor fit.

Your goal is to find values that produce the lowest possible MAE. This is exactly what a regression algorithm does, except it uses calculus rather than trial and error.

**Key observations:**

- Some residuals are positive (data point is above the line) and some are negative (below). If we summed them directly, they would partially cancel out, which would be misleading. Taking the absolute value before averaging prevents this.
- The MAE is a single number that summarises how well the line fits: the lower the MAE, the better.
- In Notebook 6, Python will find the optimal m and c automatically using an algorithm called **Ordinary Least Squares (OLS)**. That algorithm minimises the **sum of squared errors** (SSE) rather than MAE, but the intuition is identical: it is searching for the line that is, on average, closest to all the data points.
