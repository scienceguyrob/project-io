## 4. Exploratory Data Analysis: Know Your Data First
🔙 [Back to Table of Contents](#Table-of-Contents)

Before building any ML model, we need to understand the data we are working with. This process is called **Exploratory Data Analysis (EDA)**, and it is one of the most important steps in any machine learning project, as well as one of the most often skipped. Skipping it is one of the most common reasons models fail in practice.

> 📓 **Notebook 4 refresher:** We worked with pandas in Notebook 4. This section revisits the key ideas because EDA is where pandas earns its keep, and because returning to a tool in a new context is one of the best ways to make it stick.

Visualisation and summary statistics are how we *understand* a dataset before touching a model. The two most important first steps are:

- Checking the **data types and shape**: what kind of data do we have, and how much of it?
- Computing **summary statistics**: what are the typical values, ranges, and distributions?

Together, these reveal whether data is missing, whether features are on comparable scales, and whether outliers are present, all of which directly affect model performance. Let's look at each in turn.

### 4.1 Data Types and Shape

Every dataset has a **shape**: the number of rows (observations) and columns (features). In pandas, a dataset is stored as a `DataFrame`. Think of it as a spreadsheet held in memory.

```python
import pandas as pd

df = pd.read_csv('your_dataset.csv')

# How many rows and columns?
print(df.shape)          # e.g. (150, 5): 150 rows, 5 columns

# What are the column names, data types, and are there missing values?
df.info()
```

The `.info()` method is particularly useful because it tells you:

- The **name** of each column
- The **data type**, e.g. `int64` (whole numbers), `float64` (decimal numbers), `object` (text or mixed)
- The **number of non-null entries**: a quick way to spot missing data

> ⚠️ **Why data types matter:** Machine learning models work with numbers. If a column contains text categories (e.g. `"male"` / `"female"`), it needs to be converted before a model can use it. Spotting this early saves debugging later.

### 4.2 Summary Statistics

Once we know the shape and types, we want to understand the *values* themselves. The `.describe()` method produces a table of summary statistics for every numeric column in the DataFrame.

```python
df.describe()
```

For each numeric column, this returns:

| Statistic | What it tells you |
|-----------|-------------------|
| `count` | How many non-missing values exist |
| `mean` | The average value |
| `std` | Standard deviation: how spread out the values are |
| `min` | The smallest value |
| `25%` | The value below which 25% of observations fall (lower quartile) |
| `50%` | The median, the middle value |
| `75%` | The value below which 75% of observations fall (upper quartile) |
| `max` | The largest value |

> 💡 **Intuition:** Think of `.describe()` as a first health check. It won't tell you everything, but it will immediately flag if something looks wrong: a negative age, a salary of zero, a test score of 999. These are the kinds of issues that corrupt a model silently if left unchecked.

The cells below use pandas' built-in `.describe()` and `.info()` methods for this initial audit of our Breast Cancer Wisconsin dataset from scikit-learn.

````python
# Listing 3.
# .describe() computes summary statistics (count, mean, std, min, quartiles, max)
# for every numeric column. Round to 2 decimal places for readability.
df.describe().round(2)
````

---

Next we use the `.info()` method to get a structural summary of the dataset.

````python
# Listing 4.
# .info() gives a concise structural summary: column names, non-null counts, and dtypes.
# A non-null count less than the total row count indicates missing values.
df.info()
````

---

**What to look for:**

- **Large gap between `mean` and `50%` (median):** suggests the data is *skewed*, pulled in one direction by extreme values. For example, income data is often skewed because a small number of very high earners pull the average up.
- **Large `std` relative to the `mean`:** suggests high variability, or possibly outliers.
- **Very different ranges across columns:** for example, one feature ranging from 0–1 and another from 0–100,000. Many ML algorithms are sensitive to this, as features on larger scales can dominate the model unfairly. This is called the **feature scaling problem**, and we will address it later in this notebook.
- **`count` less than the total number of rows:** missing values are present in that column.

## 5. Matplotlib: The Foundation

🔙 [Back to Table of Contents](#Table-of-Contents)

Every Python visualisation library, including seaborn, pandas `.plot()`, and scikit-learn's plot utilities, is built on top of **matplotlib**. It is the engine underneath, even when you are not using it directly. Understanding how matplotlib thinks about plots saves significant debugging time when customising or combining charts.

### 5.1 How Matplotlib Structures a Plot

Matplotlib organises every plot around two core objects:

- A **Figure** (`fig`): the overall canvas. Think of it as the blank page.
- One or more **Axes** (`ax`): the individual plot panels that sit inside the figure. Each Axes has its own x-axis, y-axis, title, and data.

This distinction matters because when something looks wrong, whether that's a title appearing on the wrong panel, labels missing, or plots overlapping, the fix almost always comes down to knowing which Axes you are currently modifying.

### 5.2 Two Interfaces: Stateful vs Object-Oriented

Matplotlib offers two ways to build plots:

**Stateful interface**: functions like `plt.plot(...)` and `plt.title(...)` implicitly operate on whichever Axes was most recently created. This is convenient for a single quick plot but becomes unpredictable as soon as you have more than one panel.

**Object-oriented interface**: you explicitly create a Figure and its Axes upfront, then call methods directly on them:

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('My plot')
```

> 💡 **Rule of thumb:** Use the stateful interface only for a single throwaway plot. Use the object-oriented interface for everything else. The habits you build here will save you hours of confusion later.

### 5.3 Demonstration: Three Common Plot Types

The cells below create 3 single Figures. **Notice the pattern:** every cell follows the same structure: create the plot, label the axes, set the title. This consistency makes matplotlib code easy to read and easy to extend. Let's start with the line plot.



````python
# Listing 5.

# --- Plot 1: Line plot ---
# np.linspace(start, stop, n) generates n evenly spaced values between start and stop.
# Here we create 200 points from 0 to 2π to produce a smooth sine curve.
x = np.linspace(0, 2 * np.pi, 200)

# plt.subplots() with no arguments creates a single Figure and a single Axes.
# figsize=(width, height) sets the size of the figure in inches.
fig, ax = plt.subplots(figsize=(5, 4))

# ax.plot() draws a line connecting each (x, y) point in order.
# color= sets the line colour; linewidth= sets its thickness in points.
ax.plot(x, np.sin(x), color='steelblue', linewidth=2)

ax.set_title('Figure 1. Line plot')   # Title displayed above the plot
ax.set_xlabel('x')          # Label for the horizontal axis
ax.set_ylabel('sin(x)')     # Label for the vertical axis

plt.tight_layout()  # Adjusts spacing so labels and titles are not clipped
plt.show()  
````

Next we have a line plot with multiple lines. Here calling ```ax.plot()``` twice on the same Axes overlays both lines on the same panel.

````python
# Listing 6.
# --- Plot 2: Multiple lines with a legend ---
x = np.linspace(0, 2 * np.pi, 200)

fig, ax = plt.subplots(figsize=(5, 4))

# Calling ax.plot() twice on the same Axes overlays both lines on the same panel.
# label= assigns each line a name that will appear in the legend.
ax.plot(x, np.sin(x), color='steelblue', linewidth=2, label='sin(x)')

# linestyle='--' renders the line as a dashes rather than solid.
# Other options include ':' (dotted), '-.' (dash-dot), and '-' (solid, the default).
ax.plot(x, np.cos(x), color='tomato', linewidth=2, linestyle='--', label='cos(x)')

ax.set_title('Figure 2. Multiple lines + legend')
ax.set_xlabel('x')

# ax.legend() reads the label= arguments from every .plot() call on this Axes
# and draws a legend box automatically. Without label= arguments it produces nothing.
ax.legend()

plt.tight_layout()
plt.show()
````

Finally, here's a scatter plot, which places one marker per point rather than connecting them with a line.

````python
# Listing 7.
# --- Plot 3: Scatter plot ---
# np.random.default_rng(seed) creates a random number generator.
# Providing a fixed seed (here 0) means the same random values are produced
# every time the cell runs — essential for reproducibility.
rng = np.random.default_rng(0)

# rng.uniform(low, high, n) draws n values sampled uniformly between low and high.
xr = rng.uniform(0, 10, 80)

# rng.normal(mean, std, n) draws n values from a Gaussian distribution.
# Adding this noise to a linear function simulates realistic, imperfect data.
yr = 2 * xr + rng.normal(0, 3, 80)

fig, ax = plt.subplots(figsize=(5, 4))

# ax.scatter() places one marker per (x, y) pair — no lines are drawn.
# s=       marker size in points squared
# alpha=   transparency, from 0 (invisible) to 1 (fully opaque); helps reveal overlapping points
# edgecolors= colour of the marker border; 'k' is shorthand for black
# linewidth=  thickness of the marker border
ax.scatter(xr, yr, color='seagreen', s=40, alpha=0.7, edgecolors='k', linewidth=0.3)

ax.set_title('Figure 3. Scatter plot')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.tight_layout()
plt.show()
````


### 5.4 Combining Multiple Plots in One Figure

Each of the three plots above was built in isolation: one Figure, one Axes, one plot. In practice, placing related plots side by side on the same Figure makes comparisons much easier and is standard in EDA workflows.

To do this we pass a layout to `plt.subplots()` using two arguments: the number of rows and the number of columns. For example:

```python
fig, axes = plt.subplots(1, 3)  # 1 row, 3 columns: three panels side by side
fig, axes = plt.subplots(2, 2)  # 2 rows, 2 columns: a grid of four panels
```

When you create more than one Axes this way, `plt.subplots()` returns them as a NumPy array rather than a single object. You access each panel by indexing into that array: `axes[0]` for the first panel, `axes[1]` for the second, and so on.

> 💡 Instead of writing `ax.plot(...)` on a single Axes, you now write `axes[0].plot(...)`, `axes[1].plot(...)`, and so on, with each index targeting a specific panel. Everything else works exactly as before.

The cell below brings all three plot types together into a single Figure with three panels side by side.


````python
# Listing 8.
# Demonstration of the matplotlib object-oriented interface.
# plt.subplots(1, 3, figsize=(w, h)) creates a Figure (fig) and a 1×3 array of Axes (axes).
#   - fig:  the overall canvas
#   - axes: a NumPy array of individual subplot panels
#   - figsize: width and height in inches

x = np.linspace(0, 2 * np.pi, 200)  # np.linspace(start, stop, n) creates n evenly spaced values

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

# Panel 1 — line plot
axes[0].plot(x, np.sin(x), color='steelblue', linewidth=2)
axes[0].set_title('Line plot')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')

# Panel 2 — multiple lines with a legend
axes[1].plot(x, np.sin(x), color='steelblue', linewidth=2, label='sin(x)')
axes[1].plot(x, np.cos(x), color='tomato', linewidth=2,
             linestyle='--', label='cos(x)')  # linestyle='--' draws a dashed line
axes[1].set_title('Multiple lines + legend')
axes[1].set_xlabel('x')
axes[1].legend()  # ax.legend() draws a legend from the label= arguments above

# Panel 3 — scatter plot
rng = np.random.default_rng(0)            # Seeded random number generator for reproducibility
xr  = rng.uniform(0, 10, 80)             # 80 random x values uniform on [0, 10]
yr  = 2 * xr + rng.normal(0, 3, 80)      # Linear relationship with Gaussian noise
axes[2].scatter(xr, yr, color='seagreen', s=40, alpha=0.7,
                edgecolors='k', linewidth=0.3)  # s = marker size; alpha = transparency
axes[2].set_title('Scatter plot')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')

# plt.suptitle() sets a title above all subplots; y moves it vertically
plt.suptitle('Figure 4: matplotlib basics — line, multi-line, and scatter plots',
             fontsize=13, y=1.02)
plt.tight_layout()  # Automatically adjusts spacing to prevent subplots overlapping
plt.show()
````

The cell below illustrates a bar chart with value annotations, a common requirement in ML reporting (e.g. annotating accuracy bars with numeric values). Note the use of `ax.text()` to position labels above each bar.

### 5.5 Bar Charts

A **bar chart** represents categorical data as rectangular bars whose height corresponds to a numeric value. They are one of the most widely used chart types in ML reporting, for example comparing model accuracy across algorithms, or showing how often each class appears in a dataset.

Bar charts are appropriate when your x-axis contains **discrete categories** (e.g. algorithm names, class labels) rather than continuous numeric values. When your x-axis is continuous and you want to show the distribution of values, you would use a **histogram** instead. We will cover those in the next section.

One common requirement when presenting bar charts is annotating each bar with its exact value so the reader does not have to estimate from the axis. Matplotlib does not do this automatically, but `ax.text()` gives you precise control over where a label is placed and how it is formatted.

`ax.text(x, y, string)` places a text label at the coordinate `(x, y)` on the Axes. The key arguments are:

| Argument | What it controls |
|----------|-----------------|
| `x, y` | The coordinates where the text is anchored |
| `ha=` | Horizontal alignment relative to the anchor: `'center'`, `'left'`, or `'right'` |
| `va=` | Vertical alignment relative to the anchor: `'bottom'`, `'center'`, or `'top'` |
| `fontsize=` | Size of the text in points |

By placing the label at `(xi, yi + 0.1)`, the bar's x position and just above its top, and setting `ha='center'`, each value sits neatly centred above its bar. Here is an example of a bar chart.

````python
# Listing 9.
# Bar chart with value annotations — demonstrates ax.text() for labelling bars

fig, ax = plt.subplots(figsize=(8, 5))  # Single panel: returns one Figure and one Axes

x_vals = [1, 2, 3, 4, 5]
y_vals = [3, 7, 2, 9, 5]

ax.bar(x_vals, y_vals, color='steelblue', edgecolor='white', linewidth=0.6)

# Annotate each bar with its numeric value
# zip(x_vals, y_vals) pairs each x position with its corresponding bar height
for xi, yi in zip(x_vals, y_vals):
    ax.text(xi, yi + 0.1, str(yi),  # Place text at (xi, yi+0.1) — just above the bar top
            ha='center', va='bottom', fontsize=11)

ax.set_title('Figure 5: Bar chart with annotations', fontsize=13)
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_xticks(x_vals)
ax.set_xticklabels(['A', 'B', 'C', 'D', 'E'])  # Replace numeric ticks with category labels
ax.set_ylim(0, 11)  # Set y-axis range manually to give headroom above the tallest bar

plt.tight_layout()
plt.show()
````

### 5.6 Histograms

A **histogram** looks similar to a bar chart but serves a completely different purpose. Where a bar chart compares values across discrete categories, a histogram shows the **distribution of a single continuous numeric variable**, that is, how frequently values fall within successive ranges.

To build a histogram, the range of the data is divided into equally sized intervals called **bins**. Each bar's height shows how many observations fall within that bin. The result gives you an immediate visual sense of the shape of your data.

This matters enormously in ML. Before fitting a model you want to know:

- Is the data **roughly symmetric**, or is it skewed to one side?
- Are there **outliers**: isolated bars far from the main body of data?
- Is the data **unimodal** (one peak) or **multimodal** (multiple peaks)? Multiple peaks can suggest the data contains distinct subgroups.

The number of bins you choose affects what you see. Too few bins and the distribution looks like a flat, uninformative block. Too many bins and it becomes jagged noise. A good starting point is around 20–30 bins, then adjust based on what you see. If you want a more principled starting point, the **Freedman-Diaconis rule** calculates an optimal bin width based on the spread of the middle 50% of the data and the sample size, making it robust to outliers. NumPy implements it directly: pass `bins='fd'` to `ax.hist()` and the bin width is chosen automatically.

> 💡 **Histogram vs bar chart, the key distinction:** the bars in a bar chart have gaps between them to signal that the x-axis is categorical. The bars in a histogram touch each other to signal that the x-axis is continuous and the bins are contiguous intervals.


````python
# Listing 10.

# --- Histogram ---
# We simulate two groups of exam scores from different distributions
# to demonstrate what a bimodal (two-peaked) histogram looks like.
rng = np.random.default_rng(42)

group_a = rng.normal(loc=55, scale=8, size=200)   # Mean 55, std 8 — lower-scoring group
group_b = rng.normal(loc=75, scale=6, size=200)   # Mean 75, std 6 — higher-scoring group

# Combine both groups into one array — as if we observed 400 students without knowing their group
scores = np.concatenate([group_a, group_b])

fig, ax = plt.subplots(figsize=(8, 5))

# ax.hist() computes the bins and bar heights automatically from the raw data.
# bins=     number of equal-width bins to divide the data into
# color=    fill colour of the bars
# edgecolor= colour of the bar borders — setting this to white separates the bars visually
# alpha=    transparency; useful when overlaying multiple histograms
ax.hist(scores, bins=30, color='steelblue', edgecolor='white', alpha=0.85)

# A vertical line at the mean gives an immediate visual anchor
ax.axvline(scores.mean(), color='tomato', linewidth=1.5, linestyle='--', label=f'Mean: {scores.mean():.1f}')

ax.set_title('Figure 6: Histogram — combined exam scores (bimodal distribution)', fontsize=13)
ax.set_xlabel('Score')
ax.set_ylabel('Count')
ax.legend()

plt.tight_layout()
plt.show()
````

> **What to notice in this plot:** even though we did not tell the histogram that the data came from two groups, the two peaks are clearly visible. This is exactly the kind of signal EDA is designed to surface; a single aggregate statistic like the mean (shown by the dashed line) would completely obscure this structure.

**Key matplotlib patterns to remember:**

- `fig, ax = plt.subplots(figsize=(w, h))`: creates a single panel.
- `fig, axes = plt.subplots(rows, cols, figsize=...)`: creates a grid of panels; `axes` is a 2-D NumPy array indexed as `axes[row, col]`, or 1-D when `rows=1`.
- Always call `ax.set_*` methods (not `plt.*`) after the first `plt.subplots()` call. This makes it explicit which panel you are modifying.
- `plt.tight_layout()` before `plt.show()` automatically adjusts spacing to prevent label overlap.

## 6. Visualising Distributions

🔙 [Back to Table of Contents](#Table-of-Contents)

Understanding the distribution of each feature is the natural starting point for data exploration. For classification tasks, the most important question is: *does this feature's distribution look different for each class?* If the two class distributions overlap heavily, the feature carries little discriminative information. If they are well separated, the feature is a strong candidate for inclusion in a model.

We'll look at three different visualisations that can help us explore and understand features individually, and then combine everything into a single informative plot.

### 6.1 Feature Histograms

We introduced histograms in Section 5. Here we apply the same idea to a real feature from our dataset, `mean radius`, rather than simulated data. The first step is always a plain histogram of the overall distribution, ignoring class labels entirely. This tells us the range of the feature, whether it is symmetric or skewed, and whether anything unusual is present. Let's start with a histogram of `mean radius`:


````python
# Listing 11.
# The feature we want to inspect — change this string to explore any other column
feature = 'mean radius'

fig, ax = plt.subplots(figsize=(6, 4))

# df[feature] selects the entire column as a Series, passing all values to the histogram
# bins=30     divides the value range into 30 equal-width intervals
# edgecolor=  draws a white border between bars, making individual bins easier to distinguish
# alpha=      slight transparency so the colour is not too harsh at full opacity
ax.hist(df[feature], bins=30, color='steelblue', edgecolor='white', linewidth=0.5, alpha=0.85)

ax.set_xlabel(feature)   # Label the x-axis with the feature name
ax.set_ylabel('Count')   # Each bar's height = number of observations in that bin
ax.set_title(f'Figure 7: Histogram — {feature}')

plt.tight_layout()
plt.show()
````

### 6.2 Overlaying Histograms

Once we understand the overall distribution, the next question is whether it looks different for each class. We can answer this by plotting the histogram twice on the same Axes, once for each class, using transparency (`alpha`) so the overlapping bars remain visible.

````python
# Listing 12.
# The feature we want to inspect — change this string to explore any other column
feature = 'mean radius'

fig, ax = plt.subplots(figsize=(6, 4))

# We loop over a list of (class label, colour) pairs so both histograms are produced
# with the same logic rather than writing ax.hist() out twice.
# Each iteration filters the DataFrame to one class and plots its histogram on the same Axes.
for cls, col in [('Benign', 'steelblue'), ('Malignant', 'tomato')]:

    # df.loc[condition, column] selects rows where the condition is True, then returns
    # only the specified column — here, the feature values for one class at a time.
    ax.hist(df.loc[df['diagnosis'] == cls, feature],
            bins=25,
            color=col,
            alpha=0.55,        # Transparency below 1 so overlapping bars from both classes remain visible
            edgecolor='white',
            linewidth=0.4,
            label=cls)         # label= is used by ax.legend() to identify each class

ax.set_xlabel(feature)
ax.set_ylabel('Count')
ax.set_title(f'Figure 8: Histogram by class — {feature}')
ax.legend()  # Draws a legend box using the label= arguments assigned in the loop above

plt.tight_layout()
plt.show()
````


> 📌 **What to look for:** if the two coloured distributions sit on top of each other with heavy overlap, the feature does not separate the classes well. If they sit apart, as you should see here for `mean radius`, the feature is likely to be useful to a classifier.

### 6.3 Kernel Density Estimate (KDE) Plots

Overlapping histograms can become difficult to read, particularly when bin widths cause one bar to obscure another. A **Kernel Density Estimate (KDE)** solves this by replacing the bars with a smooth continuous curve.

The idea behind a KDE is straightforward: imagine placing a small, symmetric hill (a Gaussian curve) centred on each data point. A KDE sums all of those hills together into a single smooth curve that estimates the underlying probability distribution of the data. The result shows the same information as a histogram, namely where values are concentrated and where they are sparse, but without the sensitivity to bin width and bin alignment that can make histograms misleading.

The y-axis of a KDE plot shows **density** rather than count. Density values are scaled so that the total area under the curve equals 1, which means the y-axis values are not directly interpretable as counts. What matters is the **shape** of the curve and how the two classes compare.

Seaborn's `sns.kdeplot()` fits and plots a KDE in a single call.


> 💡 **Histogram or KDE?** Use a histogram when you want to communicate counts to a non-technical audience, since the y-axis is immediately interpretable. Use a KDE when you want to compare the shapes of two or more distributions, because the smooth curves are easier to read when overlaid.

````python
# Listing 13.
# The feature we want to inspect — change this string to explore any other column
feature = 'mean radius'

fig, ax = plt.subplots(figsize=(6, 4))

# As with the class histogram, we loop over (class label, colour) pairs to avoid
# repeating the sns.kdeplot() call twice with only the data and colour changed.
for cls, col in [('Benign', 'steelblue'), ('Malignant', 'tomato')]:

    # df[df['diagnosis'] == cls] filters the DataFrame to rows belonging to this class only.
    # sns.kdeplot() fits a kernel density estimate to the values in the specified column
    # and draws the resulting smooth curve on the Axes passed to ax=.
    # x=        the column whose values are used to fit the KDE
    # ax=       the Axes to draw on — important when working with subplots
    # fill=     shades the area under the curve, making it easier to see where density is concentrated
    # alpha=    transparency of the filled area; kept low so both classes are visible where they overlap
    # linewidth= thickness of the outline curve drawn on top of the filled area
    # label=    used by ax.legend() to identify each class
    sns.kdeplot(data=df[df['diagnosis'] == cls],
                x=feature, ax=ax, color=col, fill=True,
                alpha=0.35, linewidth=2, label=cls)

ax.set_xlabel(feature)
ax.set_ylabel('Density')  # KDE y-axis is probability density, not raw count
ax.set_title(f'Figure 9: KDE by class — {feature}')
ax.legend()

plt.tight_layout()
plt.show()
````


### 6.4 Combining All Three Views

Now that we have looked at each plot individually, the cell below combines all three into a single Figure. This is a pattern worth keeping as a reusable template: changing the `feature` variable at the top is all that is needed to inspect any other feature in the dataset.

````python
# Listing 14.
# Change this variable to explore any other feature in the dataset
feature = 'mean radius'

# 1 row, 3 columns of subplots — axes is a NumPy array of three Axes objects
# figsize is wider than usual to accommodate three panels side by side
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# --- Panel 1: plain histogram — overall distribution, no class split ---
# We assign axes[0] to a local variable ax for readability — it is the same object,
# just easier to write ax.hist() than axes[0].hist() repeatedly within a panel block
ax = axes[0]
ax.hist(df[feature], bins=30, color='steelblue',
        edgecolor='white', linewidth=0.5, alpha=0.85)
ax.set_xlabel(feature)
ax.set_ylabel('Count')
ax.set_title('Histogram')

# --- Panel 2: histogram split by class ---
# Reassigning ax to axes[1] redirects all subsequent ax. calls to the second panel
ax = axes[1]
for cls, col in [('Benign', 'steelblue'), ('Malignant', 'tomato')]:
    # df.loc[condition, column] filters to rows for this class and selects the feature column
    ax.hist(df.loc[df['diagnosis'] == cls, feature],
            bins=25, color=col,
            alpha=0.55,        # Transparency so overlapping bars from both classes remain visible
            edgecolor='white', linewidth=0.4, label=cls)
ax.set_xlabel(feature)
ax.set_ylabel('Count')
ax.set_title('Histogram by class')
ax.legend()

# --- Panel 3: KDE by class — smoothed density curves ---
ax = axes[2]
for cls, col in [('Benign', 'steelblue'), ('Malignant', 'tomato')]:
    # sns.kdeplot() fits a kernel density estimate and draws the smoothed curve
    # fill=True shades the area under the curve
    # alpha= kept low so both class curves remain visible where they overlap
    sns.kdeplot(data=df[df['diagnosis'] == cls],
                x=feature, ax=ax, color=col, fill=True,
                alpha=0.35, linewidth=2, label=cls)
ax.set_xlabel(feature)
ax.set_ylabel('Density')      # KDE y-axis is probability density, not raw count
ax.set_title('KDE by class\n(smoothed density estimate)')
ax.legend()

# plt.suptitle() places a single title above all three panels
# y=1.02 nudges it upward so it does not overlap with the individual panel titles
plt.suptitle(f'Figure 10: Distribution of "{feature}" — three views', y=1.02)
plt.tight_layout()
plt.show()

# Numerical confirmation of what the plots show — if the means differ substantially
# between classes, the feature is likely to be useful to a classifier
print('Class means:')
print(df.groupby('diagnosis')[feature].mean().round(3).to_string())
````

Having established how to read a KDE plot for a single feature, we can now scale the approach across multiple features at once. Rather than producing six separate plots, we arrange them in a grid, one panel per feature, so all six distributions are visible simultaneously and can be compared at a glance.

This is a practical feature selection technique: by visually scanning the grid, features where the two class curves sit apart with little overlap are immediately identifiable as strong candidates for a classifier. Features where the curves sit almost on top of each other are telling us that knowing that value does little to help distinguish one class from the other.

The cell below applies this approach to six features from our dataset. Pay attention to how much the two curves overlap in each panel.

````python
# Listing 15.
# A KDE survey plots multiple features at once so we can visually scan for the ones
# that best separate the two classes — a quick and effective feature selection technique.
features_to_plot = [
    'mean radius', 'mean texture', 'mean perimeter',
    'mean area', 'mean smoothness', 'mean concavity'
]

# plt.subplots(2, 3) creates a 2-row, 3-column grid — one panel per feature.
# axes is returned as a 2D NumPy array with shape (2, 3).
fig, axes = plt.subplots(2, 3, figsize=(14, 8))

# axes.ravel() flattens the 2D array of Axes into a 1D list so we can iterate
# over all six panels with a single loop rather than indexing rows and columns manually.
# zip(axes.ravel(), features_to_plot) pairs each Axes panel with its corresponding feature name.
for ax, feat in zip(axes.ravel(), features_to_plot):

    # For each feature, plot one KDE per class on the same Axes — identical pattern to Section 6.2
    for cls, col in [('Benign', 'steelblue'), ('Malignant', 'tomato')]:
        sns.kdeplot(data=df[df['diagnosis'] == cls],
                    x=feat, ax=ax, color=col, fill=True,
                    alpha=0.35, linewidth=2, label=cls)

    ax.set_title(feat, fontsize=10)
    ax.set_xlabel('')        # Feature name is already in the title so the x-label is redundant here
    ax.legend(fontsize=8)    # Smaller font keeps the legend from crowding the panel

# A single suptitle across all six panels ties the figure together
# y=1.01 nudges it above the top row of panels to avoid overlap
plt.suptitle('Figure 11: KDE distributions by class — spot the best separators',
             fontsize=12, y=1.01)
plt.tight_layout()
plt.show()

# The print statements below make the visual observation explicit:
# features where the two KDE curves barely overlap carry the most discriminative information
print('Visual inspection: features with minimal overlap between the two KDE curves')
print('are the most informative for classification (e.g. mean radius, mean perimeter).')
````


### 6.5 Box Plots

A **box plot** (also called a box-and-whisker plot) is a compact visual summary of a distribution. Where a histogram or KDE shows the full shape of the data, a box plot distils it down to five key numbers, making it particularly useful when you want to compare many features or many groups side by side without the visual noise of overlapping curves.

The five components of a box plot are:

| Component | What it shows |
|-----------|---------------|
| Bottom of box | Q1, the 25th percentile; 25% of values fall below this point |
| Top of box | Q3, the 75th percentile; 75% of values fall below this point |
| Box height | The **interquartile range (IQR)** = Q3 − Q1; the middle 50% of the data |
| Centre line | The **median** (Q2), the middle value of the distribution |
| Whiskers | Extend to the furthest data point still within 1.5 × IQR from the box edge |
| Individual dots | Points beyond the whiskers, flagged as **potential outliers** |

**Why 1.5 × IQR?** This threshold was proposed by the statistician John Tukey and has become the standard definition of a mild outlier. A point more than 1.5 × IQR beyond the box edge is unusually far from the bulk of the data, not necessarily an error, but worth investigating.

The IQR is a measure of spread that is robust to outliers. Unlike the standard deviation, which is pulled by extreme values, the IQR only describes the middle 50% of the data and is unaffected by what happens in the tails.

**When to use a box plot over a KDE:** box plots are preferable when you are comparing many groups or many features simultaneously. A grid of KDE curves can become visually cluttered; a grid of box plots remains readable because each distribution occupies a fixed, compact space on the axis.

The cell below shows two panels. The first compares the distribution of a single feature, `mean radius`, between the two diagnosis classes. The second places four features side by side on the same Axes, which requires a small but important data reshaping step using `pd.melt()`, explained in the code comments.

> ⚠️ **A note on scale:** when placing multiple features on the same axis as in Panel 2, be aware that features with very different ranges will compress the boxes of smaller-valued features. `mean area`, for example, has values in the hundreds. Placing it alongside `mean smoothness`, which ranges from 0 to 0.16, would make the smoothness boxes almost invisible. This is the feature scaling problem introduced in Section 4, and it applies to visualisation just as much as it does to modelling.


````python
# Listing 16.
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Panel 1: single feature split by class ---
# sns.boxplot() draws one box per unique value in x=, positioned along the x-axis.
# Passing x='diagnosis' means seaborn groups the data by class automatically —
# no manual filtering needed, unlike the histogram and KDE loops above.
# palette= maps each class label string to a colour.
ax = axes[0]
sns.boxplot(data=df, x='diagnosis', y='mean radius',
            hue='diagnosis', legend=False,
            palette={'Benign': 'steelblue', 'Malignant': 'tomato'}, ax=ax)
ax.set_title('mean radius by diagnosis')
ax.set_xlabel('Diagnosis')
ax.set_ylabel('mean radius')

# --- Panel 2: four features side by side ---
# To plot multiple features on the same Axes, seaborn expects the data in long format:
# one row per (observation, feature) pair, with a single column for the feature name
# and a single column for its value.
# Our DataFrame is currently in wide format — one column per feature — so we need
# to reshape it first using pd.melt().
#
# pd.melt() arguments:
#   id_vars=    column(s) to keep unchanged — here we preserve the class label
#   var_name=   name for the new column that stores the original column names (feature names)
#   value_name= name for the new column that stores the original values
#
# Example — one row of wide format:
#   diagnosis | mean radius | mean texture
#   Benign    | 13.2        | 18.4
#
# becomes two rows in long format:
#   diagnosis | feature      | value
#   Benign    | mean radius  | 13.2
#   Benign    | mean texture | 18.4
top4 = ['mean radius', 'mean texture', 'mean concavity', 'mean smoothness']
df_melt = df[top4 + ['diagnosis']].melt(
    id_vars='diagnosis', var_name='feature', value_name='value')

# hue='diagnosis' tells seaborn to split each feature's box by class,
# producing a pair of coloured boxes per feature
ax = axes[1]
sns.boxplot(data=df_melt, x='feature', y='value', hue='diagnosis',
            palette={'Benign': 'steelblue', 'Malignant': 'tomato'}, ax=ax)
ax.set_title('Four features side by side')
ax.set_xlabel('Feature')
ax.set_ylabel('Value')
ax.tick_params(axis='x', rotation=15)  # Rotate x-axis labels to prevent them overlapping

plt.suptitle('Figure 12: Box plots — distribution comparison by class', y=1.01)
plt.tight_layout()
plt.show()
````

Notice that Panel 2 illustrates an important limitation of placing multiple features on the same axis: `mean concavity` and `mean smoothness` operate on a much smaller numerical scale than `mean radius` and `mean texture`, so their boxes are compressed almost flat and are effectively unreadable. This is the feature scaling problem from Section 4 appearing in a visualisation context.

To make a meaningful multi-feature box plot comparison, the features should first be transformed onto a common scale, a process called **normalisation** or **standardisation**.


### 6.6 Violin Plots

A **violin plot** combines two things we have already covered, the five-number summary of a box plot and the smoothed density curve of a KDE, into a single chart. The result is shaped like a violin: wide where many observations are concentrated, narrow where few are.

This matters because a box plot, despite being compact and readable, can hide important structure in the data. Two distributions can have identical medians, identical IQRs, and identical whiskers yet look completely different: one could be unimodal and symmetric while the other is bimodal or heavily skewed. A violin plot exposes this structure immediately.

The trade-off is legibility. Violin plots take up more space than box plots and can be harder to read for audiences unfamiliar with density curves. A good compromise, used in the cell below, is to overlay a miniature box plot inside the violin using `inner='box'`, so both the distributional shape and the five-number summary are visible at once.

> 💡 **When to use each:** use a box plot when you need to compare many features or groups compactly, or when presenting to a non-specialist audience. Use a violin plot when the shape of the distribution matters, for example when you suspect bimodality or strong skewness that a box plot would obscure.


````python
# Listing 17.
fig, ax = plt.subplots(figsize=(10, 5))

# sns.violinplot() fits a KDE to the data for each class and mirrors it on both sides
# to produce the characteristic violin shape.
# x=        the column used to define the groups along the x-axis (one violin per class)
# y=        the column whose distribution is being visualised
# hue=      required in recent seaborn versions when using palette= with a categorical x variable
# legend=   set to False because the x-axis labels already identify the classes
# palette=  maps each class label to a colour, applied to both the violin fill and the inner box
# inner=    controls what is drawn inside the violin body:
#             'box'    — a miniature box plot showing median, IQR, and whiskers (used here)
#             'quart'  — horizontal lines at the quartiles only
#             'point'  — individual data points
#             None     — nothing, just the violin shape
# linewidth= thickness of the violin outline
sns.violinplot(data=df, x='diagnosis', y='mean area',
               hue='diagnosis', legend=False,
               palette={'Benign': 'steelblue', 'Malignant': 'tomato'},
               inner='box', linewidth=1.2, ax=ax)

# The title spans two lines using \n — the second line reminds the reader how to interpret width
ax.set_title('Figure 13: Violin plot — mean area by diagnosis\n'
             '(width reflects density: where the violin is widest, most data points lie)')
ax.set_xlabel('Diagnosis')
ax.set_ylabel('mean area')

plt.tight_layout()
plt.show()
````

## 7. Categorical Data: Class Balance

🔙 [Back to Table of Contents](#Table-of-Contents)


Before fitting any classifier, always check the **class distribution**. This is one of the most important and most commonly overlooked steps in any classification workflow.

**Class imbalance** occurs when one class has substantially more observations than another. To understand why this matters, consider a medical dataset where 95% of patients are healthy and only 5% have a disease. A model that simply predicts "healthy" for every single patient, without looking at any feature, would achieve 95% accuracy. That sounds impressive until you realise the model has never once correctly identified a sick patient. Accuracy, in this case, is a completely misleading measure of performance.

Class imbalance is common in real-world ML problems:

- **Fraud detection**: fraudulent transactions are rare relative to legitimate ones
- **Medical diagnosis**: diseases are often rare in the general population
- **Fault detection**: equipment failures are infrequent compared to normal operation

Visualising class balance early in the workflow sets appropriate expectations for which evaluation metrics to use. Accuracy alone is not sufficient when classes are imbalanced; more on that in later notebooks. The cell below shows the class distribution in two ways, raw counts and percentages, so both the absolute numbers and the relative proportions are immediately visible.


````python
# Listing 18.
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# --- Panel 1: raw counts ---
# sns.countplot() counts the occurrences of each unique value in x= automatically —
# unlike ax.bar(), it does not need pre-aggregated data; it works directly from the DataFrame.
# hue= and legend=False are required in recent seaborn versions when using palette= 
# with a categorical x variable to avoid a FutureWarning (same fix as the box plot above).
ax = axes[0]
sns.countplot(data=df, x='diagnosis',
              hue='diagnosis', legend=False,
              palette={'Benign': 'steelblue', 'Malignant': 'tomato'}, ax=ax)
ax.set_title('Class distribution (raw counts)')
ax.set_xlabel('Diagnosis')
ax.set_ylabel('Count')

# ax.patches is a list of the rectangle objects that make up the bars.
# Iterating over it lets us annotate each bar individually without hardcoding positions.
# p.get_height()              — the height of the bar, which equals the count
# p.get_x() + p.get_width()/2 — the horizontal centre of the bar
# ha='center'                 — centres the text label over that position
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2, p.get_height() + 4),
                ha='center', fontsize=11)

# --- Panel 2: proportions as percentages ---
# We compute percentages manually here because we want to annotate with percentage
# strings — sns.countplot() gives us raw counts but not the proportions directly.
# value_counts(normalize=True) returns each class as a proportion of the total (sums to 1).
# Multiplying by 100 converts those proportions to percentages.
ax = axes[1]
counts = df['diagnosis'].value_counts(normalize=True) * 100

# ax.bar() is used directly rather than a seaborn function because the data is
# already aggregated — seaborn's count-based functions expect raw unaggregated data.
# color= accepts a list, assigning one colour per bar in order.
bars = ax.bar(counts.index, counts.values,
              color=['steelblue', 'tomato'],
              edgecolor='white', linewidth=0.5)

ax.set_title('Class balance (percentage)')
ax.set_xlabel('Diagnosis')
ax.set_ylabel('%')
ax.set_ylim(0, 80)  # Manual limit gives headroom above the tallest bar for the annotation

# zip(bars, counts.values) pairs each bar rectangle object with its computed percentage value.
# bar.get_x() + bar.get_width()/2 finds the horizontal centre of each bar,
# and val + 0.8 places the label just above the bar top.
for bar, val in zip(bars, counts.values):
    ax.text(bar.get_x() + bar.get_width() / 2, val + 0.8,
            f'{val:.1f}%', ha='center', fontsize=11)

plt.suptitle('Figure 14: Class balance — raw counts and percentages', y=1.01)
plt.tight_layout()
plt.show()

# A concrete reminder of why class imbalance matters in practice —
# a classifier that always predicts the majority class would score high accuracy
# on this dataset while completely failing to detect malignant cases
print('Class imbalance matters: if Malignant cases are rare, a naive model')
print('that always predicts Benign would score high accuracy but miss every cancer.')
````

## 8. Relationships Between Features: Scatter Plots

🔙 [Back to Table of Contents](#Table-of-Contents)

Scatter plots reveal the **joint distribution** of two continuous features, that is, not just how each feature behaves individually, but how they relate to each other. Each point represents one observation, plotted at the coordinates formed by its two feature values.

This matters for ML because models do not learn from features in isolation. A feature that appears uninformative on its own can become highly useful when combined with another. Conversely, two features that each look useful individually might carry largely the same information, in which case including both adds redundancy without adding predictive power. Scatter plots give you an immediate visual sense of both of these situations.

Colouring points by class label turns a scatter plot into a two-dimensional **decision boundary preview**. If the two colours form distinct clusters with little overlap, those two features together carry strong discriminative information and a classifier should be able to separate them well. If the colours are thoroughly mixed, no straight line or simple boundary drawn in that two-dimensional space will cleanly separate the classes.

You can encode a **third variable** by mapping it to point size (`s=`) or colour intensity (`alpha=`), effectively turning a two-dimensional plot into a three-variable display. This can be powerful, but use it with care: overloading a scatter plot with too many visual encodings makes it harder to read, not easier.

The cell below shows two panels for the same feature pair. The first plots the raw relationship with no class information; the second colours each point by diagnosis. Comparing the two panels illustrates how much information the class label adds to the same underlying data.

> 📌 **What to notice:** the plain scatter in Panel 1 already shows a clear positive relationship between `mean radius` and `mean area`: larger radii correspond to larger areas, which is geometrically expected. Panel 2 reveals that malignant tumours tend to occupy the upper end of both scales, forming a cluster that is somewhat, though not perfectly, separable from the benign cases.

````python
# Listing 19.
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Panel 1: plain scatter — overall relationship, no class information ---
# ax.scatter() places one marker at (x, y) for each observation.
# s=         marker size in points squared — kept small (30) as we have hundreds of points
# alpha=     transparency; values below 1 prevent dense regions from appearing as a solid block,
#            making it easier to see where points are concentrated
# edgecolors= thin black border on each marker helps distinguish overlapping points
# linewidth=  thickness of that border — kept very thin so it does not dominate at small marker sizes
ax = axes[0]
ax.scatter(df['mean radius'], df['mean area'],
           color='steelblue', s=30, alpha=0.5,
           edgecolors='k', linewidth=0.2)
ax.set_xlabel('mean radius')
ax.set_ylabel('mean area')
ax.set_title('mean radius vs mean area')

# --- Panel 2: coloured by diagnosis — class separation becomes visible ---
# A dictionary maps each class label to a colour, making it easy to loop over
# both classes with consistent colour assignments throughout the notebook.
ax = axes[1]
palette = {'Benign': 'steelblue', 'Malignant': 'tomato'}

for cls, col in palette.items():
    # Filter the DataFrame to this class only, then plot its points separately.
    # Plotting each class in its own ax.scatter() call is necessary because
    # ax.scatter() does not group by a categorical variable automatically —
    # unlike seaborn functions, it expects a single colour for the entire call.
    sub = df[df['diagnosis'] == cls]
    ax.scatter(sub['mean radius'], sub['mean area'],
               color=col, s=30, alpha=0.6,
               edgecolors='k', linewidth=0.2,
               label=cls)   # label= assigns this class to a legend entry

ax.set_xlabel('mean radius')
ax.set_ylabel('mean area')
ax.set_title('Coloured by diagnosis')
ax.legend()

# Figure number updated to match the section numbering used throughout this notebook
plt.suptitle('Figure 15: Scatter plots — feature relationships with and without class colouring', y=1.01)
plt.tight_layout()
plt.show()
````

The cell below demonstrates two further techniques that extend the basic scatter plot.

**Panel 1, regression lines with confidence bands:** `sns.regplot()` fits a linear regression line to the data and overlays it on the scatter plot, with a shaded band representing the 95% confidence interval around that line. Plotting one regression line per class reveals not just whether the classes occupy different regions of the feature space, but whether the *relationship* between the two features differs between them. A confidence band that is wide indicates uncertainty in the fitted line, usually because there are few observations in that region.

**Panel 2, size encoding a third variable:** by mapping a third feature to the `s=` (size) parameter of `ax.scatter()`, we can display three variables simultaneously in a two-dimensional plot. Here, point size encodes `mean area`, with larger points representing observations with a larger area. Because raw feature values can span a wide range, the sizes are first **normalised** to a fixed pixel range (20–170) so that no single point dominates the figure visually.

**A note on readability:** size encoding works well when the third variable has a wide, continuous range and the relationship is approximately monotonic. It becomes harder to read when points overlap heavily or when the size differences are subtle. If the third variable is categorical, colour is almost always a better choice than size.

````python
# Listing 20.
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Panel 1: per-class regression lines with 95% confidence bands ---
# We loop over both classes and call sns.regplot() once per class on the same Axes.
# sns.regplot() does two things in a single call:
#   1. Draws a scatter plot of the raw data points
#   2. Fits a linear regression line and overlays it with a shaded 95% confidence band
# The confidence band is computed by bootstrapping — the wider it is, the less
# certain the fitted line is, usually because there are fewer points in that region.
# scatter_kws= passes a dictionary of keyword arguments through to the underlying
# ax.scatter() call, allowing us to control point size and transparency independently
# of the line and band styling.
for cls, col, ax_label in [('Benign', 'steelblue', True), ('Malignant', 'tomato', True)]:
    sns.regplot(data=df[df['diagnosis'] == cls],
                x='mean radius', y='mean perimeter',
                ax=axes[0],
                scatter_kws={'s': 25, 'alpha': 0.5},  # Controls the scatter points only
                color=col,                              # Controls both the line and the band
                label=cls)

axes[0].set_title('Scatter + regression line per class')
axes[0].legend()

# --- Panel 2: three-variable display — x, y, and point size encoding a third feature ---
ax = axes[1]

# df.sample(n, random_state=) draws n rows at random from the DataFrame.
# We sample 120 rows rather than plotting all observations to reduce overplotting —
# with size encoding, heavily overlapping points of different sizes become difficult to read.
# random_state= fixes the random seed so the same 120 rows are selected every time.
sub = df.sample(120, random_state=1)

# ax.scatter() expects s= to be a numeric value or array of values in points squared.
# Raw feature values are rarely suitable directly — mean area ranges into the hundreds,
# which would produce enormous points that dominate the figure.
# We normalise to the range [20, 170] using min-max scaling:
#   (value - min) / (max - min) maps all values to [0, 1]
#   multiplying by 150 stretches that to [0, 150]
#   adding 20 shifts the range to [20, 170], ensuring no point disappears entirely
sizes = (sub['mean area'] - sub['mean area'].min()) / \
        (sub['mean area'].max() - sub['mean area'].min()) * 150 + 20

# .map() replaces each class label string with its corresponding colour string,
# producing a Series of colour values aligned with the DataFrame index
colours = sub['diagnosis'].map({'Benign': 'steelblue', 'Malignant': 'tomato'})

# c= accepts an array of colour values — one per point — unlike color= which takes a single value
ax.scatter(sub['mean radius'], sub['mean concavity'],
           s=sizes, c=colours, alpha=0.6,
           edgecolors='k', linewidth=0.2)

ax.set_xlabel('mean radius')
ax.set_ylabel('mean concavity')
ax.set_title('Three variables: x, y, and point size = mean area')

# When c= is used with an array of colour strings, matplotlib does not generate
# legend entries automatically — there is no label= to attach to individual points.
# mpatches.Patch() creates a plain coloured rectangle that can be passed to ax.legend()
# as a manual legend entry, giving us class labels without needing to re-plot anything.
ax.legend(handles=[
    mpatches.Patch(color='steelblue', label='Benign'),
    mpatches.Patch(color='tomato',    label='Malignant'),
])

plt.suptitle('Figure 16: Advanced scatter — regression lines and size encoding', y=1.01)
plt.tight_layout()
plt.show()
````

## 9. Correlation: Heatmaps and Target Alignment

🔙 [Back to Table of Contents](#Table-of-Contents)

In Notebook 3 we introduced the **Pearson correlation coefficient** $r$ and the **correlation matrix**. As a brief recap: $r$ measures the strength and direction of the *linear* relationship between two variables, and always lies between $-1$ and $+1$. A value close to $+1$ means the two variables tend to increase together; close to $-1$ means one increases as the other decreases; close to $0$ means no detectable linear relationship. The correlation matrix extends this to an entire dataset by computing $r$ for every possible pair of features at once and arranging the results in a square grid. If either concept feels unfamiliar, it is worth revisiting Sections 7.4 and 7.5 of Notebook 3 before continuing.

Here we apply the same idea to our real dataset. The correlation matrix gives a bird's-eye view of all pairwise feature relationships simultaneously, and in the context of feature selection it answers two questions that are central to building a good classifier:

1. **Which features correlate with the target?** A feature with a high absolute $r$ value relative to the diagnosis label is potentially useful: it means the feature value tends to differ systematically between benign and malignant cases.

2. **Which features correlate with each other?** A high $r$ between two input features indicates **multicollinearity**, meaning the two features are carrying largely the same information. Including both in a model adds complexity without adding predictive power, and in some models it can actively destabilise the parameter estimates.

As a reminder, $r$ always lies in the range $[-1, +1]$:

| Value | Meaning |
|-------|---------|
| $r = +1$ | Perfect positive linear relationship |
| $r = -1$ | Perfect negative linear relationship |
| $r \approx 0$ | No linear relationship detectable |
| $\|r\| > 0.8$ | Strong correlation, worth investigating for redundancy |

One important limitation to keep in mind: **Pearson $r$ only detects linear relationships**. Two features can have a strong non-linear dependency and still produce an $r$ close to zero. We will address this later in this notebook using **mutual information**, which captures non-linear associations and is a more general alternative for feature selection.

A correlation matrix is most useful when visualised as a **heatmap**, a grid where each cell is filled with a colour that encodes the $r$ value it contains. This makes patterns of high and low correlation immediately visible without having to read individual numbers. Seaborn's `sns.heatmap()` handles this in a single call, but it has several important parameters worth understanding before looking at the full example:

```python
sns.heatmap(
    corr,
    annot=True,              # Print the r value as text inside each cell
    fmt='.2f',               # Format each value to 2 decimal places
    annot_kws={'size': 8},   # Font size for the annotations, kept small to fit in each cell
    cmap='RdBu_r',           # Diverging colour map: deep blue = +1, deep red = -1, white = 0
                             # The _r suffix reverses the map so blue = positive (intuitive)
    vmin=-1, vmax=1,         # Anchor the colour scale to the full theoretical range of r
                             # so colours are comparable across different datasets
    linewidths=0.5,          # Width of the grid lines drawn between cells
    linecolor='white',       # White grid lines make individual cells easier to distinguish
    ax=ax
)
```

The choice of `cmap='RdBu_r'` is deliberate. A **diverging colour map** is appropriate here because $r$ has a meaningful centre point at zero: colours should diverge away from a neutral midpoint in two opposite directions, one for positive correlations and one for negative. A sequential colour map (which runs from light to dark in one direction) would be misleading because it would suggest that all correlations are on a one-directional scale rather than centred on zero.

Setting `vmin=-1` and `vmax=1` is equally important. Without these, seaborn scales the colours to the range of values actually present in the matrix. If the strongest correlation in your dataset is only $r = 0.7$, that cell would be coloured as deep blue even though 0.7 is not a particularly strong correlation in absolute terms. Anchoring to $[-1, +1]$ ensures the colours are always interpretable on the same absolute scale.

To keep the heatmap readable, the cell below computes the correlation matrix for the `mean` features only (the ten features whose names begin with `mean`) rather than all thirty features in the dataset. The same approach applies to any subset of columns.

````python
# Listing 21.
# A list comprehension selects only the columns whose names begin with 'mean'.
# Using a subset keeps the heatmap to a manageable size — plotting all 30 features
# would produce a grid too dense to read comfortably.
mean_cols = [c for c in df.columns if c.startswith('mean')]

# .corr() computes the Pearson r between every pair of columns in the DataFrame
# and returns a square symmetric matrix with 1.0 on the diagonal.
corr = df[mean_cols].corr()

fig, ax = plt.subplots(figsize=(11, 9))

sns.heatmap(
    corr,
    annot=True,              # Print the r value as text inside each cell
    fmt='.2f',               # Format each value to 2 decimal places
    annot_kws={'size': 8},   # Font size for the annotations — kept small to fit in each cell
    cmap='RdBu_r',           # Diverging colour map: deep blue = +1, deep red = -1, white = 0
                             # The _r suffix reverses the map so blue = positive (intuitive)
    vmin=-1, vmax=1,         # Anchor the colour scale to the full theoretical range of r
                             # so colours are comparable across different datasets
    linewidths=0.5,          # Width of the grid lines drawn between cells
    linecolor='white',       # White grid lines make individual cells easier to distinguish
    ax=ax
)

ax.set_title('Figure 17: Pearson correlation matrix — mean features', fontsize=13)
ax.tick_params(axis='x', rotation=40)  # Rotate x-axis labels to prevent overlap
ax.tick_params(axis='y', rotation=0)   # Keep y-axis labels horizontal for readability

plt.tight_layout()
plt.show()

# The heatmap shows the full matrix, but the upper and lower triangles are mirror images
# of each other — every pair appears twice. To list the most correlated pairs without
# duplicates we extract the upper triangle only, excluding the diagonal.
#
# np.triu(np.ones(corr.shape), k=1) creates a matrix of the same shape as corr
# filled with 1s in the upper triangle and 0s elsewhere.
# k=1 means the diagonal itself is excluded (k=0 would include it).
# .astype(bool) converts 1s to True and 0s to False.
# corr.where(mask) retains values where the mask is True and replaces the rest with NaN.
# .stack() converts the 2D matrix to a Series, automatically dropping NaN entries —
# leaving only the upper triangle values with their row and column labels as the index.
# .reset_index() promotes those index labels back into regular columns.
pairs = (
    corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
    .stack()
    .reset_index()
)

pairs.columns = ['feature_1', 'feature_2', 'r']

# Sort by absolute r value so the strongest correlations appear first regardless of direction
pairs = pairs.reindex(pairs['r'].abs().sort_values(ascending=False).index)

print('Top 5 most correlated feature pairs:')
print(pairs.head(5).to_string(index=False))
````

---

The heatmap reveals relationships *between* features, but it does not directly tell us which features are most relevant to what we are actually trying to predict. For that we need to measure each feature's correlation with the **target variable** directly.

Computing this produces a simple ranking of features by their linear association with the class label, a quick and interpretable first pass at feature selection. Features with a high absolute $r$ with the target are worth prioritising; features with $r$ close to zero have little linear relationship with the outcome and may not contribute much to a classifier.

The bar chart below visualises this ranking. Because the target variable is encoded as a binary value (1 = Benign, 0 = Malignant), the sign of $r$ tells us something useful about the direction of the relationship:

- A **positive $r$** (blue) means higher values of that feature are associated with benign cases: as the feature value increases, the target value tends toward 1.
- A **negative $r$** (red) means higher values of that feature are associated with malignant cases: as the feature value increases, the target value tends toward 0.

> ⚠️ **Keep the limitation in mind:** this ranking is based entirely on linear correlation. A feature with a low $r$ here is not necessarily uninformative: it may have a strong *non-linear* relationship with the target that Pearson $r$ cannot detect. We will address this shortly using mutual information, which captures both linear and non-linear associations.


````python
# Listing 22.
# .corr() computes the full correlation matrix for the selected columns.
# Selecting the 'target' column from the result gives us a Series of r values —
# one per feature — representing each feature's linear association with the target.
# .drop('target') removes the target's self-correlation (always 1.0) from the Series.
target_corr = df[mean_cols + ['target']].corr()['target'].drop('target')

# Sort by absolute r value so the most strongly correlated features appear at the top
# of the bar chart, regardless of whether the correlation is positive or negative.
target_corr_sorted = target_corr.reindex(target_corr.abs().sort_values(ascending=False).index)

fig, ax = plt.subplots(figsize=(9, 6))

# Assign a colour to each bar based on the direction of the correlation.
# A list comprehension iterates over the sorted values and picks the appropriate colour:
# steelblue = positive r (higher feature value → more likely Benign)
# tomato    = negative r (higher feature value → more likely Malignant)
colours = ['steelblue' if v > 0 else 'tomato' for v in target_corr_sorted]

# ax.barh() draws horizontal bars — preferable to vertical bars here because
# the feature names are long strings that would overlap on a vertical x-axis.
ax.barh(target_corr_sorted.index, target_corr_sorted.values,
        color=colours, edgecolor='white', linewidth=0.4)

# A vertical line at x=0 provides a clear visual anchor separating positive
# and negative correlations and makes bar lengths easier to judge.
ax.axvline(0, color='black', linewidth=0.8)

# The x-axis label clarifies the binary encoding of the target variable —
# without this the sign of r would be ambiguous to the reader.
ax.set_xlabel('Pearson r with target  (1 = Benign, 0 = Malignant)')

ax.set_title('Figure 18: Feature–target correlation\n'
             'Blue = positive (higher value → more likely Benign)\n'
             'Red = negative (higher value → more likely Malignant)')

# Grid lines on the x-axis only — makes bar lengths easier to compare
# without adding visual clutter to the chart.
ax.grid(True, alpha=0.25, axis='x')

plt.tight_layout()
plt.show()
````

## 10. The Pairplot: A Complete Multivariate Overview

🔙 [Back to Table of Contents](#Table-of-Contents)

Everything we have done so far has examined either one feature at a time (histograms, KDEs, box plots, violin plots) or one pair of features at a time (scatter plots). A **pairplot** (also called a scatter matrix) scales this up by visualising every possible feature pair simultaneously in a single figure.

For $p$ features, a pairplot produces a $p \times p$ grid of subplots:

- **Diagonal panels** show the marginal distribution of each individual feature, here displayed as per-class KDE curves, so you can immediately see how each feature's distribution differs between benign and malignant cases.
- **Off-diagonal panels** show a scatter plot of every pair of features, with points coloured by class label.

The result is the most information-dense single plot in the exploratory toolkit. Reading across a row or down a column tells you how one feature relates to every other feature. Scanning the entire grid gives you an immediate multi-feature separability assessment: panels where the two colours form distinct, non-overlapping clusters identify feature pairs that together carry strong discriminative information, the kind that a classifier can exploit.

`sns.pairplot()` generates the entire grid in a single call. Unlike most seaborn functions which draw onto an existing `ax`, `sns.pairplot()` creates its own figure internally and returns a `PairGrid` object, here assigned to `g`. The underlying matplotlib Figure is accessible via `g.fig`, which is how we attach a `suptitle` to the overall figure.

The key parameters are:

```python
g = sns.pairplot(
    df[subset_cols],
    hue='diagnosis',                                        # Colour points by class label across all off-diagonal panels
    palette={'Benign': 'steelblue', 'Malignant': 'tomato'}, # Colour mapping for each class
    diag_kind='kde',                                        # Draw KDE curves on the diagonal instead of histograms
    plot_kws={'alpha': 0.5, 's': 25, 'edgecolor': 'none'}, # Keyword arguments passed to each off-diagonal scatter plot
    diag_kws={'fill': True, 'alpha': 0.4}                  # Keyword arguments passed to each diagonal KDE plot
)
```

The `plot_kws` and `diag_kws` parameters are dictionaries that pass styling arguments through to the underlying plot functions: `plot_kws` controls the scatter plots and `diag_kws` controls the KDEs. This is the same pattern as `scatter_kws` in `sns.regplot()` encountered in Section 8. Keeping `alpha` low in both is important here: with hundreds of points across many panels, full opacity would turn the scatter plots into solid blobs and the KDE fills would obscure each other where the classes overlap.

> ⚠️ **Keep the feature count small.** A pairplot with $p$ features produces $p^2$ subplots. With 5 features that is 25 panels, already a large figure. With 10 features it is 100 panels, which is unreadable. Always select a meaningful subset of 4–6 features, guided by what the correlation heatmap and feature–target bar chart revealed in the previous sections.

The cell below produces a pairplot for four features identified as informative in the earlier analysis, with per-class KDEs on the diagonal and class-coloured scatter plots off the diagonal.

````python
# Listing 23.
# The feature subset is defined explicitly here — guided by the correlation heatmap
# and feature–target bar chart above. Including 'diagnosis' in the column selection
# is required so that hue='diagnosis' has a column to read class labels from.
subset_cols = ['mean radius', 'mean texture',
               'mean concavity', 'mean symmetry', 'diagnosis']

# sns.pairplot() creates its own Figure internally and returns a PairGrid object.
# Unlike most seaborn functions, it does not accept an ax= argument — it manages
# its own grid of subplots. The returned object g gives access to the underlying
# Figure via g.fig, which we use below to attach a suptitle.
g = sns.pairplot(
    df[subset_cols],
    hue='diagnosis',                                         # Colour all off-diagonal scatter points by class
    palette={'Benign': 'steelblue', 'Malignant': 'tomato'},  # Class-to-colour mapping
    diag_kind='kde',                                         # KDE curves on the diagonal rather than histograms
                                                             # — better for comparing class distributions
    plot_kws={'alpha': 0.5, 's': 25, 'edgecolor': 'none'},  # Passed through to each off-diagonal ax.scatter() call:
                                                             #   alpha= low to prevent overplotting
                                                             #   s=     small marker size for a dense grid
                                                             #   edgecolor='none' removes marker borders at this scale
    diag_kws={'fill': True, 'alpha': 0.4}                   # Passed through to each diagonal sns.kdeplot() call:
                                                             #   fill=  shades the area under each KDE curve
                                                             #   alpha= low so both class curves are visible where they overlap
)

# g.fig is the underlying matplotlib Figure — suptitle must be set here rather
# than via plt.suptitle() because pairplot manages its own Figure object.
# y=1.01 nudges the title above the top row of panels.
g.fig.suptitle('Figure 19: Pairplot — scatter and KDE for a feature subset', y=1.01)
plt.show()

# A plain-text reading guide printed below the figure — useful for readers
# who have not encountered a pairplot before.
print('How to read a pairplot:')
print('  Off-diagonal: scatter of row feature vs column feature, coloured by class.')
print('  Diagonal:     KDE of that feature for each class.')
print('  Look for panels where the two colours separate clearly — those feature')
print('  pairs are strong discriminators.')
````

## 11. ML-Specific Visualisations

🔙 [Back to Table of Contents](#Table-of-Contents)

The visualisations covered so far have all been concerned with understanding the data before any model is built. This section shifts focus to **model evaluation**: visualisations that help us understand how well a trained classifier is performing, where it is making mistakes, and whether those mistakes could be reduced with more data or a different approach.

Three visualisation types appear in almost every ML evaluation workflow:

- **Confusion matrix**: a grid showing exactly where a classifier makes errors, broken down by class. Rather than summarising performance as a single number, it shows which classes are being confused with which.
- **ROC curve**: plots the trade-off between the true positive rate and false positive rate across every possible decision threshold, giving a threshold-independent view of classifier performance.
- **Learning curve**: shows how model performance changes as the amount of training data increases, which is the primary diagnostic tool for identifying whether a model is underfitting, overfitting, or simply data-hungry.

> 📓 **Coming up:** each of these is introduced visually here. Notebooks 10, 11, and 12 cover the underlying theory in depth, including accuracy, precision, recall, F1 score, the confusion matrix, ROC curves, and the AUC metric.

Before we can produce any of these plots, we need a trained model and a set of predictions to evaluate. The cell below prepares the data, trains a Logistic Regression classifier on the `mean` features, and generates predictions on a held-out test set. Two steps in this pipeline deserve particular attention:

**Train/test splitting**: the model is trained on 80% of the data and evaluated on the remaining 20%, which it has never seen. This is the minimum requirement for an honest estimate of generalisation performance. Evaluating on the same data used for training would give an artificially optimistic result.

**Feature scaling**: Logistic Regression uses a gradient-based optimisation algorithm to find its parameters. When features are on very different numerical scales, the optimisation landscape becomes poorly conditioned and the algorithm converges slowly or not at all. `StandardScaler` addresses this by transforming each feature to have zero mean and unit variance, a process called **standardisation**.

> ⚠️ **Data leakage:** the scaler must be fitted on the training data only, then applied to the test data using the training statistics. Fitting the scaler on the full dataset before splitting, a common mistake, would allow information about the test set to influence the scaling, producing an overly optimistic evaluation. This is called **data leakage** and it is one of the most common sources of inflated performance estimates in practice.

````python
# Listing 24.
# Prepare data and train a Logistic Regression classifier.
# This model's predictions are used in the three evaluation visualisations below.

mean_cols = [c for c in df.columns if c.startswith('mean')]  # Re-define for clarity
X = df[mean_cols].values
y = df['target'].values

# Split into training and test sets: 80% train, 20% test
# random_state=0 ensures the same split every time this cell is run
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2, random_state=0)

# Standardise features: subtract the mean and divide by the standard deviation
# IMPORTANT: fit the scaler ONLY on training data, then apply to test data
# Fitting on the full dataset would leak test information into training (data leakage)
scaler = StandardScaler()
X_tr_s = scaler.fit_transform(X_tr)  # fit_transform: compute mean/std and scale
X_te_s = scaler.transform(X_te)      # transform: apply the already-fitted scaler

lr = LogisticRegression(max_iter=1000, random_state=0)
lr.fit(X_tr_s, y_tr)
y_pred = lr.predict(X_te_s)

print('Logistic Regression trained.')
print(f'Test accuracy: {lr.score(X_te_s, y_te):.4f}')
````

### 11.1 Confusion Matrix

When a classifier makes a prediction, there are four possible outcomes depending on the combination of the true label and the predicted label. A **confusion matrix** arranges these four outcomes into a table, giving a complete picture of where the model is right and where it is wrong:

| | Predicted Negative | Predicted Positive |
|---|---|---|
| **Actual Negative** | **True Negative (TN)**: correctly predicted negative | **False Positive (FP)**: incorrectly predicted positive |
| **Actual Positive** | **False Negative (FN)**: incorrectly predicted negative | **True Positive (TP)**: correctly predicted positive |

The diagonal cells (TN and TP) represent correct predictions. The off-diagonal cells (FP and FN) represent errors. A perfect classifier would have zeros in both off-diagonal cells.

The distinction between the two error types matters enormously in practice, and the appropriate balance between them depends entirely on the problem:

- A **False Positive** in a cancer screening context means telling a healthy patient they may have cancer, causing anxiety and triggering unnecessary follow-up tests. Costly, but recoverable.
- A **False Negative** means telling a patient with cancer that they are healthy, a missed diagnosis that could be fatal. In most medical screening contexts, this error is far more serious.

This asymmetry is exactly why a single accuracy figure is insufficient. A model could achieve 90% accuracy while producing an unacceptable number of false negatives, and the confusion matrix is the only standard summary that makes this visible.

> 📓 **Precision, recall, and F1:** the confusion matrix is the foundation for the metrics covered in Notebooks 11 and 12. Precision measures how often a positive prediction is correct (TP / (TP + FP)). Recall measures how often actual positives are correctly identified (TP / (TP + FN)). F1 score balances the two. These are printed below the heatmap by `classification_report()`.

Displaying the confusion matrix as a heatmap rather than a plain table makes the pattern of errors immediately visible: darker cells contain more predictions, so a well-performing model will have dark diagonal cells and light off-diagonal cells.


````python
# Listing 25.
# confusion_matrix(y_true, y_pred) computes the 2×2 count matrix from the true
# and predicted labels. The rows correspond to actual classes and the columns
# to predicted classes — in the order they appear in the label array.
cm = confusion_matrix(y_te, y_pred)

fig, ax = plt.subplots(figsize=(6, 5))

# fmt='d' formats the annotated cell values as integers rather than floats —
# counts should never be displayed as 42.0 when 42 is clearer.
# cmap='Blues' is a sequential colour map appropriate here because the values
# (counts) have a natural zero baseline and no meaningful negative direction.
# xticklabels and yticklabels must be specified explicitly to replace the
# default 0/1 integers with meaningful class names.
sns.heatmap(cm, annot=True, fmt='d',
            cmap='Blues',
            xticklabels=['Malignant', 'Benign'],
            yticklabels=['Malignant', 'Benign'],
            linewidths=0.5, linecolor='white', ax=ax)

ax.set_xlabel('Predicted label')
ax.set_ylabel('True label')
ax.set_title('Figure 20: Confusion matrix — Logistic Regression')

plt.tight_layout()
plt.show()

# classification_report() prints a per-class summary of precision, recall, F1 score,
# and support (the number of true instances for each class).
# target_names= replaces the default 0/1 labels with meaningful class name strings.
print(classification_report(y_te, y_pred, target_names=['Malignant', 'Benign']))
````

### 11.2 ROC Curve and AUC

The confusion matrix we looked at in the previous section is computed at a single fixed decision threshold: by default, a classifier predicts the positive class when its estimated probability exceeds 0.5. But that threshold is arbitrary. In a cancer screening context, we might deliberately lower the threshold to catch more malignant cases, accepting more false positives in exchange for fewer missed cancers. Raising the threshold has the opposite effect.

The **Receiver Operating Characteristic (ROC) curve** makes this trade-off explicit by plotting classifier performance across *every possible threshold*, not just 0.5:

- The **x-axis** shows the **False Positive Rate (FPR)**: the proportion of actual negative cases incorrectly predicted as positive. Also called 1 − specificity.
- The **y-axis** shows the **True Positive Rate (TPR)**: the proportion of actual positive cases correctly identified. Also called sensitivity or recall.

As the threshold is lowered from 1 to 0, the classifier becomes more aggressive: it predicts positive more often, which increases both TPR (more true positives caught) and FPR (more false positives introduced). Each threshold value produces one point on the curve; the ROC curve is the line connecting all of them.

Two reference points anchor interpretation:

- A **perfect classifier** would immediately reach TPR = 1 with FPR = 0; its curve hugs the top-left corner of the plot.
- A **random classifier** performs no better than a coin flip; its curve follows the diagonal from (0, 0) to (1, 1).

The further the curve bows toward the top-left corner, the better the classifier is at separating the classes across all thresholds.

**Area Under the Curve (AUC)** summarises the entire ROC curve as a single number, which makes it easy to compare classifiers. It has a clean probabilistic interpretation:

$$
\text{AUC} = P(\text{score}_{\text{positive}} > \text{score}_{\text{negative}})
$$

This is the probability that the model assigns a higher confidence score to a randomly selected positive example than to a randomly selected negative example. An AUC of 1.0 is perfect; 0.5 is equivalent to random guessing; anything below 0.5 means the model is performing worse than chance.

| AUC range | Interpretation |
|-----------|---------------|
| 1.00 | Perfect classifier |
| 0.90 – 0.99 | Excellent |
| 0.80 – 0.89 | Good |
| 0.70 – 0.79 | Fair |
| 0.60 – 0.69 | Poor |
| 0.50 | Random guessing, no discriminative ability |

The AUC is generally a more informative metric than accuracy, particularly when classes are imbalanced. Because it evaluates the model across all thresholds rather than at a single fixed point, it reflects the model's underlying discriminative ability rather than its behaviour at one arbitrary cutoff.

````python
# Listing 26.
# lr.predict_proba() returns a 2D array of shape (n_samples, 2).
# Column 0 contains P(Malignant | x) and column 1 contains P(Benign | x).
# We extract column 1 — the probability of the positive class — because
# roc_curve() needs a continuous confidence score, not a hard 0/1 prediction.
# Using probabilities rather than hard predictions is what allows the curve to
# be evaluated across all possible thresholds rather than just the default 0.5.
y_prob = lr.predict_proba(X_te_s)[:, 1]

# roc_curve() sweeps through every unique value in y_prob as a candidate threshold.
# At each threshold it computes how many positives are correctly caught (TPR)
# and how many negatives are incorrectly flagged (FPR).
# It returns three aligned arrays — one value per threshold point on the curve:
#   fpr:        False Positive Rate at each threshold
#   tpr:        True Positive Rate at each threshold
#   thresholds: the probability threshold that produced each (fpr, tpr) pair
fpr, tpr, thresholds = roc_curve(y_te, y_prob)

# auc() computes the area under the ROC curve by numerical integration (trapezoidal rule).
# The result is a single scalar between 0 and 1.
roc_auc = auc(fpr, tpr)

fig, ax = plt.subplots(figsize=(7, 6))

# The ROC curve itself — each point is one (FPR, TPR) pair at a different threshold.
# The AUC value is embedded in the legend label so it is visible without reading the print output.
ax.plot(fpr, tpr, color='steelblue', linewidth=2.5,
        label=f'Logistic Regression  (AUC = {roc_auc:.3f})')

# The diagonal reference line represents a random classifier (AUC = 0.50) —
# one that assigns scores independently of the true class label.
# Any curve above this line is performing better than chance.
ax.plot([0, 1], [0, 1], color='gray', linewidth=1.5,
        linestyle='--', label='Random classifier  (AUC = 0.50)')

# Shading the area under the ROC curve gives a visual sense of the AUC magnitude —
# the larger the shaded region, the further the curve bows toward the top-left corner.
ax.fill_between(fpr, tpr, alpha=0.12, color='steelblue')

ax.set_xlabel('False Positive Rate  (1 − specificity)')
ax.set_ylabel('True Positive Rate  (sensitivity)')
ax.set_title('Figure 21: ROC curve — Logistic Regression')
ax.legend(fontsize=10)

# Fix both axes to [0, 1] so the plot is always square and the diagonal reference
# line runs exactly corner to corner. The y-axis upper limit is set to 1.02 to
# prevent the curve from being clipped at TPR = 1.
ax.set_xlim(0, 1)
ax.set_ylim(0, 1.02)

plt.tight_layout()
plt.show()

# The AUC printed here is the probability that the model assigns a higher score
# to a randomly chosen benign case than to a randomly chosen malignant case.
print(f'AUC = {roc_auc:.4f}')
print('Interpretation: probability that the model ranks a random Benign')
print('example higher than a random Malignant example.')
````

### 11.3 Learning Curve

The confusion matrix and ROC curve tell us how well the model performs on the test set we have. The **learning curve** asks a different and equally important question: *why* is the model performing the way it is, and what would most improve it?

A learning curve plots two accuracy scores against training set size: the accuracy measured on the training data itself, and the accuracy measured on held-out validation data via cross-validation. By watching how these two lines behave as more training data is added, we can diagnose the model's fundamental limitation:

| Pattern | Diagnosis | Likely remedy |
|---------|-----------|---------------|
| Validation accuracy still rising at maximum training size | **Data-hungry**: the model has not yet seen enough examples to generalise well | Collect more training data |
| Training and validation accuracy have converged but both are low | **Underfitting** (high bias): the model is too simple to capture the structure in the data | Use a more complex model or add more informative features |
| Training accuracy is high but there is a large persistent gap to validation accuracy | **Overfitting** (high variance): the model has memorised the training data rather than learning general patterns | Regularisation, more training data, or feature reduction |
| Training and validation accuracy have converged and both are high | **Well-fitted**: the model is generalising effectively | No action needed |

> 💡 **Bias and variance:** underfitting is also called **high bias**, meaning the model makes strong, incorrect assumptions about the data. Overfitting is also called **high variance**, meaning the model is too sensitive to the specific training examples it saw. The learning curve is the standard visual tool for distinguishing between these two failure modes, which require opposite remedies. We will return to the bias-variance trade-off in detail later in the series.

The code cell below creates a learning curve for our Logistic Regression classifier. The shaded bands around each curve show ±1 standard deviation across the cross-validation folds at each training size. Wide bands indicate that performance is unstable across folds, meaning the model's behaviour depends heavily on which specific examples it happened to be trained on.

> 📓 **Cross-validation** is covered in Notebook 10. Briefly: rather than evaluating on a single fixed validation set, $k$-fold cross-validation splits the training data into $k$ equal parts, trains on $k-1$ of them, and evaluates on the remaining one, repeating this $k$ times so every part serves as the validation set exactly once. The scores are then averaged. This gives a more reliable estimate of generalisation performance than a single split, particularly when the dataset is small.


````python
# Listing 27.
# learning_curve() trains and evaluates the model at a series of increasing training set sizes.
# Rather than a single train/test split, it uses cross-validation at each size so the
# reported scores are averaged over multiple splits — giving a more reliable estimate
# than any single evaluation would.
#
# train_sizes=np.linspace(0.1, 1.0, 10) specifies ten evenly spaced fractions of the
# training set: 10%, 20%, 30%, ... 100%. At each fraction, the model is trained on that
# proportion of X_tr_s and evaluated via cv=5 cross-validation.
#
# The function returns three arrays:
#   train_sizes:  the actual number of training examples used at each step (not fractions)
#   train_scores: a 2D array of shape (n_sizes, n_folds) — training accuracy at each size and fold
#   val_scores:   a 2D array of shape (n_sizes, n_folds) — validation accuracy at each size and fold
train_sizes, train_scores, val_scores = learning_curve(
    LogisticRegression(max_iter=1000, random_state=0),
    X_tr_s, y_tr,
    train_sizes=np.linspace(0.1, 1.0, 10),
    cv=5, scoring='accuracy', n_jobs=-1   # n_jobs=-1 uses all available CPU cores in parallel
)

# Average the 5 fold scores at each training size to get a single mean accuracy per point.
# The standard deviation quantifies how much performance varied across folds —
# used below to draw the shaded uncertainty bands.
train_mean = train_scores.mean(axis=1)
train_std  = train_scores.std(axis=1)
val_mean   = val_scores.mean(axis=1)
val_std    = val_scores.std(axis=1)

fig, ax = plt.subplots(figsize=(9, 5))

# Plot training accuracy — typically starts high (model fits small training sets easily)
# and decreases slightly as the training set grows and becomes harder to fit perfectly.
ax.plot(train_sizes, train_mean, 'o-', color='steelblue', lw=2, label='Train accuracy')

# Shaded band shows ±1 standard deviation across folds.
# A wide band means performance was unstable across folds at that training size.
ax.fill_between(train_sizes, train_mean - train_std, train_mean + train_std,
                alpha=0.15, color='steelblue')

# Plot cross-validation accuracy — typically starts low and rises as training set grows.
# The gap between this line and the training accuracy line is the key diagnostic signal.
ax.plot(train_sizes, val_mean, 's-', color='tomato', lw=2, label='Cross-val accuracy')
ax.fill_between(train_sizes, val_mean - val_std, val_mean + val_std,
                alpha=0.15, color='tomato')

ax.set_xlabel('Training set size')
ax.set_ylabel('Accuracy')
ax.set_title('Figure 22: Learning curve — Logistic Regression\n'
             'Gap between curves = variance; low val accuracy = bias')
ax.legend(fontsize=10)

# y-axis lower limit set to 0.8 to zoom into the relevant range —
# differences in the 0.85–1.0 range are far more informative than the full 0–1 scale
ax.set_ylim(0.8, 1.02)

plt.tight_layout()
plt.show()
````

## 12. Feature Selection

🔙 [Back to Table of Contents](#Table-of-Contents)

With the visualisation toolkit established, we can now apply it directly to one of the most practically important decisions in any ML project: **which features should we include in our model?**

This question matters more than it might initially seem. It is tempting to assume that giving a model more information is always better, but in practice, including irrelevant or redundant features actively harms performance. Every additional feature adds dimensions to the space the model must search, increases the risk of fitting noise, and makes the model harder to interpret and maintain.

Feature selection matters for three interconnected reasons:

1. **Accuracy**: irrelevant features add noise and redundant features introduce multicollinearity, both of which can degrade model performance. Distance-based models such as k-NN are particularly sensitive to this. Recall the *curse of dimensionality* from Notebook 4: as the number of features grows, the volume of the feature space grows exponentially, data becomes increasingly sparse, and distance metrics lose their meaning.

2. **Speed**: fewer features means smaller input matrices, faster training, and faster prediction. At production scale, where a model may be called millions of times per day, this difference is significant.

3. **Interpretability**: a model built on 5 well-chosen features is far easier to explain to a stakeholder, audit for bias, or debug when it fails than one built on 30. Simpler models are also easier to trust.

Feature selection methods fall into four broad families, each with different trade-offs:

| Family | How it works | Speed | Captures non-linear relationships? |
|--------|-------------|-------|-------------------------------------|
| **Visual / statistical** | Inspect distributions and compute effect sizes manually | Very fast | No |
| **Filter** | Score each feature independently of any model using a statistical criterion | Fast | Mutual information does; Pearson $r$ does not |
| **Wrapper** | Repeatedly train a model on different feature subsets and compare performance | Slow | Yes, depends on the choice of model |
| **Embedded** | Feature importance is computed as a by-product of the model's own training process | Medium | Yes, particularly tree-based methods |

In practice, a good feature selection workflow combines methods from more than one family: start with fast visual and filter methods to eliminate obvious weak features, then apply wrapper or embedded methods to fine-tune the final set.

This section works through three approaches in order of increasing sophistication: visual inspection with Cohen's $d$, filter-based selection using mutual information, and embedded selection using a Random Forest.

### 12.1 Visual Inspection and Cohen's d

Before applying any formal statistical method, visual inspection of class-separated distributions is always the first step, as we did extensively in Sections 6 through 10. The plots we have already produced give a strong qualitative sense of which features separate the classes well. **Cohen's $d$** converts that visual impression into a single number.

Cohen's $d$ measures the separation between two group means in units of the pooled standard deviation, that is, how many standard deviations apart the two class distributions are:

$$
d = \frac{|\mu_{\text{Benign}} - \mu_{\text{Malignant}}|}{\sqrt{(\sigma_{\text{Benign}}^2 + \sigma_{\text{Malignant}}^2) / 2}}
$$

The denominator is the **pooled standard deviation**, a weighted average of the spread within each class. Dividing by it puts the mean difference on a standardised scale that is independent of the original units of measurement, making $d$ values comparable across features that operate on completely different scales.

Conventional interpretation guidelines:

| $d$ value | Effect size |
|-----------|-------------|
| $d < 0.2$ | Negligible: the distributions almost completely overlap |
| $d \approx 0.5$ | Medium: moderate separation between classes |
| $d \geq 0.8$ | Large: clear separation between classes |
| $d \geq 1.5$ | Very large: strong candidate for inclusion in a classifier |

> 💡 **Cohen's $d$ and the KDE plots:** if you look back at the KDE plots in Section 6, features where the two curves sit clearly apart will have high $d$ values, and features where the curves heavily overlap will have low $d$ values. Cohen's $d$ is simply a way of quantifying what those plots show visually.

The cell below computes Cohen's $d$ for every `mean` feature and visualises the results as a ranked bar chart, making it straightforward to identify the strongest candidate features at a glance.


````python
# Listing 28.
# Compute Cohen's d for every mean feature — quantifying the separation between
# the Benign and Malignant distributions for each feature independently.
benign    = df[df['diagnosis'] == 'Benign'][mean_cols]
malignant = df[df['diagnosis'] == 'Malignant'][mean_cols]

def cohens_d(a, b):
    # Compute the pooled standard deviation — a weighted average of the within-class spread.
    # This is the denominator in the Cohen's d formula and puts the mean difference
    # on a scale that is independent of the original units of measurement.
    pooled_std = np.sqrt((a.std()**2 + b.std()**2) / 2)
    return (a.mean() - b.mean()).abs() / pooled_std

d_scores = cohens_d(benign, malignant).sort_values(ascending=False)

# --- Bar chart: Cohen's d ranked by effect size ---
# Displaying features in descending order of d makes it immediately clear
# which features have the strongest class separation.
fig, ax = plt.subplots(figsize=(9, 5))

ax.barh(d_scores.index[::-1], d_scores.values[::-1],
        color='steelblue', edgecolor='white', linewidth=0.4)

# Reference lines at conventional Cohen's d thresholds help the reader
# interpret the magnitude of each bar without needing to read exact values.
ax.axvline(0.8, color='tomato',   linewidth=1.2, linestyle='--', label='Large effect  (d = 0.8)')
ax.axvline(1.5, color='darkred',  linewidth=1.2, linestyle='--', label='Very large    (d = 1.5)')

ax.set_xlabel("Cohen's d  (pooled-SD units)")
ax.set_title("Figure 23: Cohen's d — class separation per feature")
ax.legend(fontsize=9)

plt.tight_layout()
plt.show()

print("Features with d ≥ 1.5 are strong candidates for inclusion in a classifier.")
print(d_scores[d_scores >= 1.5].to_string())
````

### 12.2 Filter Method 1: Pearson Correlation

We have already used Pearson $r$ to examine the relationship between individual features and the target in Section 9. Here we apply the same idea as a formal **filter method**, a feature selection technique that scores each feature independently of any model and ranks them by that score.

The Pearson filter is the simplest filter method available. It requires no model, involves no training, and runs almost instantaneously even on large datasets, making it a practical and low-cost first pass over a feature set of any size. Features that score highly are worth prioritising; features that score near zero can often be safely discarded before any modelling begins.

The score used here is the **absolute Pearson correlation** $|r|$ between each feature and the target variable. Taking the absolute value means we treat strong negative correlations (higher feature value → more likely Malignant) as equally informative as strong positive ones (higher feature value → more likely Benign), which is correct, since the sign of the relationship does not affect how useful a feature is to a classifier.

In the cell below we apply this to all 30 features in the dataset, not just the `mean` features, to produce a complete ranking. Features above the $|r| = 0.5$ threshold are highlighted as strong linear candidates.

The key limitation of this approach is fundamental rather than incidental: **Pearson $r$ only detects linear associations**. A feature could have a strong quadratic relationship with the target, or a threshold effect where values above a certain point are highly predictive but values below it are not, and still produce an $|r|$ close to zero. Such a feature would be incorrectly discarded by this method. The next section addresses this limitation directly using mutual information, which captures non-linear associations of any shape.


````python
# Listing 30.
# Select all feature columns, excluding the numeric target and the string diagnosis label.
# This gives us all 30 features rather than just the 'mean' subset used earlier.
all_features = [c for c in df.columns if c not in ['target', 'diagnosis']]
X_all = df[all_features]
y_ser = df['target']

# .corrwith(y_ser) computes the Pearson correlation of each column in X_all
# with y_ser in a single vectorised call — equivalent to calling np.corrcoef()
# on each column individually but far more concise.
# .abs() takes the absolute value so positive and negative correlations are
# treated as equally informative — the sign of the relationship does not affect
# how useful a feature is to a classifier.
# .sort_values() ranks from strongest to weakest.
pearson_r = X_all.corrwith(y_ser).abs().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(10, 8))
fig.canvas.header_visible = False
fig.canvas.resizable = True
fig.canvas.toolbar_visible = True
fig.canvas.toolbar_position = 'right'

# Features above the 0.5 threshold are drawn in full steelblue to highlight them;
# weaker features are drawn in a lighter shade to de-emphasise them visually.
colours_p = ['steelblue' if v >= 0.5 else 'lightsteelblue' for v in pearson_r]

# [::-1] reverses the order so the strongest feature appears at the top of the chart.
# The colours list is reversed in the same step to keep colours aligned with their bars.
ax.barh(pearson_r.index[::-1], pearson_r.values[::-1],
        color=colours_p[::-1], edgecolor='white', linewidth=0.3)

# Annotate each bar with its |r| value, placed just beyond the end of the bar.
# Iterating over the reversed Series keeps the annotation order aligned with the bars.
for feature, val in zip(pearson_r.index[::-1], pearson_r.values[::-1]):
    ax.text(val + 0.005,                  # Slightly to the right of the bar tip
            feature,                       # y position matches the bar's category label
            f'{val:.3f}',                  # Display to 3 decimal places
            va='center', ha='left',
            fontsize=7.5, color='#333333')

# A reference line at |r| = 0.5 provides a visual cut-off separating strong
# from weak linear associations — not a hard rule, but a useful working threshold.
ax.axvline(0.5, color='red', linewidth=1.5, linestyle='--',
           label='|r| = 0.5 threshold')

# Extend the x-axis limit slightly to give the annotations room beyond the longest bar
ax.set_xlim(0, pearson_r.max() + 0.08)

ax.set_xlabel('|Pearson r| with target')
ax.set_title('Figure 25: Filter method — all 30 features ranked by\n'
             '|Pearson correlation| with target')
ax.legend(fontsize=9)

plt.tight_layout()
plt.show()

# Print a summary count and a reminder of the method's core limitation
n_above = (pearson_r >= 0.5).sum()
print(f'Features with |r| >= 0.5: {n_above} / {len(pearson_r)}')
print()
print('Limitation: Pearson r only captures LINEAR associations.')
print('A feature could be highly informative but non-linearly related')
print('to the target — and would be missed by this method.')
````

### 12.3 Filter Method 2: Mutual Information

**Mutual Information (MI)** addresses the core limitation of the Pearson filter directly. Where Pearson $r$ only detects linear associations, MI captures *any* statistical dependence between a feature and the target, whether linear, non-linear, or otherwise.

> 📓 **Notebook 6 recap:** when we introduced decision trees, we covered **entropy** and **information gain** in detail. Mutual information is built on exactly the same ideas. If those concepts feel unfamiliar, it is worth revisiting Section 7 of Notebook 6 before continuing, as the intuition developed there transfers directly here.

Recall that entropy $H(Y)$ measures the amount of uncertainty in a variable $Y$. For a binary class label it is maximised at 1 bit when both classes are equally probable, and equals 0 when one class occurs with certainty:

$$H(Y) = -\sum_{c=1}^{k} P(Y = c) \log_2 P(Y = c)$$

Mutual information uses this idea to measure how much a feature $X$ reduces our uncertainty about the target $Y$:

$$I(X; Y) = H(Y) - H(Y \mid X)$$

where $H(Y)$ is the entropy of the target before observing $X$, and $H(Y \mid X)$ is the **conditional entropy**, the uncertainty that remains in $Y$ after observing $X$. The difference between them is the reduction in uncertainty that knowing $X$ provides.

This should look familiar: it is the same logic as information gain in a decision tree. In that context, a split was evaluated by how much it reduced the entropy of the child nodes relative to the parent. MI applies the same principle to the relationship between a single feature and the target label across the entire dataset.

Two special cases make the interpretation concrete:

- **$I(X; Y) = H(Y)$:** knowing $X$ eliminates all uncertainty about $Y$, so the feature is perfectly informative.
- **$I(X; Y) = 0$:** knowing $X$ tells us nothing about $Y$, so the feature is completely uninformative.

Because MI makes no assumption about the form of the relationship between $X$ and $Y$, it will assign a high score to a feature that has a strong threshold effect, a quadratic relationship, or any other non-linear pattern, cases where Pearson $r$ would produce a value close to zero.

> ⚠️ **One practical limitation:** MI values are not bounded between 0 and 1 the way Pearson $r$ is. The maximum value depends on the entropy of the target, which varies between datasets. This means MI scores are useful for *ranking* features within a dataset but are not directly comparable across different datasets. Scikit-learn's `mutual_info_classif()` returns scores in nats (using natural logarithm rather than $\log_2$), so the absolute values will differ from the entropy calculations in Notebook 6 which used bits, but the ranking is what matters here.

The cell below applies mutual information scoring to all 30 features and ranks them in a single bar chart. Every feature is scored and sorted from strongest to weakest, with features above the working threshold of 0.2 highlighted in blue and weaker features shown in a lighter shade.

Compare this ranking with the Pearson ranking from the previous section. For features with a strong linear relationship with the target, the two rankings will be broadly similar. Differences are where mutual information earns its place as a separate method: a feature that ranks much higher here than it did in the Pearson chart has a non-linear relationship with the target that Pearson $r$ was missing entirely but MI is detecting.

````python
# Listing 31.
# mutual_info_classif() estimates the mutual information between each feature
# and the target variable using a k-nearest-neighbour density estimator
# (the Kraskov estimator). This approach makes no assumption about the form
# of the relationship — linear, quadratic, threshold effects, and any other
# non-linear pattern are all captured in the same score.
# random_state= fixes the seed used internally by the estimator so results
# are reproducible across runs.
X_np = X_all.values
y_np = y_ser.values

mi_scores = mutual_info_classif(X_np, y_np, random_state=0)

# Wrap the scores in a Series indexed by feature name so we can sort and
# align with the Pearson results computed in the previous section.
mi_series = pd.Series(mi_scores, index=all_features).sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(9, 8))
fig.canvas.header_visible = False
fig.canvas.resizable      = True
fig.canvas.toolbar_visible = False

# Features above the MI = 0.2 threshold are highlighted in full steelblue;
# weaker features are shown in a lighter shade to de-emphasise them.
# The threshold of 0.2 is a working heuristic — adjust based on the dataset.
colours_mi = ['steelblue' if v >= 0.2 else 'lightsteelblue'
              for v in mi_series.values[::-1]]

# [::-1] reverses the order so the strongest feature appears at the top of the chart
ax.barh(mi_series.index[::-1], mi_series.values[::-1],
        color=colours_mi, edgecolor='white', linewidth=0.3)

# Annotate each bar with its MI score, placed just beyond the end of the bar
for feature, val in zip(mi_series.index[::-1], mi_series.values[::-1]):
    ax.text(val + 0.003,
            feature,
            f'{val:.3f}',
            va='center', ha='left',
            fontsize=7.5, color='#333333')

ax.axvline(0.2, color='red', linewidth=1.5, linestyle='--',
           label='MI = 0.2 threshold')

# Extend the x-axis limit to give the bar annotations room beyond the longest bar
ax.set_xlim(0, mi_series.max() + 0.08)

ax.set_xlabel('Mutual Information score')
ax.set_title('Figure 26: Mutual Information scores — all 30 features\n'
             'Captures linear and non-linear associations with the target')
ax.legend(fontsize=9)

plt.tight_layout()
plt.show()

# The top 10 features by MI provide a ranked shortlist for the next stage
# of feature selection or direct use as model inputs.
print('Top 10 features by Mutual Information:')
print(mi_series.head(10).round(4).to_string())
````

The Pearson filter and the MI filter are complementary rather than competing. A feature that scores highly on both is a confident inclusion. A feature that scores low on Pearson but high on MI is worth keeping, since it carries non-linear signal that a classifier can exploit. A feature that scores low on both is a strong candidate for removal. Running both filters takes seconds and together they cover the full range of univariate associations between features and the target.


### 12.4 Wrapper Method: Recursive Feature Elimination (RFE)

Filter methods score features *independently* of each other and independently of any model. This is their strength, speed and simplicity, but also their limitation. A feature that appears weak in isolation can become highly useful in combination with another, and a filter method will never detect this. Two features that are individually strong but highly correlated with each other may together add less value than either one paired with a weaker but complementary feature. Filter methods cannot see any of this.

**Wrapper methods** address this by evaluating feature *subsets* rather than individual features. Instead of computing a score per feature, a wrapper method trains an actual model on a candidate subset, measures its predictive performance on held-out data, and uses that performance as the score for the subset. Feature interactions are accounted for automatically because the model sees all selected features simultaneously during training.

The trade-off is computational cost. Where a filter method runs in milliseconds, a wrapper method trains and evaluates a model once for every subset it considers, which can be expensive when the number of features is large.

**Recursive Feature Elimination (RFE)** is the most widely used wrapper approach. Rather than evaluating every possible subset, which would be computationally intractable, it uses a principled greedy strategy to work backwards from the full feature set:

1. Train the model on all available features.
2. Rank features by their importance to the model. For a logistic regression, this is the absolute magnitude of the learned coefficients after scaling.
3. Remove the least important feature.
4. Repeat steps 1–3 until the desired number of features remains.

By running this process for each possible number of features from 1 to $n$ and measuring 5-fold cross-validated accuracy at each step, we can plot an **accuracy vs feature count curve**. This curve answers the practical question directly: *how many features do we actually need, and which ones are they?*

The point where accuracy plateaus is the sweet spot: adding more features beyond this point produces negligible improvement, while removing features below it causes a meaningful drop in performance.

**Feature scaling is essential here.** RFE with Logistic Regression ranks features by coefficient magnitude, which is only a fair comparison when all features are on the same scale. A feature with large raw values will produce a small coefficient even if it is genuinely informative, and vice versa. `StandardScaler` is applied to all 30 features before RFE begins, consistent with the scaling step used when training the classifier in Section 11.

The cell below runs RFE across feature counts from 1 to 15, plots the accuracy curve with a shaded ±1 standard deviation band across folds, and then refits using the optimal number of features to print the final selected set.

````python
# Listing 32.
# StandardScaler transforms each feature to zero mean and unit variance.
# This is essential before RFE with Logistic Regression because the model
# ranks features by coefficient magnitude — a fair comparison only when all
# features are on the same scale. Without scaling, features with large raw
# values produce small coefficients and would be incorrectly ranked as unimportant.
# fit_transform() on the full feature matrix is acceptable here because we are
# performing feature selection on the full dataset, not evaluating generalisation.
scaler_rfe = StandardScaler()
X_scaled   = scaler_rfe.fit_transform(X_all)

# We evaluate every feature count from 1 to 15 rather than jumping straight to
# a fixed number — this lets us plot the full accuracy curve and identify the
# point where adding more features stops producing meaningful improvement.
n_feature_range = list(range(1, 16))
cv_means, cv_stds = [], []

for n_feat in n_feature_range:
    # RFE(estimator, n_features_to_select) wraps the estimator in a transformer
    # that internally eliminates features one by one until n_features_to_select remain.
    # Passing the RFE object directly to cross_val_score means the full elimination
    # process is repeated independently inside each fold — preventing data leakage
    # from the feature ranking step.
    rfe = RFE(LogisticRegression(max_iter=1000, random_state=0),
              n_features_to_select=n_feat)

    # cross_val_score() splits X_scaled into 5 folds, trains and evaluates the
    # RFE-wrapped model on each, and returns one accuracy score per fold.
    scores = cross_val_score(rfe, X_scaled, y_np, cv=5, scoring='accuracy')

    cv_means.append(scores.mean())   # Average accuracy across the 5 folds
    cv_stds.append(scores.std())     # Spread across folds — wider = less stable

cv_means = np.array(cv_means)
cv_stds  = np.array(cv_stds)

fig, ax = plt.subplots(figsize=(10, 5))
fig.canvas.header_visible = False
fig.canvas.resizable = True
fig.canvas.toolbar_visible = False

# The accuracy curve — each point is the mean 5-fold accuracy for that feature count
ax.plot(n_feature_range, cv_means, 'o-', color='steelblue', lw=2, ms=7,
        label='Mean cross-val accuracy')

# Shaded band shows ±1 standard deviation across folds.
# A wide band at low feature counts is normal — small feature sets are more
# sensitive to which specific examples appear in each fold.
ax.fill_between(n_feature_range, cv_means - cv_stds, cv_means + cv_stds,
                alpha=0.15, color='steelblue')

# Identify and mark the feature count that produced the highest mean accuracy.
# .argmax() returns the index of the maximum value in cv_means;
# indexing into n_feature_range converts that back to an actual feature count.
best_n = n_feature_range[cv_means.argmax()]
ax.axvline(best_n, color='red', linewidth=2, linestyle='--',
           label=f'Best: {best_n} features  (acc = {cv_means.max():.3f})')

ax.set_xlabel('Number of features kept')
ax.set_ylabel('5-fold cross-validation accuracy')
ax.set_title('Figure 27: RFE — cross-validated accuracy vs number of features kept\n'
             'Wrapper method: directly optimises model performance')
ax.legend(fontsize=10)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Refit RFE on the full dataset using the optimal feature count identified above.
# This final fit determines which specific features survive the elimination process.
rfe_best = RFE(LogisticRegression(max_iter=1000, random_state=0),
               n_features_to_select=best_n)
rfe_best.fit(X_scaled, y_np)

# rfe_best.support_ is a boolean array — True for features that were retained,
# False for those that were eliminated. Zipping it with the feature names lets
# us extract the names of the surviving features in a single list comprehension.
selected = [f for f, s in zip(all_features, rfe_best.support_) if s]

print(f'RFE selected {best_n} features:')
for f in selected:
    print(f'  {f}')
````

### 12.5 Embedded Method 1: Lasso (L1 Regularisation)

**Embedded methods** take a fundamentally different approach to feature selection from the filter and wrapper methods above. Rather than selecting features before training, or repeatedly retraining on different subsets, embedded methods build feature selection directly into the model's training objective: selection and learning happen simultaneously.

The **Lasso** (Least Absolute Shrinkage and Selection Operator) achieves this by adding an $\ell_1$ penalty on the model's coefficients to the loss function. For logistic regression, this gives:

$$\text{Loss}_{\text{Lasso}} = \text{NLL} + \frac{1}{C} \sum_{j=1}^{p} |\beta_j|$$

where NLL is the negative log-likelihood (the standard logistic regression loss), $\beta_j$ are the model coefficients, and $C$ is the regularisation parameter. The penalty term charges the model for every unit of coefficient magnitude it uses, forcing it to justify the inclusion of each feature by the improvement it brings to the loss.

**Why does this drive coefficients to exactly zero?** The answer is geometric. The $\ell_1$ penalty constrains the sum of absolute coefficient values to lie within a diamond-shaped region in coefficient space. This shape has sharp corners aligned with the axes. When the optimisation finds the point where the loss surface touches the constraint boundary, it frequently lands on one of these corners, where one or more coefficients are exactly zero. This is in contrast to the $\ell_2$ (Ridge) penalty, whose circular constraint region has no corners, so the optimal solution is almost never exactly zero.

The parameter $C$ controls the strength of the penalty:

| $C$ value | Effect |
|-----------|--------|
| Small (e.g. 0.001) | Strong penalty: only the most important features survive |
| Large (e.g. 5.0) | Weak penalty: most features are retained, behaviour approaches unregularised logistic regression |

By sweeping $C$ across a range of values and tracking how many features survive and what test accuracy is achieved, we can read off the penalty level that gives the best trade-off between model simplicity and predictive performance.

---

**Embedded vs wrapper:** both RFE and Lasso produce a ranked or filtered feature set, but Lasso does so in a single training run rather than the repeated retraining that RFE requires. For large datasets or high-dimensional feature spaces this difference in speed is significant. The cost is that Lasso is specific to linear models: it cannot be applied to a decision tree or k-NN classifier in the same way.

The cell below launches an interactive Lasso explorer. A slider controls the regularisation parameter $C$ across a log-spaced range from 0.001 to 10. Move it left to increase the penalty and force more coefficients to zero, move it right to relax the penalty and allow more features to survive.

Three panels update live as the slider moves:

- **Left, surviving features:** a horizontal bar chart showing only the features with non-zero coefficients at the current $C$, sorted by absolute magnitude. Blue bars indicate positive coefficients (associated with the Benign class); red bars indicate negative coefficients (associated with Malignant). Watch features disappear from this panel as the penalty strengthens; their disappearance is the selection signal.
- **Centre, regularisation path:** the full curve of surviving feature count across the entire $C$ range, with a red marker tracking your current position. This gives context for where you are on the sparsification curve.
- **Right, accuracy path:** test accuracy across the full $C$ range with a red marker. Reading the centre and right panels together reveals the practical sweet spot: the region where the feature count is substantially reduced but accuracy has not yet fallen.

> 💡 **What to look for:** find the lowest $C$ value at which accuracy remains competitive. The features visible in the left panel at that point are the ones the Lasso considers most important, a compact, model-derived feature set that balances simplicity with predictive performance.

### 12.6 Embedded Method 2: Random Forest Importances

> 📓 **Notebook 6 recap:** Random Forests were introduced in detail in Section 7 of Notebook 6, covering bootstrap sampling, random feature selection, out-of-bag scoring, and majority vote prediction. If any of those concepts feel unfamiliar, it is worth revisiting that section before continuing, as the feature importance mechanism described here builds directly on how the forest is trained.

Recall that at each node in each tree, the Random Forest selects the feature and split threshold that maximally reduces impurity, as measured by the Gini index. This selection happens thousands of times across all trees and all bootstrap samples. A feature that is genuinely useful will be selected frequently and will produce large impurity reductions when it is; a feature that is irrelevant will rarely be selected and will produce small reductions when it is.

The **mean decrease in impurity (MDI)**, also called the feature importance, formalises this intuition. For each feature $j$, it averages the impurity reduction produced by all splits that used $j$, across all trees in the forest:

$$\text{importance}(j) = \frac{1}{T} \sum_{t=1}^{T} \sum_{\text{splits on } j \text{ in tree } t} \Delta \text{Gini}_t$$

where $T$ is the number of trees, and $\Delta \text{Gini}_t$ is the reduction in Gini impurity achieved at a split that used feature $j$ in tree $t$. Scikit-learn normalises the final values so they sum to 1 across all features, making them straightforward to interpret as relative contributions.

This is a fundamentally different measure of importance from anything the filter methods could produce:

- **It is non-linear**: because it is computed from tree splits, it captures threshold effects, interactions between features, and any other non-linear pattern that contributes to reducing impurity. Pearson $r$ would miss all of this.
- **It is stable**: because importances are averaged over all trees and all bootstrap samples rather than computed from a single model fit, they are far less sensitive to the particular training sample than the root split of a single decision tree.
- **It is free**: no additional computation is needed beyond training the forest itself. Feature importances are a by-product of the training process, accessible via `rf.feature_importances_` after calling `fit()`.

> ⚠️ **One known limitation:** MDI importances can overestimate the importance of high-cardinality features, that is, features that take many distinct values, because such features offer more possible split points and therefore more opportunities to be selected. For datasets with a mix of continuous and categorical features with very different cardinalities, permutation importance is a more reliable alternative. For our dataset, where all 30 features are continuous measurements on comparable scales, MDI is appropriate.

The cell below trains a Random Forest on all 30 features and visualises the resulting importances as a ranked bar chart, comparable in format to the Pearson and MI charts produced earlier in this section.

````python
# Listing 34.
# Build a Random Forest on the same scaled features used for the Lasso above,
# so the three embedded methods (RFE, Lasso, RF) are directly comparable.
# n_estimators=200 — 200 trees gives stable importance estimates; the variance
#   in importances across runs decreases as more trees are added.
# n_jobs=-1 — trees are independent of each other so training parallelises
#   cleanly across all available CPU cores, with no effect on the result.
# random_state= — fixes the bootstrap sampling and feature selection seeds
#   so results are reproducible across runs.
rf = RandomForestClassifier(n_estimators=200, random_state=0, n_jobs=-1)
rf.fit(X_scaled, y_np)

# rf.feature_importances_ is a 1D array of length n_features.
# Each value is the mean decrease in Gini impurity produced by splits on that
# feature, averaged across all trees and all nodes — normalised to sum to 1.
# Wrapping in a Series indexed by feature name makes sorting and plotting cleaner.
importances = pd.Series(rf.feature_importances_, index=all_features)

# Sort ascending so the most important feature appears at the top of the
# horizontal bar chart (barh plots the last entry at the top by default).
importances = importances.sort_values(ascending=True)

# Highlight the top 30% of features by importance — those above the 70th
# percentile — in full steelblue; the remaining 70% are shown in a lighter
# shade to direct the reader's attention to the most informative features.
colours_rf = ['steelblue' if v >= importances.quantile(0.7)
              else 'lightsteelblue' for v in importances]

fig, ax = plt.subplots(figsize=(10, 9))
fig.canvas.header_visible = False
fig.canvas.resizable = True
fig.canvas.toolbar_visible = True
fig.canvas.toolbar_position = 'right'

ax.barh(importances.index, importances.values,
        color=colours_rf, edgecolor='white', linewidth=0.3)

# Annotate each bar with its importance score, placed just beyond the bar tip
for feature, val in zip(importances.index, importances.values):
    ax.text(val + 0.001,
            feature,
            f'{val:.4f}',
            va='center', ha='left',
            fontsize=7.5, color='#333333')

# Extend the x-axis to give annotation labels room beyond the longest bar
ax.set_xlim(0, importances.max() + 0.03)

ax.set_xlabel('Mean decrease in impurity  (normalised, sums to 1)')
ax.set_title('Figure 29: Random Forest feature importances — 200 trees\n'
             'Captures non-linear relationships — compare with Pearson r and MI above')
ax.grid(True, alpha=0.25, axis='x')

plt.tight_layout()
plt.show()

# Print the top 10 features as a numerical summary.
# .tail(10) takes the 10 largest values from the ascending-sorted Series;
# [::-1] reverses the order so the most important feature is listed first.
print('Top 10 features by Random Forest importance:')
print(importances.tail(10)[::-1].round(4).to_string())
````
